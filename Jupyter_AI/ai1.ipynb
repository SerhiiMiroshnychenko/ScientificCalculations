```python
# Приклад спискового генератора: створення списку квадратів чисел від 0 до 9
squares = [x**2 for x in range(10)]
print(squares)

# Ще один приклад: створення списку парних чисел від 0 до 9, піднесених до квадрату
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)

# Приклад з рядками: створення списку довжин слів
words = ["apple", "banana", "cherry", "date"]
word_lengths = [len(word) for word in words]
print(word_lengths)
```

```python
# Приклад словникового генератора: створення словника, де ключі - числа, а значення - їх квадрати
squares_dict = {x: x**2 for x in range(10)}
print(squares_dict)

# Ще один приклад: створення словника з довжинами слів
words = ["apple", "banana", "cherry"]
word_lengths_dict = {word: len(word) for word in words}
print(word_lengths_dict)

# Приклад з умовою: створення словника тільки для слів, довжина яких більша за 5
long_words_dict = {word: len(word) for word in words if len(word) > 5}
print(long_words_dict)
```
