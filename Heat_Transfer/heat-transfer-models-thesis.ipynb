# ПОРІВНЯЛЬНИЙ АНАЛІЗ ОДНОВИМІРНОЇ ТА ДВОВИМІРНОЇ МОДЕЛЕЙ ТЕПЛОПРОВІДНОСТІ ПРИ МОДЕЛЮВАННІ ПРОЦЕСІВ ТЕПЛОПЕРЕДАЧІ

## Анотація

В даній роботі проведено порівняльний аналіз одновимірної та двовимірної моделей теплопровідності при моделюванні процесів теплопередачі в різних фізичних системах. Досліджено математичні формулювання моделей, розглянуто основні методи їх чисельного розв'язання, проаналізовано точність та ефективність обчислень, а також визначено раціональні області застосування кожної з моделей. На основі проведеного аналізу встановлено кореляцію між геометричними особливостями досліджуваних об'єктів та доцільністю використання певної моделі теплопровідності.

## Вступ

Моделювання процесів теплопередачі є важливим інструментом при проектуванні та оптимізації різноманітних технічних систем, зокрема теплообмінників, систем охолодження, будівельних конструкцій тощо. Точність моделювання цих процесів безпосередньо впливає на ефективність розроблюваних систем та їх відповідність експлуатаційним вимогам. При цьому вибір відповідної моделі теплопровідності — одновимірної чи двовимірної — має ґрунтуватися на раціональному балансі між точністю отримуваних результатів та обчислювальними витратами.

## Теоретичні основи моделей теплопровідності

Одновимірна модель теплопровідності базується на припущенні, що температурне поле змінюється лише вздовж однієї просторової координати. Математично вона описується рівнянням [1]:

$$\rho c_p \frac{\partial T}{\partial t} = \frac{\partial}{\partial x} \left( \lambda \frac{\partial T}{\partial x} \right) + q_v$$

де $\rho$ — густина матеріалу, $c_p$ — питома теплоємність, $T$ — температура, $t$ — час, $\lambda$ — коефіцієнт теплопровідності, $x$ — просторова координата, $q_v$ — об'ємна густина джерела тепла.

Двовимірна модель теплопровідності враховує зміну температури у двох просторових координатах і описується рівнянням [2]:

$$\rho c_p \frac{\partial T}{\partial t} = \frac{\partial}{\partial x} \left( \lambda_x \frac{\partial T}{\partial x} \right) + \frac{\partial}{\partial y} \left( \lambda_y \frac{\partial T}{\partial y} \right) + q_v$$

де $\lambda_x$ та $\lambda_y$ — коефіцієнти теплопровідності в напрямках $x$ та $y$ відповідно.

Для стаціонарних задач, коли температурне поле не змінюється з часом, рівняння спрощуються до:

Одновимірний випадок:
$$\frac{d}{dx} \left( \lambda \frac{dT}{dx} \right) + q_v = 0$$

Двовимірний випадок:
$$\frac{\partial}{\partial x} \left( \lambda_x \frac{\partial T}{\partial x} \right) + \frac{\partial}{\partial y} \left( \lambda_y \frac{\partial T}{\partial y} \right) + q_v = 0$$

## Методи чисельного розв'язання

Для розв'язання рівнянь теплопровідності застосовуються різні чисельні методи, зокрема метод скінченних різниць, метод скінченних елементів та метод граничних елементів. В таблиці 1 наведено порівняння обчислювальної складності цих методів для одновимірної та двовимірної моделей [3].

**Таблиця 1. Порівняння обчислювальної складності методів розв'язання**

| Метод | Одновимірна модель | Двовимірна модель |
|-------|-------------------|-------------------|
| Метод скінченних різниць | $O(n)$ | $O(n^2)$ |
| Метод скінченних елементів | $O(n \log n)$ | $O(n^2 \log n)$ |
| Метод граничних елементів | $O(n^2)$ | $O(n^3)$ |

де $n$ — кількість вузлів дискретизації в одному напрямку.

При використанні методу скінченних різниць для одновимірної стаціонарної задачі дискретизація приводить до системи лінійних рівнянь [4]:

$$\frac{\lambda_{i+1/2}}{\Delta x^2}(T_{i+1} - T_i) - \frac{\lambda_{i-1/2}}{\Delta x^2}(T_i - T_{i-1}) + q_{vi} = 0$$

Для двовимірної моделі аналогічна система має вигляд [5]:

$$\frac{\lambda_{x,i+1/2,j}}{\Delta x^2}(T_{i+1,j} - T_{i,j}) - \frac{\lambda_{x,i-1/2,j}}{\Delta x^2}(T_{i,j} - T_{i-1,j}) + \frac{\lambda_{y,i,j+1/2}}{\Delta y^2}(T_{i,j+1} - T_{i,j}) - \frac{\lambda_{y,i,j-1/2}}{\Delta y^2}(T_{i,j} - T_{i,j-1}) + q_{vi,j} = 0$$

## Порівняльний аналіз моделей

Проведено порівняльний аналіз одновимірної та двовимірної моделей теплопровідності за кількома ключовими параметрами, результати якого наведено в таблиці 2.

**Таблиця 2. Порівняння одновимірної та двовимірної моделей теплопровідності**

| Параметр | Одновимірна модель | Двовимірна модель |
|----------|-------------------|-------------------|
| Точність моделювання для тіл зі співвідношенням розмірів > 10:1 | Висока | Надмірна |
| Точність моделювання для тіл зі співвідношенням розмірів < 10:1 | Недостатня | Висока |
| Обчислювальні витрати | Низькі | Середні до високих |
| Складність задання граничних умов | Низька | Середня |
| Можливість моделювання анізотропних матеріалів | Обмежена | Широка |
| Час розрахунку (відносний) | 1 | 10-100 |

Аналіз показує, що одновимірна модель є ефективною для систем з вираженою одновимірною геометрією, таких як стержні, труби з тонкими стінками, та теплоізоляційні шари. Математично це відповідає випадкам, коли градієнт температури в одному напрямку значно перевищує градієнти в інших напрямках, що можна виразити як:

$$\left| \frac{\partial T}{\partial x} \right| \gg \left| \frac{\partial T}{\partial y} \right|, \left| \frac{\partial T}{\partial z} \right|$$

Двовимірна модель є необхідною для систем з вираженою двовимірною геометрією, таких як пластини, оребрені поверхні, та поперечні перерізи протяжних конструкцій. Особливо важливо застосовувати двовимірну модель у випадках наявності теплових містків, концентраторів теплового потоку, та при складній конфігурації граничних умов.

Експериментальні дослідження показують, що похибка одновимірної моделі порівняно з двовимірною для об'єктів зі співвідношенням розмірів менше 5:1 може перевищувати 15%, що є неприйнятним для багатьох інженерних задач.

## Аналіз ефективності обчислень

Дослідження ефективності обчислень показало, що час розрахунку для двовимірної моделі зростає квадратично відносно кількості вузлів дискретизації в одному напрямку, тоді як для одновимірної моделі зростання є лінійним. Це означає, що при однаковій щільності сітки дискретизації двовимірна модель потребує значно більших обчислювальних ресурсів.

При проведенні параметричних досліджень або оптимізації, що вимагають багаторазових розрахунків з різними вхідними параметрами, оптимальним підходом є використання одновимірної моделі на етапі попереднього аналізу з подальшим уточненням результатів за допомогою двовимірної моделі для найбільш перспективних варіантів.

Для нестаціонарних задач різниця в обчислювальних витратах стає ще більш значущою, оскільки розрахунок необхідно проводити для кожного кроку за часом. Відношення часу обчислень може бути оцінено як:

$$\frac{t_{2D}}{t_{1D}} \approx \frac{n_y \cdot \alpha_{2D}}{\alpha_{1D}}$$

де $n_y$ — кількість вузлів дискретизації в напрямку $y$, а $\alpha_{2D}$ та $\alpha_{1D}$ — коефіцієнти, що характеризують ефективність алгоритму розв'язання систем лінійних рівнянь для двовимірної та одновимірної моделей відповідно.

## Практичні рекомендації щодо вибору моделі

На основі проведеного аналізу можна сформулювати наступні практичні рекомендації щодо вибору моделі теплопровідності:

Одновимірну модель доцільно застосовувати у таких випадках:
- Для попередніх оціночних розрахунків на етапі концептуального проектування
- Для систем з вираженою одновимірною геометрією (співвідношення характерних розмірів більше 10:1)
- При проведенні параметричних досліджень, що вимагають багаторазових розрахунків
- У випадках, коли граничні умови не змінюються вздовж неосновних координат
- При моделюванні процесів теплопередачі в композитних шаруватих структурах з переважним напрямком теплового потоку перпендикулярно до шарів

Двовимірну модель слід використовувати у таких випадках:
- Для систем з порівнянними характерними розмірами у двох напрямках
- При наявності неоднорідних граничних умов
- У випадках значної анізотропії теплофізичних властивостей матеріалів
- При моделюванні теплових містків та концентраторів теплового потоку
- У задачах оптимізації форми конструкції для мінімізації теплових втрат

## Висновки

Проведений порівняльний аналіз одновимірної та двовимірної моделей теплопровідності показав, що вибір моделі має ґрунтуватися на балансі між точністю результатів та обчислювальними витратами з урахуванням геометричних особливостей досліджуваного об'єкта та характеру теплових процесів.

Одновимірна модель забезпечує достатню точність для об'єктів з вираженою одновимірною геометрією при суттєво менших обчислювальних витратах, що робить її ефективною для попередніх та параметричних досліджень. Двовимірна модель є необхідною для точного моделювання об'єктів зі складною геометрією та неоднорідними граничними умовами, проте потребує значно більших обчислювальних ресурсів.

Раціональний підхід до моделювання процесів теплопередачі полягає у комбінуванні обох моделей: використання одновимірної моделі на етапі попереднього аналізу з подальшим уточненням результатів за допомогою двовимірної моделі для найбільш критичних або складних ділянок конструкції.

Подальші дослідження мають бути спрямовані на розробку адаптивних алгоритмів, які автоматично обирають оптимальну модель для різних ділянок досліджуваного об'єкта, а також на вдосконалення методів скорочення обчислювальних витрат при використанні двовимірної моделі.
