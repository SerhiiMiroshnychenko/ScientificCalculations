## Вхідні дані

Згідно з варіантом 11, розглядаємо крок 11 (час 110 хв) з наступними параметрами:

### Склад газової суміші (об'ємні %)
- CO: 8.8%
- CO₂: 91.0%
- O₂: 0.1%
- N₂: 0.1%

### Температура
- T = 110°C


## Крок 1: Перетворення об'ємних відсотків у частки

### Формула
$$r_i = \frac{V_i}{V_{total}} = \frac{\%_i}{100\%}$$

де:
- $r_i$ - об'ємна частка i-го компонента
- $\%_i$ - об'ємний відсоток i-го компонента

```python
# Перетворення відсотків у частки
r_co = 73.5 / 100
r_co2 = 25 / 100
r_o2 = 0.9 / 100
r_n2 = 0.6 / 100

print(f"Об'ємна частка CO: {r_co:.3f}")
print(f"Об'ємна частка CO₂: {r_co2}")
print(f"Об'ємна частка O₂: {r_o2}")
print(f"Об'ємна частка N₂: {r_n2}")
print(f"Сума часток: {r_co + r_co2 + r_o2 + r_n2}")
```

### Результат
$$r_{CO} = \frac{8.8\%}{100\%} = 0.088$$
$$r_{CO_2} = \frac{91.0\%}{100\%} = 0.91$$
$$r_{O_2} = \frac{0.1\%}{100\%} = 0.001$$
$$r_{N_2} = \frac{0.1\%}{100\%} = 0.001$$
$$r_{CO} + r_{CO_2} + r_{O_2} + r_{N_2} = 0.088 + 0.91 + 0.001 + 0.001 = 1.0$$

## Крок 2: Перетворення температури з градусів Цельсія в Кельвіни

### Формула
$$T(K) = T(°C) + 273.15$$

де:
- $T(K)$ - температура в Кельвінах
- $T(°C)$ - температура в градусах Цельсія

```python
# Перетворення температури в Кельвіни
T_celsius = 141
T_kelvin = T_celsius + 273.15

print(f"Температура в градусах Цельсія: {T_celsius}°C")
print(f"Температура в Кельвінах: {T_kelvin} K")
```

### Результат
$$T(K) = 110°C + 273.15 = 383.15 K$$

## Крок 3: Розрахунок мольних теплоємностей компонентів при сталому тиску

### Формула
$$c_{p,i} = a_i + b_i \cdot T + \frac{c_i}{T^2}$$

де:
- $c_{p,i}$ - мольна теплоємність i-го компонента при сталому тиску, кал/(моль·К)
- $a_i$, $b_i$, $c_i$ - емпіричні коефіцієнти для i-го компонента
- $T$ - температура, К

```python
# Коефіцієнти для газів (в кал/(моль·К))
# CO
a_co = 6.79
b_co = 0.98e-3  # Коефіцієнт b·10³ з таблиці
c_co = -0.11*1e5  # Множимо на 1e5 згідно з таблицею

# CO₂
a_co2 = 10.55
b_co2 = 2.16e-3
c_co2 = -2.04*1e5

# O₂
a_o2 = 7.16
b_o2 = 1.0e-3
c_o2 = -0.4*1e5

# N₂
a_n2 = 6.66
b_n2 = 1.02e-3
c_n2 = 0  # для N₂ немає коефіцієнта c

# Розрахунок мольних теплоємностей при сталому тиску (cp) в кал/(моль·К)
cp_co_cal = a_co + b_co*T_kelvin + c_co/(T_kelvin*T_kelvin)
cp_co2_cal = a_co2 + b_co2*T_kelvin + c_co2/(T_kelvin*T_kelvin)
cp_o2_cal = a_o2 + b_o2*T_kelvin + c_o2/(T_kelvin*T_kelvin)
cp_n2_cal = a_n2 + b_n2*T_kelvin + c_n2/(T_kelvin*T_kelvin)

print(f"Мольна теплоємність CO: {cp_co_cal:.4f} кал/(моль·К)")
print(f"Мольна теплоємність CO₂: {cp_co2_cal:.4f} кал/(моль·К)")
print(f"Мольна теплоємність O₂: {cp_o2_cal:.4f} кал/(моль·К)")
print(f"Мольна теплоємність N₂: {cp_n2_cal:.4f} кал/(моль·К)")
```

### Результат

Для CO:
$$c_{p,CO} = 6.79 + 0.98 \cdot 10^{-3} \cdot 383.15 + \frac{-0.11 \cdot 10^5}{383.15^2} = 6.79 + 0.3755 - 0.0749 = 7.0906 \text{ кал/(моль·К)}$$

Для CO₂:
$$c_{p,CO_2} = 10.55 + 2.16 \cdot 10^{-3} \cdot 383.15 + \frac{-2.04 \cdot 10^5}{383.15^2} = 10.55 + 0.8276 - 1.3896 = 9.9880 \text{ кал/(моль·К)}$$

Для O₂:
$$c_{p,O_2} = 7.16 + 1.0 \cdot 10^{-3} \cdot 383.15 + \frac{-0.4 \cdot 10^5}{383.15^2} = 7.16 + 0.3832 - 0.2725 = 7.2707 \text{ кал/(моль·К)}$$

Для N₂:
$$c_{p,N_2} = 6.66 + 1.02 \cdot 10^{-3} \cdot 383.15 + \frac{0}{383.15^2} = 6.66 + 0.3908 + 0 = 7.0508 \text{ кал/(моль·К)}$$

## Крок 4: Перетворення мольних теплоємностей з кал/(моль·К) в Дж/(моль·К)

### Формула
$$c_{p,i}(\text{Дж/(моль·К)}) = c_{p,i}(\text{кал/(моль·К)}) \cdot 4.1868$$

де:
- $c_{p,i}(\text{Дж/(моль·К)})$ - мольна теплоємність в Дж/(моль·К)
- $c_{p,i}(\text{кал/(моль·К)})$ - мольна теплоємність в кал/(моль·К)
- $4.1868$ - коефіцієнт перетворення калорій в джоулі

```python
# Коефіцієнт перетворення кал в Дж
CAL_TO_JOULE = 4.1868

# Переведення в Дж/(моль·К)
cp_co = cp_co_cal * CAL_TO_JOULE
cp_co2 = cp_co2_cal * CAL_TO_JOULE
cp_o2 = cp_o2_cal * CAL_TO_JOULE
cp_n2 = cp_n2_cal * CAL_TO_JOULE

print(f"Мольна теплоємність CO: {cp_co:.4f} Дж/(моль·К)")
print(f"Мольна теплоємність CO₂: {cp_co2:.4f} Дж/(моль·К)")
print(f"Мольна теплоємність O₂: {cp_o2:.4f} Дж/(моль·К)")
print(f"Мольна теплоємність N₂: {cp_n2:.4f} Дж/(моль·К)")
```

### Результат

$$c_{p,CO}(\text{Дж/(моль·К)}) = 7.0906 \text{ кал/(моль·К)} \cdot 4.1868 = 29.6867 \text{ Дж/(моль·К)}$$
$$c_{p,CO_2}(\text{Дж/(моль·К)}) = 9.9880 \text{ кал/(моль·К)} \cdot 4.1868 = 41.8177 \text{ Дж/(моль·К)}$$
$$c_{p,O_2}(\text{Дж/(моль·К)}) = 7.2707 \text{ кал/(моль·К)} \cdot 4.1868 = 30.4409 \text{ Дж/(моль·К)}$$
$$c_{p,N_2}(\text{Дж/(моль·К)}) = 7.0508 \text{ кал/(моль·К)} \cdot 4.1868 = 29.5203 \text{ Дж/(моль·К)}$$

## Крок 5: Розрахунок середньої мольної теплоємності суміші

### Формула
$$c_{p,mix} = \sum_{i=1}^n r_i \cdot c_{p,i}$$

де:
- $c_{p,mix}$ - середня мольна теплоємність суміші, Дж/(моль·К)
- $r_i$ - об'ємна частка i-го компонента
- $c_{p,i}$ - мольна теплоємність i-го компонента, Дж/(моль·К)
- $n$ - кількість компонентів

```python
# Розрахунок середньої мольної теплоємності суміші
cp_mix = r_co * cp_co + r_co2 * cp_co2 + r_o2 * cp_o2 + r_n2 * cp_n2

print(f"Середня мольна теплоємність суміші: {cp_mix:.4f} Дж/(моль·К)")
```

### Результат

$$c_{p,mix} = 0.088 \cdot 29.6867 + 0.91 \cdot 41.8177 + 0.001 \cdot 30.4409 + 0.001 \cdot 29.5203$$
$$c_{p,mix} = 2.6124 + 38.0541 + 0.0304 + 0.0295 = 40.7265 \text{ Дж/(моль·К)}$$

## Крок 6: Розрахунок мольного об'єму газу при заданій температурі і тиску

### Формула
$$V_m = \frac{R \cdot T}{P}$$

де:
- $V_m$ - мольний об'єм газу, м³/моль
- $R$ - універсальна газова стала, 8.31446 Дж/(моль·К)
- $T$ - температура, К
- $P$ - тиск, Па (приймаємо стандартний атмосферний тиск 101325 Па)

```python
# Константи
R = 8.31446261815  # Дж/(моль·К)
P_STANDARD = 101325  # Па (1 атм)

# Розрахунок мольного об'єму
V_m = R * T_kelvin / P_STANDARD

print(f"Мольний об'єм газу при {T_celsius}°C і стандартному тиску: {V_m:.8f} м³/моль")
```

### Результат

$$V_m = \frac{8.31446 \cdot 383.15}{101325} = \frac{3185.7}{101325} = 0.03144 \text{ м³/моль}$$

## Крок 7: Розрахунок об'ємної теплоємності суміші

### Формула
$$c_{p,v} = \frac{c_{p,mix}}{V_m}$$

де:
- $c_{p,v}$ - об'ємна теплоємність суміші, Дж/(м³·К)
- $c_{p,mix}$ - середня мольна теплоємність суміші, Дж/(моль·К)
- $V_m$ - мольний об'єм газу, м³/моль

```python
# Розрахунок об'ємної теплоємності
cp_vol = cp_mix / V_m

# Переведення в кДж/(м³·К)
cp_vol_kj = cp_vol / 1000

print(f"Об'ємна теплоємність суміші: {cp_vol:.2f} Дж/(м³·К)")
print(f"Об'ємна теплоємність суміші: {cp_vol_kj:.4f} кДж/(м³·К)")
```

### Результат

$$c_{p,v} = \frac{40.7265 \text{ Дж/(моль·К)}}{0.03144 \text{ м³/моль}} = 1295.36 \text{ Дж/(м³·К)} = 1.2954 \text{ кДж/(м³·К)}$$

Середня об'ємна теплоємність газової суміші для варіанту 11 (крок 11, час 110 хв) складає 1.2954 кДж/(м³·К).
