# Задача 8: Розрахунок середньої питомої теплоємності міксерного шлаку

## Умова задачі
Міксерний шлак має склад, мас. %:
- СаO – 45
- SіO₂ – 40
- Al₂O₃ – 15

Обчислити середню питому теплоємність міксерного шлаку в інтервалі температур від 0 до 500 °С:
1. За правилом адитивності
2. З урахуванням температурної залежності

## Теоретичні відомості

При дослідженні термодинамічних властивостей сумішей матеріалів, зокрема міксерного шлаку, важливо правильно оцінити їх питомі теплоємності. Для цього застосовуються різні методи розрахунку, що враховують як прості адитивні властивості компонентів, так і складні температурні залежності, які виникають через взаємодію компонентів у суміші. У цьому звіті розглядаються два підходи:
1. **Метод адитивності**, який базується на принципі лінійної комбінації теплоємностей компонентів з урахуванням їх масових часток.
2. **Метод з урахуванням температурної залежності**, який використовує експериментально визначені залежності теплоємності суміші від температури, що враховують складні фізико-хімічні процеси в матеріалі.

---

## 1. Метод адитивності

### 1.1. Визначення та принцип

Метод адитивності ґрунтується на припущенні, що сумарна питома теплоємність суміші визначається як зважена сума теплоємностей її окремих компонентів. Якщо $x_i$ – масова частка $i$-го компонента, а $c_i$ – його питома теплоємність, тоді середня питома теплоємність суміші $c$ обчислюється за формулою:

$$
c = \sum_{i=1}^n x_i \cdot c_i
$$

Цей підхід передбачає, що компоненти не взаємодіють між собою, а їх внески в загальну теплоємність не змінюються при змішуванні.

### 1.2. Переваги та обмеження

**Переваги:**
- **Простота розрахунків:** Метод дозволяє швидко та просто оцінити теплоємність суміші, використовуючи відомі значення теплоємностей окремих компонентів.
- **Лінійність:** У разі відсутності взаємодії між компонентами розрахунки за адитивним принципом дають достатньо точні результати.

**Обмеження:**
- **Ігнорування взаємодій:** У реальних системах компоненти можуть вступати в хімічні чи фізичні взаємодії, що змінюють їхні термодинамічні властивості.
- **Температурні ефекти:** Адитивний метод не враховує зміну теплоємності компонентів зі зміною температури, що може бути критичним у високотемпературних процесах.
- **Утворення нових фаз:** При змішуванні матеріалів можуть утворюватися нові сполуки, що не враховані в простій лінійній моделі.

---

## 2. Метод з урахуванням температурної залежності

### 2.1. Визначення та принцип

Метод з урахуванням температурної залежності передбачає використання емпіричних або теоретичних поліноміальних залежностей теплоємності від температури для конкретних матеріалів або сумішей. Загальна форма такої залежності може мати вигляд:

$$
\bar{c}(T) = a + bT + cT^2 + dT^3,
$$

де $a$, $b$, $c$ та $d$ – коефіцієнти, визначені на основі експериментальних даних, а $T$ – температура. Ця формула дозволяє враховувати нелінійний характер зміни теплоємності з температурою, що особливо актуально для систем, де відбуваються структурні зміни або утворюються нові хімічні сполуки.

### 2.2. Переваги та обмеження

**Переваги:**
- **Вища точність:** Метод дозволяє більш точно врахувати вплив температури на теплоємність, що є важливим у високотемпературних умовах.
- **Врахування взаємодій:** Емпіричні залежності можуть відображати вплив взаємодій між компонентами, структурних змін та утворення нових фаз.
- **Ширше застосування:** Підхід дозволяє моделювати поведінку матеріалу в широких температурних інтервалах, що робить його корисним для практичних технологічних процесів.

**Обмеження:**
- **Складність визначення коефіцієнтів:** Отримання точних коефіцієнтів потребує широкого набору експериментальних даних та аналізу.
- **Обмеженість діапазону:** Зазвичай емпіричні залежності розробляються для конкретного температурного інтервалу, і їх застосування поза цим діапазоном може бути некоректним.
- **Залежність від умов експерименту:** Отримані залежності можуть варіюватися в залежності від умов проведення експерименту та характеристик досліджуваного матеріалу.

---

## 3. Порівняльний аналіз

### 3.1. Теоретичні основи

- **Метод адитивності:**  
  Ґрунтується на припущенні лінійної комбінації властивостей окремих компонентів без врахування їх взаємодії. Застосовується за умови, що компоненти зберігають свої властивості незалежно від того, чи знаходяться вони в чистій формі, чи в суміші.

- **Метод з урахуванням температурної залежності:**  
  Базується на емпіричних або теоретичних моделях, що описують зміну теплоємності в залежності від температури. Такий підхід враховує нелінійний характер цієї залежності, а також можливі зміни в структурі матеріалу при нагріванні.

### 3.2. Практичне застосування

- **Приклад застосування адитивного методу:**  
  Для розрахунку питомої теплоємності міксерного шлаку за простим правилом адитивності використовуються масові частки та теплоємності окремих компонентів. Результати можуть бути прийнятними при відсутності значних взаємодій між компонентами, проте в умовах високих температур чи при змішуванні матеріалів з різною структурою отримані значення можуть бути заниженими або завищеними.

- **Приклад застосування методу температурної залежності:**  
  При використанні температурної залежності враховується вплив температури на кожен з компонентів або на суміш в цілому. Це дозволяє отримати більш реалістичну оцінку теплоємності в умовах практичних технологічних процесів, де температура може суттєво змінювати фізико-хімічні властивості матеріалу.

### 3.3. Аналіз точності та застосовності

- **Точність розрахунків:**  
  Методи, що враховують температурну залежність, зазвичай дають більш точні результати в умовах значних температурних змін. Адитивний метод може бути спрощеним і не враховувати деяких ефектів, таких як зміна структури чи взаємодія компонентів, що може призводити до систематичних похибок.

- **Комплексність підходів:**  
  Адитивний метод є більш простим та інтуїтивно зрозумілим, тому його часто застосовують для попередніх оцінок. Метод температурної залежності, хоч і вимагає більше даних та розрахункових зусиль, забезпечує більш глибоке розуміння термодинамічної поведінки системи.

- **Вибір методу:**  
  Вибір методу залежить від конкретної задачі. Якщо необхідно швидко оцінити теплоємність суміші за умов, де взаємодії можна знехтувати, адитивний підхід є доцільним. У випадках, коли система піддається значним температурним коливанням або коли взаємодії між компонентами мають критичне значення, доцільніше застосовувати метод з урахуванням температурної залежності.

## Висновки

1. **Метод адитивності** – простий і зручний підхід, який дозволяє розрахувати питомі теплоємності сумішей шляхом лінійної комбінації властивостей окремих компонентів. Проте цей метод не враховує можливих змін теплоємності внаслідок взаємодій між компонентами чи структурних змін при нагріванні.

2. **Метод з урахуванням температурної залежності** – дозволяє точніше відобразити вплив температури на термодинамічні властивості сумішей, враховуючи нелінійність зміни теплоємності та можливі зміни фізико-хімічного стану матеріалу. Хоча цей метод потребує додаткових експериментальних даних та складніших розрахунків, він є більш адекватним для умов високотемпературних технологічних процесів.

3. **Порівняльний аналіз** показує, що для систем, де взаємодії між компонентами відіграють значну роль, метод температурної залежності забезпечує кращу відповідність реальним умовам. Проте у випадках, коли такі ефекти можна знехтувати, адитивний підхід може бути використаний як спрощена модель для первинної оцінки.

Таким чином, вибір методу залежить від конкретних умов дослідження та вимог до точності розрахунків, а комбіноване застосування обох підходів може надати більш комплексне розуміння термодинамічних процесів у складних сумішах.

### Температурні залежності теплоємностей

Для міксерного шлаку (0-1360 °С):
$$ \bar{c} = 0,694 + 8,95 \cdot 10^{-4}T - 1,18 \cdot 10^{-6}T^2 + 5,72 \cdot 10^{-10}T^3, \text{ кДж/(кг·°С)} $$

Для компонентів:
1. СаО (0-800 °С):
$$ \bar{c} = 0,749 + 3,78 \cdot 10^{-4}T - 1,533 \cdot 10^{-7}T^2, \text{ кДж/(кг·°С)} $$

2. SіО₂ (0-575 °С):
$$ \bar{c} = 0,794 + 9,4 \cdot 10^{-4}T - 7,15 \cdot 10^{-7}T^2, \text{ кДж/(кг·°С)} $$

3. Al₂O₃ (0-500 °С):
$$ \bar{c} = 0,786 + 5,97 \cdot 10^{-4}T - 2,98 \cdot 10^{-7}T^2, \text{ кДж/(кг·°С)} $$

### Метод адитивності
$$ c = \sum_{i=1}^n x_i c_i $$

де:
- $c$ - середня питома теплоємність суміші
- $x_i$ - масова частка i-го компонента
- $c_i$ - середня питома теплоємність i-го компонента

## Розв'язання

### 1. Метод адитивності

#### 1.1 Переведення масових часток у десятковий вигляд
- СаО: $x_1 = 45\% = 0,45$
- SіО₂: $x_2 = 40\% = 0,40$
- Al₂O₃: $x_3 = 15\% = 0,15$

#### 1.2 Розрахунок середніх теплоємностей компонентів при 500°С

Для СаО:
$$ \bar{c}_1 = 0,749 + 3,78 \cdot 10^{-4} \cdot (500+273.15) - 1,533 \cdot 10^{-7} \cdot (500+273.15)^2 = 0,9496 \text{ кДж/(кг·°С)} $$

Для SіО₂:
$$ \bar{c}_2 = 0,794 + 9,4 \cdot 10^{-4} \cdot (500+273.15) - 7,15 \cdot 10^{-7} \cdot (500+273.15)^2 = 1,0934 \text{ кДж/(кг·°С)} $$

Для Al₂O₃:
$$ \bar{c}_3 = 0,786 + 5,97 \cdot 10^{-4} \cdot (500+273.15) - 2,98 \cdot 10^{-7} \cdot (500+273.15)^2 = 1,0694 \text{ кДж/(кг·°С)} $$

#### 1.3 Розрахунок середньої теплоємності шлаку за правилом адитивності

Внески компонентів:
- СаО: $0,45 \cdot 0,9496 = 0,4273 \text{ кДж/(кг·°С)}$
- SіО₂: $0,40 \cdot 1,0934 = 0,4373 \text{ кДж/(кг·°С)}$
- Al₂O₃: $0,15 \cdot 1,0694 = 0,1604 \text{ кДж/(кг·°С)}$

Сумарна теплоємність:
$$ c_{ад} = 0,4273 + 0,4373 + 0,1604 = 1,0251 \text{ кДж/(кг·°С)} $$

### 2. Метод з урахуванням температурної залежності

Підставляємо T = 500°С у формулу для шлаку:

$$ c_{зал} = 0,694 + 8,95 \cdot 10^{-4} \cdot (500+273.15) - 1,18 \cdot 10^{-6} \cdot (500+273.15)^2 + 5,72 \cdot 10^{-10} \cdot (500+273.15)^3 = 0,945 \text{ кДж/(кг·°С)} $$

### 3. Порівняння результатів

1. За правилом адитивності: 1,0251 кДж/(кг·°С)
2. З урахуванням температурної залежності: 0,945 кДж/(кг·°С)
3. Абсолютна різниця: 0,08012 кДж/(кг·°С)
4. Відносна різниця: 8.48%

## Відповідь
1. Середня питома теплоємність міксерного шлаку за правилом адитивності: 1,0251 кДж/(кг·°С)
2. Середня питома теплоємність міксерного шлаку з урахуванням температурної залежності: 0,945 кДж/(кг·°С)
3. Різниця між методами становить 8.48%

## Аналіз результатів
1. Метод адитивності дає більше значення теплоємності, ніж метод з урахуванням температурної залежності
2. Різниця між методами досить значна (8.48%), що може бути пов'язано з:
   - Взаємодією компонентів у шлаку
   - Утворенням нових хімічних сполук
   - Структурними змінами при нагріванні

## Перевірка розмірностей
- Масові частки: [безрозмірні]
- Теплоємності: [кДж/(кг·°С)]
- Температура: [°С]
- Результат: [кДж/(кг·°С)] ✓

```python
# Масові частки компонентів, %
x_cao = 45    # CaO
x_sio2 = 40   # SiO2
x_al2o3 = 15  # Al2O3

# Температура, К
T = 500+273.15

print("1. Вхідні дані:")
print(f"   Температура: {T} К")
print("\n   Масові частки компонентів, %:")
print(f"   CaO: {x_cao}")
print(f"   SiO2: {x_sio2}")
print(f"   Al2O3: {x_al2o3}")
```

```python
# Переведення масових часток у десятковий вигляд
x_cao_dec = x_cao/100
x_sio2_dec = x_sio2/100
x_al2o3_dec = x_al2o3/100

print("\n2. Масові частки у десятковому вигляді:")
print(f"   CaO: {x_cao_dec:.4f}")
print(f"   SiO2: {x_sio2_dec:.4f}")
print(f"   Al2O3: {x_al2o3_dec:.4f}")
```

```python
# Перевірка суми масових часток
x_sum = x_cao_dec + x_sio2_dec + x_al2o3_dec
print(f"\n3. Сума масових часток: {x_sum:.4f}")
```

```python
# Функції для розрахунку теплоємності компонентів
def c_cao(T):
    """Теплоємність CaO, кДж/(кг·°С)"""
    return 0.749 + 3.78e-4*T - 1.533e-7*T**2

def c_sio2(T):
    """Теплоємність SiO2, кДж/(кг·°С)"""
    return 0.794 + 9.4e-4*T - 7.15e-7*T**2

def c_al2o3(T):
    """Теплоємність Al2O3, кДж/(кг·°С)"""
    return 0.786 + 5.97e-4*T - 2.98e-7*T**2

def c_slag(T):
    """Теплоємність шлаку, кДж/(кг·°С)"""
    return 0.694 + 8.95e-4*T - 1.18e-6*T**2 + 5.72e-10*T**3
```

```python
# Розрахунок теплоємностей компонентів при заданій температурі
c_cao_T = c_cao(T)
c_sio2_T = c_sio2(T)
c_al2o3_T = c_al2o3(T)

print("\n4. Теплоємності компонентів при {T} К, кДж/(кг·°С):")
print(f"   CaO: {c_cao_T:.4f}")
print(f"   SiO2: {c_sio2_T:.4f}")
print(f"   Al2O3: {c_al2o3_T:.4f}")
```

```python
# Метод адитивності
c_additive = x_cao_dec * c_cao_T + x_sio2_dec * c_sio2_T + x_al2o3_dec * c_al2o3_T

print("\n5. Внески компонентів у теплоємність суміші (метод адитивності), кДж/(кг·°С):")
print(f"   CaO: {x_cao_dec * c_cao_T:.4f}")
print(f"   SiO2: {x_sio2_dec * c_sio2_T:.4f}")
print(f"   Al2O3: {x_al2o3_dec * c_al2o3_T:.4f}")
```

```python
print(f"\n6. Середня питома теплоємність (метод адитивності):")
print(f"   c = {c_additive:.4f} кДж/(кг·°С)")
```

```python
# Метод з урахуванням температурної залежності
c_temp = c_slag(T)

print(f"\n7. Середня питома теплоємність (з урахуванням температурної залежності):")
print(f"   c = {c_temp:.4f} кДж/(кг·°С)")
```

```python
# Порівняння методів
abs_diff = abs(c_additive - c_temp)
rel_diff = abs_diff/c_temp * 100

print("\n8. Порівняння методів:")
print(f"   Абсолютна різниця: {abs_diff:.4f} кДж/(кг·°С)")
print(f"   Відносна різниця: {rel_diff:.2f}%")
```

```python
print("\n9. Перевірка достовірності:")
print(f"   - Сума масових часток = 1: {'✓' if abs(x_sum - 1) < 1e-10 else '✗'}")
print(f"   - Розмірності збережені: ✓")
```
