## Методика розв'язання

### Крок 1: Розрахунок початкової кількості речовини

Формула для розрахунку кількості речовини

$$ n = \frac{PV}{RT} $$

де:
- $P$ - тиск газу, Па (101325 Па)
- $V$ - об'єм газу, м³ (100 м³)
- $R$ - універсальна газова стала, Дж/(моль·К) (8.31446261815 Дж/(моль·К))
- $T$ - температура газу, К (911°C = 1184.15 К)

```python
# Константи
P = 101325  # Па
V = 100  # м³
R = 8.31446261815  # Дж/(моль·К)
T_C_start = 734  # °C
T_K_start = T_C_start + 273.15  # К

# Розрахунок кількості речовини
n_start = (P * V) / (R * T_K_start)
print(f"Початкова кількість речовини: {n_start:.2f} моль")
```

Результат розрахунку

$$ n = \frac{101325 \cdot 100}{8.31446261815 \cdot 1184.15} = 1029.14 \text{ моль} $$

Початкова кількість речовини: 1029.14 моль

### Крок 2: Розрахунок кількості речовини кожного компонента

Формула для розрахунку кількості речовини компонентів

$$ n_i = r_i \cdot n $$

де:
- $n_i$ - кількість речовини i-го компонента, моль
- $r_i$ - об'ємна частка i-го компонента (в долях одиниці)
- $n$ - загальна кількість речовини, моль (1029.14 моль)

```python
# Об'ємні частки компонентів (початкові)
r_co_start = 25.9/100  # 12.5%
r_co2_start = 73.4/100  # 87.1%
r_o2_start = 0.4/100  # 0.2%
r_n2_start = 0.3/100  # 0.2%

# Розрахунок кількості речовини кожного компонента
n_co_start = r_co_start * n_start
n_co2_start = r_co2_start * n_start
n_o2_start = r_o2_start * n_start
n_n2_start = r_n2_start * n_start

print(f"Початкова кількість CO: {n_co_start:.2f} моль")
print(f"Початкова кількість CO₂: {n_co2_start:.2f} моль")
print(f"Початкова кількість O₂: {n_o2_start:.2f} моль")
print(f"Початкова кількість N₂: {n_n2_start:.2f} моль")
```

Результат розрахунку

$$ n_{CO} = 0.125 \cdot 1029.14 = 128.64 \text{ моль} $$
$$ n_{CO_2} = 0.871 \cdot 1029.14 = 896.38 \text{ моль} $$
$$ n_{O_2} = 0.002 \cdot 1029.14 = 2.06 \text{ моль} $$
$$ n_{N_2} = 0.002 \cdot 1029.14 = 2.06 \text{ моль} $$

Початкова кількість CO: 128.64 моль

Початкова кількість CO₂: 896.38 моль

Початкова кількість O₂: 2.06 моль

Початкова кількість N₂: 2.06 моль

### Крок 3: Розрахунок коефіцієнтів теплоємності

Формула для розрахунку теплоємності при постійному тиску

$$ c_p = a + bT + cT^{-2} \text{ кал/(моль·К)} $$

де:
- $a$, $b$, $c$ - коефіцієнти з таблиці
- $T$ - температура, К

Переведення в Дж/(моль·К):
$$ c_p[\text{Дж/(моль·К)}] = c_p[\text{кал/(моль·К)}] \cdot 4.1868 $$

Теплоємність при постійному об'ємі:
$$ c_v = c_p - R $$

де $R$ - універсальна газова стала, Дж/(моль·К) (8.31446261815 Дж/(моль·К))

```python
# Коефіцієнти теплоємності (кал/(моль·К))
# CO
a_co = 6.79
b_co = 0.98e-3
c_co = -0.11e5
# CO₂
a_co2 = 10.55
b_co2 = 2.16e-3
c_co2 = -2.04e5
# O₂
a_o2 = 7.16
b_o2 = 1.0e-3
c_o2 = -0.40e5
# N₂
a_n2 = 6.66
b_n2 = 1.02e-3
c_n2 = 0

# Коефіцієнт переведення з кал в Дж
CAL_TO_JOULE = 4.1868

# Температури для розрахунку (К)
T_C_end = 823  # °C
T_K_end = T_C_end + 273.15  # К

# Універсальна газова стала в кал/(моль·К)
R_cal = R / CAL_TO_JOULE

# Коефіцієнти для розрахунку теплоємності при постійному об'ємі
# CO
a_v_co = (a_co - R_cal) * CAL_TO_JOULE
b_v_co = b_co * CAL_TO_JOULE
c_v_co = c_co * CAL_TO_JOULE
# CO₂
a_v_co2 = (a_co2 - R_cal) * CAL_TO_JOULE
b_v_co2 = b_co2 * CAL_TO_JOULE
c_v_co2 = c_co2 * CAL_TO_JOULE
# O₂
a_v_o2 = (a_o2 - R_cal) * CAL_TO_JOULE
b_v_o2 = b_o2 * CAL_TO_JOULE
c_v_o2 = c_o2 * CAL_TO_JOULE
# N₂
a_v_n2 = (a_n2 - R_cal) * CAL_TO_JOULE
b_v_n2 = b_n2 * CAL_TO_JOULE
c_v_n2 = 0  # для N₂ коефіцієнт c = 0

print(f"Коефіцієнти теплоємності при постійному об'ємі:")
print(f"CO: a_v = {a_v_co:.2f}, b_v = {b_v_co:.5f}, c_v = {c_v_co:.2f}")
print(f"CO₂: a_v = {a_v_co2:.2f}, b_v = {b_v_co2:.5f}, c_v = {c_v_co2:.2f}")
print(f"O₂: a_v = {a_v_o2:.2f}, b_v = {b_v_o2:.5f}, c_v = {c_v_o2:.2f}")
print(f"N₂: a_v = {a_v_n2:.2f}, b_v = {b_v_n2:.5f}, c_v = {c_v_n2:.2f}")
```

Результат розрахунку

Коефіцієнти теплоємності при постійному тиску в Дж/(моль·К):
$$ c_p^{CO} = (6.79 + 0.98 \cdot 10^{-3} \cdot T - 0.11 \cdot 10^{5} \cdot T^{-2}) \cdot 4.1868 $$
$$ c_p^{CO_2} = (10.55 + 2.16 \cdot 10^{-3} \cdot T - 2.04 \cdot 10^{5} \cdot T^{-2}) \cdot 4.1868 $$
$$ c_p^{O_2} = (7.16 + 1.0 \cdot 10^{-3} \cdot T - 0.40 \cdot 10^{5} \cdot T^{-2}) \cdot 4.1868 $$
$$ c_p^{N_2} = (6.66 + 1.02 \cdot 10^{-3} \cdot T) \cdot 4.1868 $$

Коефіцієнти теплоємності при постійному об'ємі в Дж/(моль·К):
$$ c_v^{CO} = ((6.79 - 1.987) + 0.98 \cdot 10^{-3} \cdot T - 0.11 \cdot 10^{5} \cdot T^{-2}) \cdot 4.1868 $$
$$ c_v^{CO_2} = ((10.55 - 1.987) + 2.16 \cdot 10^{-3} \cdot T - 2.04 \cdot 10^{5} \cdot T^{-2}) \cdot 4.1868 $$
$$ c_v^{O_2} = ((7.16 - 1.987) + 1.0 \cdot 10^{-3} \cdot T - 0.40 \cdot 10^{5} \cdot T^{-2}) \cdot 4.1868 $$
$$ c_v^{N_2} = ((6.66 - 1.987) + 1.02 \cdot 10^{-3} \cdot T) \cdot 4.1868 $$

Де 1.987 кал/(моль·К) - це універсальна газова стала в кал/(моль·К).

### Крок 4: Розрахунок теплоти нагрівання при постійному об'ємі (ізохоричний процес)

Формула для розрахунку теплоти нагрівання

Для кожного компонента газової суміші теплота нагрівання при постійному об'ємі:

$$ Q_{v,i} = n_i \cdot [(a_i-R)(T_2 - T_1) + \frac{b_i}{2}(T_2^2 - T_1^2) - c_i(\frac{1}{T_2} - \frac{1}{T_1})] $$

де:
- $n_i$ - кількість речовини i-го компонента, моль
- $a_i$, $b_i$, $c_i$ - коефіцієнти теплоємності i-го компонента
- $R$ - універсальна газова стала, Дж/(моль·К)
- $T_1$ - початкова температура, К (911°C = 1184.15 К)
- $T_2$ - кінцева температура, К (1000°C = 1273.15 К)

Загальна теплота нагрівання суміші:

$$ Q_{v,сум} = Q_{v,CO} + Q_{v,CO_2} + Q_{v,O_2} + Q_{v,N_2} $$

```python
print(f"{n_co_start = }")
print(f"{T_K_end = }")
print(f"{T_K_start = }")
print(f"{c_v_co = }")
print(f"{b_v_co = }")
print(f"{a_v_co = }")
```

```python
# Розрахунок теплоти нагрівання при постійному об'ємі

# Функція для розрахунку теплоти нагрівання при постійному об'ємі
def calculate_heat_const_volume(n, a, b, c, T1, T2):
    return n * ((a * (T2 - T1)) + (b/2 * (T2**2 - T1**2)) - (c * (1/T2 - 1/T1)))

# Розрахунок теплоти нагрівання для кожного компонента
Q_v_co = calculate_heat_const_volume(n_co_start, a_v_co, b_v_co, c_v_co, T_K_start, T_K_end)
Q_v_co2 = calculate_heat_const_volume(n_co2_start, a_v_co2, b_v_co2, c_v_co2, T_K_start, T_K_end)
Q_v_o2 = calculate_heat_const_volume(n_o2_start, a_v_o2, b_v_o2, c_v_o2, T_K_start, T_K_end)
Q_v_n2 = calculate_heat_const_volume(n_n2_start, a_v_n2, b_v_n2, c_v_n2, T_K_start, T_K_end)

# Загальна теплота нагрівання суміші
Q_v_total = Q_v_co + Q_v_co2 + Q_v_o2 + Q_v_n2

print(f"Теплота нагрівання CO при постійному об'ємі: {Q_v_co/1000:.2f} кДж")
print(f"Теплота нагрівання CO₂ при постійному об'ємі: {Q_v_co2/1000:.2f} кДж")
print(f"Теплота нагрівання O₂ при постійному об'ємі: {Q_v_o2/1000:.2f} кДж")
print(f"Теплота нагрівання N₂ при постійному об'ємі: {Q_v_n2/1000:.2f} кДж")
print(f"Загальна теплота нагрівання при постійному об'ємі: {Q_v_total/1000:.2f} кДж")
```

Результат розрахунку

Для CO:
$$ Q_{v,CO} = 128.64 \cdot [(20.10 \cdot (1273.15 - 1184.15)) + (\frac{4.10 \cdot 10^{-3}}{2} \cdot (1273.15^2 - 1184.15^2)) - (-460.55 \cdot (\frac{1}{1273.15} - \frac{1}{1184.15}))] = 287.66 \text{ кДж} $$

Для CO₂:
$$ Q_{v,CO_2} = 896.38 \cdot [(35.86 \cdot (1273.15 - 1184.15)) + (\frac{9.04 \cdot 10^{-3}}{2} \cdot (1273.15^2 - 1184.15^2)) - (-854107.20 \cdot (\frac{1}{1273.15} - \frac{1}{1184.15}))] = 3701.79 \text{ кДж} $$

Для O₂:
$$ Q_{v,O_2} = 2.06 \cdot [(21.65 \cdot (1273.15 - 1184.15)) + (\frac{4.19 \cdot 10^{-3}}{2} \cdot (1273.15^2 - 1184.15^2)) - (-1674.72 \cdot (\frac{1}{1273.15} - \frac{1}{1184.15}))] = 4.89 \text{ кДж} $$

Для N₂:
$$ Q_{v,N_2} = 2.06 \cdot [(19.56 \cdot (1273.15 - 1184.15)) + (\frac{4.27 \cdot 10^{-3}}{2} \cdot (1273.15^2 - 1184.15^2)) - (0 \cdot (\frac{1}{1273.15} - \frac{1}{1184.15}))] = 4.55 \text{ кДж} $$

Загальна теплота нагрівання суміші при постійному об'ємі:
$$ Q_{v,сум} = 287.66 + 3701.79 + 4.89 + 4.55 = 3998.88 \text{ кДж} $$

Теплота нагрівання CO при постійному об'ємі: 287.66 кДж

Теплота нагрівання CO₂ при постійному об'ємі: 3701.79 кДж

Теплота нагрівання O₂ при постійному об'ємі: 4.89 кДж

Теплота нагрівання N₂ при постійному об'ємі: 4.55 кДж

Загальна теплота нагрівання при постійному об'ємі: 3998.88 кДж

### Крок 5: Розрахунок кінцевої кількості речовини для ізобаричного процесу

Формула для розрахунку кінцевої кількості речовини

Для ізобаричного процесу (при постійному тиску) закон ідеального газу можна записати як:

$$ \frac{n_1}{T_1} = \frac{n_2}{T_2} $$

Звідси кінцева кількість речовини:

$$ n_2 = n_1 \cdot \frac{T_1}{T_2} $$

де:
- $n_2$ - кінцева кількість речовини, моль
- $n_1$ - початкова кількість речовини, моль (1029.14 моль)
- $T_1$ - початкова температура, К (911°C = 1184.15 К)
- $T_2$ - кінцева температура, К (1000°C = 1273.15 К)

Кінцева кількість речовини кожного компонента визначається за кінцевими об'ємними частками:

$$ n_{i,2} = r_{i,2} \cdot n_2 $$

де $r_{i,2}$ - кінцева об'ємна частка i-го компонента.

```python
# Розрахунок кінцевої кількості речовини для ізобаричного процесу
n_end_isobaric = n_start * (T_K_start / T_K_end)

# Об'ємні частки компонентів (кінцеві)
r_co_end = 20.3/100  # 8.8%
r_co2_end = 79.2/100  # 91.0%
r_o2_end = 0.3/100  # 0.1%
r_n2_end = 0.2/100  # 0.1%

# Розрахунок кінцевої кількості речовини кожного компонента
n_co_end = r_co_end * n_end_isobaric
n_co2_end = r_co2_end * n_end_isobaric
n_o2_end = r_o2_end * n_end_isobaric
n_n2_end = r_n2_end * n_end_isobaric

print(f"Кінцева кількість речовини (ізобаричний процес): {n_end_isobaric:.2f} моль")
print(f"Кінцева кількість CO: {n_co_end:.2f} моль")
print(f"Кінцева кількість CO₂: {n_co2_end:.2f} моль")
print(f"Кінцева кількість O₂: {n_o2_end:.2f} моль")
print(f"Кінцева кількість N₂: {n_n2_end:.2f} моль")
```

Результат розрахунку

$$ n_2 = 1029.14 \cdot \frac{1184.15}{1273.15} = 957.20 \text{ моль} $$

$$ n_{CO,end} = 0.088 \cdot 957.20 = 84.23 \text{ моль} $$
$$ n_{CO_2,end} = 0.910 \cdot 957.20 = 871.05 \text{ моль} $$
$$ n_{O_2,end} = 0.001 \cdot 957.20 = 0.96 \text{ моль} $$
$$ n_{N_2,end} = 0.001 \cdot 957.20 = 0.96 \text{ моль} $$

Кінцева кількість речовини (ізобаричний процес): 957.20 моль

Кінцева кількість CO: 84.23 моль

Кінцева кількість CO₂: 871.05 моль

Кінцева кількість O₂: 0.96 моль

### Крок 6: Розрахунок теплоти нагрівання при постійному тиску (ізобаричний процес)

Формула для розрахунку теплоти нагрівання при постійному тиску

Для кожного компонента газової суміші теплота нагрівання при постійному тиску:

$$ Q_{p,i} = \bar{n}_i \cdot [a_i(T_2 - T_1) + \frac{b_i}{2}(T_2^2 - T_1^2) - c_i(\frac{1}{T_2} - \frac{1}{T_1})] $$

де:
- $\bar{n}_i$ - середня кількість речовини i-го компонента, моль
- $a_i$, $b_i$, $c_i$ - коефіцієнти теплоємності i-го компонента при постійному тиску
- $T_1$ - початкова температура, К (911°C = 1184.15 К)
- $T_2$ - кінцева температура, К (1000°C = 1273.15 К)

Середня кількість речовини i-го компонента:

$$ \bar{n}_i = \frac{n_{i,1} + n_{i,2}}{2} $$

де $n_{i,1}$ - початкова кількість речовини i-го компонента, $n_{i,2}$ - кінцева кількість речовини i-го компонента.

Загальна теплота нагрівання суміші:

$$ Q_{p,сум} = Q_{p,CO} + Q_{p,CO_2} + Q_{p,O_2} + Q_{p,N_2} $$

```python
# Розрахунок середньої кількості речовини для кожного компонента
n_co_avg = (n_co_start + n_co_end) / 2
n_co2_avg = (n_co2_start + n_co2_end) / 2
n_o2_avg = (n_o2_start + n_o2_end) / 2
n_n2_avg = (n_n2_start + n_n2_end) / 2

# Функція для розрахунку теплоти нагрівання при постійному тиску
def calculate_heat_const_pressure(n, a, b, c, T1, T2):
    return n * ((a * (T2 - T1)) + (b/2 * (T2**2 - T1**2)) - (c * (1/T2 - 1/T1)))

# Розрахунок теплоти нагрівання для кожного компонента
Q_p_co = calculate_heat_const_pressure(n_co_avg, a_co * CAL_TO_JOULE, b_co * CAL_TO_JOULE, c_co * CAL_TO_JOULE, T_K_start, T_K_end)
Q_p_co2 = calculate_heat_const_pressure(n_co2_avg, a_co2 * CAL_TO_JOULE, b_co2 * CAL_TO_JOULE, c_co2 * CAL_TO_JOULE, T_K_start, T_K_end)
Q_p_o2 = calculate_heat_const_pressure(n_o2_avg, a_o2 * CAL_TO_JOULE, b_o2 * CAL_TO_JOULE, c_o2 * CAL_TO_JOULE, T_K_start, T_K_end)
Q_p_n2 = calculate_heat_const_pressure(n_n2_avg, a_n2 * CAL_TO_JOULE, b_n2 * CAL_TO_JOULE, 0, T_K_start, T_K_end)

# Загальна теплота нагрівання суміші
Q_p_total = Q_p_co + Q_p_co2 + Q_p_o2 + Q_p_n2

print(f"Теплота нагрівання CO при постійному тиску: {Q_p_co/1000:.2f} кДж")
print(f"Теплота нагрівання CO₂ при постійному тиску: {Q_p_co2/1000:.2f} кДж")
print(f"Теплота нагрівання O₂ при постійному тиску: {Q_p_o2/1000:.2f} кДж")
print(f"Теплота нагрівання N₂ при постійному тиску: {Q_p_n2/1000:.2f} кДж")
print(f"Загальна теплота нагрівання при постійному тиску: {Q_p_total/1000:.2f} кДж")
```
