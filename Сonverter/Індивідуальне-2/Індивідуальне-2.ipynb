## Методика розв'язання

### Крок 1: Розрахунок початкової кількості речовини

Формула для розрахунку кількості речовини

$$ n = \frac{PV}{RT} $$

де:
- $P$ - тиск газу, Па (101325 Па)
- $V$ - об'єм газу, м³ (100 м³)
- $R$ - універсальна газова стала, Дж/(моль·К) (8.31446261815 Дж/(моль·К))
- $T$ - температура газу, К (911°C = 1184.15 К)

```python
# Константи
P = 101325  # Па
V = 100  # м³
R = 8.31446261815  # Дж/(моль·К)
T_C_start = 0  # °C
T_K_start = T_C_start + 273.15  # К

# Розрахунок кількості речовини
n_start = (P * V) / (R * T_K_start)
print(f"Початкова кількість речовини: {n_start:.2f} моль")
```

Результат розрахунку

$$ n = \frac{101325 \cdot 100}{8.31446261815 \cdot 1184.15} = 1029.14 \text{ моль} $$

Початкова кількість речовини: 1029.14 моль

### Крок 2: Розрахунок кількості речовини кожного компонента

Формула для розрахунку кількості речовини компонентів

$$ n_i = r_i \cdot n $$

де:
- $n_i$ - кількість речовини i-го компонента, моль
- $r_i$ - об'ємна частка i-го компонента (в долях одиниці)
- $n$ - загальна кількість речовини, моль (1029.14 моль)

```python
# Об'ємні частки компонентів (початкові)
r_co_start = 0.125  # 12.5%
r_co2_start = 0.871  # 87.1%
r_o2_start = 0.002  # 0.2%
r_n2_start = 0.002  # 0.2%

# Розрахунок кількості речовини кожного компонента
n_co_start = r_co_start * n_start
n_co2_start = r_co2_start * n_start
n_o2_start = r_o2_start * n_start
n_n2_start = r_n2_start * n_start

print(f"Початкова кількість CO: {n_co_start:.2f} моль")
print(f"Початкова кількість CO₂: {n_co2_start:.2f} моль")
print(f"Початкова кількість O₂: {n_o2_start:.2f} моль")
print(f"Початкова кількість N₂: {n_n2_start:.2f} моль")
```

Результат розрахунку

$$ n_{CO} = 0.125 \cdot 1029.14 = 128.64 \text{ моль} $$
$$ n_{CO_2} = 0.871 \cdot 1029.14 = 896.38 \text{ моль} $$
$$ n_{O_2} = 0.002 \cdot 1029.14 = 2.06 \text{ моль} $$
$$ n_{N_2} = 0.002 \cdot 1029.14 = 2.06 \text{ моль} $$

Початкова кількість CO: 128.64 моль

Початкова кількість CO₂: 896.38 моль

Початкова кількість O₂: 2.06 моль

Початкова кількість N₂: 2.06 моль

### Крок 3: Розрахунок коефіцієнтів теплоємності при постійному тиску

Формула для розрахунку теплоємності при постійному тиску

$$ c_p = a + bT + cT^{-2} \text{ кал/(моль·К)} $$

де:
- $a$, $b$, $c$ - коефіцієнти з таблиці
- $T$ - температура, К

Переведення в Дж/(моль·К):
$$ c_p[\text{Дж/(моль·К)}] = c_p[\text{кал/(моль·К)}] \cdot 4.1868 $$

```python
# Коефіцієнти теплоємності (кал/(моль·К))
# CO
a_co = 6.79
b_co = 0.98e-3
c_co = -0.11e5
# CO₂
a_co2 = 10.55
b_co2 = 2.16e-3
c_co2 = -2.04e5
# O₂
a_o2 = 7.16
b_o2 = 1.0e-3
c_o2 = -0.40e5
# N₂
a_n2 = 6.66
b_n2 = 1.02e-3
c_n2 = 0

# Коефіцієнт переведення з кал в Дж
CAL_TO_JOULE = 4.1868

# Середня температура для розрахунку (К)
T_C_end = 1000  # °C
T_K_end = T_C_end + 273.15  # К
T_avg = (T_K_start + T_K_end) / 2

# Розрахунок теплоємності при постійному тиску (Дж/(моль·К))
cp_co = (a_co + b_co * T_avg + c_co * T_avg**(-2)) * CAL_TO_JOULE
cp_co2 = (a_co2 + b_co2 * T_avg + c_co2 * T_avg**(-2)) * CAL_TO_JOULE
cp_o2 = (a_o2 + b_o2 * T_avg + c_o2 * T_avg**(-2)) * CAL_TO_JOULE
cp_n2 = (a_n2 + b_n2 * T_avg) * CAL_TO_JOULE

print(f"Теплоємність CO при постійному тиску: {cp_co:.2f} Дж/(моль·К)")
print(f"Теплоємність CO₂ при постійному тиску: {cp_co2:.2f} Дж/(моль·К)")
print(f"Теплоємність O₂ при постійному тиску: {cp_o2:.2f} Дж/(моль·К)")
print(f"Теплоємність N₂ при постійному тиску: {cp_n2:.2f} Дж/(моль·К)")
```

Результат розрахунку

Середня температура: $T_{avg} = (1184.15 + 1273.15) / 2 = 1228.65$ К

$$ c_p^{CO} = (6.79 + 0.98 \cdot 10^{-3} \cdot 1228.65 - 0.11 \cdot 10^{5} \cdot 1228.65^{-2}) \cdot 4.1868 = 33.44 \text{ Дж/(моль·К)} $$

$$ c_p^{CO_2} = (10.55 + 2.16 \cdot 10^{-3} \cdot 1228.65 - 2.04 \cdot 10^{5} \cdot 1228.65^{-2}) \cdot 4.1868 = 54.72 \text{ Дж/(моль·К)} $$

$$ c_p^{O_2} = (7.16 + 1.0 \cdot 10^{-3} \cdot 1228.65 - 0.40 \cdot 10^{5} \cdot 1228.65^{-2}) \cdot 4.1868 = 35.01 \text{ Дж/(моль·К)} $$

$$ c_p^{N_2} = (6.66 + 1.02 \cdot 10^{-3} \cdot 1228.65) \cdot 4.1868 = 33.13 \text{ Дж/(моль·К)} $$

Теплоємність CO при постійному тиску: 33.44 Дж/(моль·К)

Теплоємність CO₂ при постійному тиску: 54.72 Дж/(моль·К)

Теплоємність O₂ при постійному тиску: 35.01 Дж/(моль·К)

Теплоємність N₂ при постійному тиску: 33.13 Дж/(моль·К)

### Крок 4: Розрахунок коефіцієнтів теплоємності при постійному об'ємі

Формула для розрахунку теплоємності при постійному об'ємі

$$ c_v = c_p - R $$

де:
- $c_p$ - теплоємність при постійному тиску, Дж/(моль·К)
- $R$ - універсальна газова стала, Дж/(моль·К) (8.31446261815 Дж/(моль·К))

```python
# Розрахунок теплоємності при постійному об'ємі (Дж/(моль·К))
cv_co = cp_co - R
cv_co2 = cp_co2 - R
cv_o2 = cp_o2 - R
cv_n2 = cp_n2 - R

print(f"Теплоємність CO при постійному об'ємі: {cv_co:.2f} Дж/(моль·К)")
print(f"Теплоємність CO₂ при постійному об'ємі: {cv_co2:.2f} Дж/(моль·К)")
print(f"Теплоємність O₂ при постійному об'ємі: {cv_o2:.2f} Дж/(моль·К)")
print(f"Теплоємність N₂ при постійному об'ємі: {cv_n2:.2f} Дж/(моль·К)")
```

Результат розрахунку

$$ c_v^{CO} = 33.44 - 8.314 = 25.12 \text{ Дж/(моль·К)} $$
$$ c_v^{CO_2} = 54.72 - 8.314 = 46.40 \text{ Дж/(моль·К)} $$
$$ c_v^{O_2} = 35.01 - 8.314 = 26.70 \text{ Дж/(моль·К)} $$
$$ c_v^{N_2} = 33.13 - 8.314 = 24.82 \text{ Дж/(моль·К)} $$

Теплоємність CO при постійному об'ємі: 25.12 Дж/(моль·К)

Теплоємність CO₂ при постійному об'ємі: 46.40 Дж/(моль·К)

Теплоємність O₂ при постійному об'ємі: 26.70 Дж/(моль·К)

Теплоємність N₂ при постійному об'ємі: 24.82 Дж/(моль·К)

### Крок 5: Розрахунок теплоти нагрівання при постійному об'ємі (ізохоричний процес)

Формула для розрахунку теплоти нагрівання при постійному об'ємі

$$ Q_v = \sum_{i} n_i \cdot c_{v,i} \cdot (T_2 - T_1) $$

де:
- $n_i$ - кількість речовини i-го компонента, моль
- $c_{v,i}$ - теплоємність i-го компонента при постійному об'ємі, Дж/(моль·К)
- $T_2$ - кінцева температура, К (1000°C = 1273.15 К)
- $T_1$ - початкова температура, К (911°C = 1184.15 К)

```python
# Різниця температур (К)
delta_T = T_K_end - T_K_start

# Розрахунок теплоти нагрівання при постійному об'ємі (Дж)
Q_v_co = n_co_start * cv_co * delta_T
Q_v_co2 = n_co2_start * cv_co2 * delta_T
Q_v_o2 = n_o2_start * cv_o2 * delta_T
Q_v_n2 = n_n2_start * cv_n2 * delta_T

Q_v_total = Q_v_co + Q_v_co2 + Q_v_o2 + Q_v_n2

print(f"Теплота нагрівання CO при постійному об'ємі: {Q_v_co/1000:.2f} кДж")
print(f"Теплота нагрівання CO₂ при постійному об'ємі: {Q_v_co2/1000:.2f} кДж")
print(f"Теплота нагрівання O₂ при постійному об'ємі: {Q_v_o2/1000:.2f} кДж")
print(f"Теплота нагрівання N₂ при постійному об'ємі: {Q_v_n2/1000:.2f} кДж")
print(f"Загальна теплота нагрівання при постійному об'ємі: {Q_v_total/1000:.2f} кДж")
```

Результат розрахунку

$$ \Delta T = 1273.15 - 1184.15 = 89 \text{ К} $$

$$ Q_{v,CO} = 128.64 \cdot 25.12 \cdot 89 = 287.66 \text{ кДж} $$
$$ Q_{v,CO_2} = 896.38 \cdot 46.40 \cdot 89 = 3701.85 \text{ кДж} $$
$$ Q_{v,O_2} = 2.06 \cdot 26.70 \cdot 89 = 4.89 \text{ кДж} $$
$$ Q_{v,N_2} = 2.06 \cdot 24.82 \cdot 89 = 4.55 \text{ кДж} $$

$$ Q_{v,total} = 287.66 + 3701.85 + 4.89 + 4.55 = 3998.94 \text{ кДж} $$

Теплота нагрівання CO при постійному об'ємі: 287.66 кДж

Теплота нагрівання CO₂ при постійному об'ємі: 3701.85 кДж

Теплота нагрівання O₂ при постійному об'ємі: 4.89 кДж

Теплота нагрівання N₂ при постійному об'ємі: 4.55 кДж

Загальна теплота нагрівання при постійному об'ємі: 3998.94 кДж

### Крок 6: Розрахунок кінцевої кількості речовини для ізобаричного процесу

Формула для розрахунку кінцевої кількості речовини при постійному тиску

$$ n_2 = n_1 \cdot \frac{T_1}{T_2} $$

де:
- $n_2$ - кінцева кількість речовини, моль
- $n_1$ - початкова кількість речовини, моль (1029.14 моль)
- $T_1$ - початкова температура, К (911°C = 1184.15 К)
- $T_2$ - кінцева температура, К (1000°C = 1273.15 К)


```python
# Розрахунок кінцевої кількості речовини для ізобаричного процесу
n_end_isobaric = n_start * (T_K_start / T_K_end)

# Розрахунок кінцевої кількості речовини кожного компонента
# Використовуємо кінцеві об'ємні частки
r_co_end = 0.088  # 8.8%
r_co2_end = 0.910  # 91.0%
r_o2_end = 0.001  # 0.1%
r_n2_end = 0.001  # 0.1%

n_co_end = r_co_end * n_end_isobaric
n_co2_end = r_co2_end * n_end_isobaric
n_o2_end = r_o2_end * n_end_isobaric
n_n2_end = r_n2_end * n_end_isobaric

print(f"Кінцева кількість речовини (ізобаричний процес): {n_end_isobaric:.2f} моль")
print(f"Кінцева кількість CO: {n_co_end:.2f} моль")
print(f"Кінцева кількість CO₂: {n_co2_end:.2f} моль")
print(f"Кінцева кількість O₂: {n_o2_end:.2f} моль")
print(f"Кінцева кількість N₂: {n_n2_end:.2f} моль")
```

Результат розрахунку

$$ n_2 = 1029.14 \cdot \frac{1184.15}{1273.15} = 957.20 \text{ моль} $$

$$ n_{CO,end} = 0.088 \cdot 957.20 = 84.23 \text{ моль} $$
$$ n_{CO_2,end} = 0.910 \cdot 957.20 = 871.05 \text{ моль} $$
$$ n_{O_2,end} = 0.001 \cdot 957.20 = 0.96 \text{ моль} $$
$$ n_{N_2,end} = 0.001 \cdot 957.20 = 0.96 \text{ моль} $$

Кінцева кількість речовини (ізобаричний процес): 957.20 моль

Кінцева кількість CO: 84.23 моль

Кінцева кількість CO₂: 871.05 моль

Кінцева кількість O₂: 0.96 моль

Кінцева кількість N₂: 0.96 моль

### Крок 7: Розрахунок теплоти нагрівання при постійному тиску (ізобаричний процес)

#### 7.1 Формула для розрахунку теплоти нагрівання при постійному тиску

$$ Q_p = \sum_{i} \bar{n}_i \cdot c_{p,i} \cdot (T_2 - T_1) $$

де:
- $\bar{n}_i$ - середня кількість речовини i-го компонента, моль
- $c_{p,i}$ - теплоємність i-го компонента при постійному тиску, Дж/(моль·К)
- $T_2$ - кінцева температура, К (1000°C = 1273.15 К)
- $T_1$ - початкова температура, К (911°C = 1184.15 К)

```python
# Розрахунок середньої кількості речовини для кожного компонента
n_co_avg = (n_co_start + n_co_end) / 2
n_co2_avg = (n_co2_start + n_co2_end) / 2
n_o2_avg = (n_o2_start + n_o2_end) / 2
n_n2_avg = (n_n2_start + n_n2_end) / 2

# Розрахунок теплоти нагрівання при постійному тиску (Дж)
Q_p_co = n_co_avg * cp_co * delta_T
Q_p_co2 = n_co2_avg * cp_co2 * delta_T
Q_p_o2 = n_o2_avg * cp_o2 * delta_T
Q_p_n2 = n_n2_avg * cp_n2 * delta_T

Q_p_total = Q_p_co + Q_p_co2 + Q_p_o2 + Q_p_n2

print(f"Теплота нагрівання CO при постійному тиску: {Q_p_co/1000:.2f} кДж")
print(f"Теплота нагрівання CO₂ при постійному тиску: {Q_p_co2/1000:.2f} кДж")
print(f"Теплота нагрівання O₂ при постійному тиску: {Q_p_o2/1000:.2f} кДж")
print(f"Теплота нагрівання N₂ при постійному тиску: {Q_p_n2/1000:.2f} кДж")
print(f"Загальна теплота нагрівання при постійному тиску: {Q_p_total/1000:.2f} кДж")
```

Результат розрахунку

$$ \bar{n}_{CO} = \frac{128.64 + 84.23}{2} = 106.44 \text{ моль} $$
$$ \bar{n}_{CO_2} = \frac{896.38 + 871.05}{2} = 883.72 \text{ моль} $$
$$ \bar{n}_{O_2} = \frac{2.06 + 0.96}{2} = 1.51 \text{ моль} $$
