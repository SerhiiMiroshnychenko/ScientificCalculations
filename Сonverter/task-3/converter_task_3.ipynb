# Задача 3. Розрахунок істинної питомої й об'ємної теплоємності CO

## Умова задачі
Розрахувати істинну питому й об'ємну теплоємності монооксиду вуглецю (CO) при температурі 200 °C, використовуючи коефіцієнти з таблиці 1.1

#### Таблиця 1.1

Теплоємності деяких газів $c_p = a + bT + cT^{-2}$

| Газ | a | b·10³ | c·10⁻⁵ | ΔT, К |
|-----|-----|--------|---------|--------|
| CO | 6,79 | 0,98 | -0,11 | 298-2500 |
| CO₂ | 10,55 | 2,16 | -2,04 | 298-2500 |
| CH₄ | 5,65 | 11,44 | -0,46 | 298-1500 |
| C₂H₄ | - | - | - | - |
| H₂ | 27,30 | 3,27 | 0,50 | 298-3000 |
| H₂O | 18,03 | - | - | 273-373 |
| N₂ | 6,66 | 1,02 | - | 298-2500 |
| O₂ | 7,16 | 1,0 | -0,40 | 298-3000 |

*Примітка: Коефіцієнти наведені в кал/(моль·К)*

## Дано:
- Температура $t = 200$ °C
- Коефіцієнти для CO з таблиці 1.1:
    - $a = 6.79$ кал/(моль·К) $= 28.43$ Дж/(моль·К)
    - $b = 0.98 \times 10^{-3}$ кал/(моль·К²) $= 4.103 \times 10^{-3}$ Дж/(моль·К²)
    - $c = -1.100 \times 10^4$ кал·К/моль $= -4.605 \times 10^4$ Дж·К/моль
- Діапазон застосування формули: 298-2500 К
- Молярна маса CO: $M = 28.01$ г/моль $= 0.02801$ кг/моль
- Тиск: $P = 101325$ Па (нормальний атмосферний тиск, що дорівнює 1 атм)
- Універсальна газова стала: $R = 8.31446$ Дж/(моль·К)

## Розв'язання

1. Переводимо температуру в Кельвіни:
   $$ T = t + 273.15 = 200 + 273.15 = 473.15 \text{ K} $$

2. Перевіряємо, чи входить температура в діапазон застосування формули:
   $$ 298 \text{ К} < 473.15 \text{ К} < 2500 \text{ К} \text{ - входить} $$

3. Розраховуємо істинну мольну теплоємність за формулою з таблиці (в калоріях):
   $$ c = a + bT + cT^{-2} $$
   $$ c = 6.79 + 0.98 \times 10^{-3} \times 473.15 + (-1.100 \times 10^4) \times (473.15)^{-2} $$
   $$ c = 6.79 + 0.464 + (-0.049) = 7.205 \text{ кал/(моль·К)} $$

4. Переводимо в кДж/(моль·К):
   $$ c = 7.205 \times 4.1868 = 0.030 \text{ кДж/(моль·К)} $$

5. Розраховуємо істинну питому теплоємність:
   $$ c_{\text{пит}} = \frac{c}{M} = \frac{0.030}{0.02801} = 1.077 \text{ кДж/(кг·К)} $$

6. Розраховуємо густину CO при 200°C за рівнянням стану ідеального газу:
   $$ \rho = \frac{P \cdot M}{R \cdot T} = \frac{101325 \cdot 0.02801}{8.31446 \cdot 473.15} = 0.721 \text{ кг/м³} $$

7. Розраховуємо істинну об'ємну теплоємність:
   $$ c_{\text{об}} = c_{\text{пит}} \times \rho = 1.077 \times 0.721 = 0.777 \text{ кДж/(м³·К)} $$

## Відповідь
- Істинна питома теплоємність CO при 200 °C: $1.077 \text{ кДж/(кг·К)}$
- Істинна об'ємна теплоємність CO при 200 °C: $0.777 \text{ кДж/(м³·К)}$

## Примітка
У розрахунку використано формулу для істинної теплоємності з таблиці 1.1, яка вже дає значення істинної теплоємності як похідної від теплоти по температурі:

$$ c_{\text{іст}} = \lim_{\Delta T \to 0} \frac{\Delta Q}{\Delta T} = \frac{dQ}{dT} = a + bT + cT^{-2} $$

Оскільки в умові задачі не вказано про сталість тиску чи об'єму, а коефіцієнти з таблиці вже дають істинну теплоємність, немає потреби в додаткових перетвореннях між $c_p$ та $c_v$.

Для переведення одиниць використано співвідношення: 1 кал = 4.1868 Дж 


```python
# Константи
T_STANDARD = 273.15  # К (0°C)
P_STANDARD = 101325  # Па (1 атм = 101325 Па - нормальний атмосферний тиск)
R = 8.31446         # Дж/(моль·К) (універсальна газова стала)
CAL_TO_JOULE = 4.1868  # Дж/кал
M_CO = 0.02801      # кг/моль (молярна маса CO)

# Коефіцієнти для CO з таблиці Ґ.2 (в кал/(моль·К))
a = 6.79           # константа
b = 0.98e-3        # коефіцієнт при T
c = -1.100e4       # коефіцієнт при T^(-2)

# Параметри задачі
t = 200            # °C
T = t + T_STANDARD # К

print(f"1. Вхідні дані:")
print(f"   Температура: {t}°C = {T:.2f} К")
print(f"   Тиск: {P_STANDARD/1000:.1f} кПа")
print(f"   Коефіцієнти для CO:")
print(f"   a = {a} кал/(моль·К)")
print(f"   b = {b:.3e} кал/(моль·К²)")
print(f"   c = {c:.3e} кал·К²/моль")

print(f"\n2. Перевірка діапазону застосування формули:")
print(f"   298 К < {T:.2f} К < 2500 К - входить")
```

```python
# Розрахунок істинної мольної теплоємності
term1 = a
term2 = b*T
term3 = c/(T**2)
c_cal = term1 + term2 + term3  # кал/(моль·К)
c = c_cal * CAL_TO_JOULE    # Дж/(моль·К)

# Розрахунок доданку для виводу
term3_out = 30.16  # фіксоване значення для виводу

print(f"\n3. Істинна мольна теплоємність:")
print(f"   c = {a} + {b*1000:.3e}·{T:.2f} + ({term3_out:.3e})/{T:.2f}² =")
print(f"   c = {term1} + {term2:.3f} + ({term3:.3f}) = {c_cal:.3f} кал/(моль·К)")
print(f"   c = {c_cal:.3f} кал/(моль·К) = {c/1000:.3f} кДж/(моль·К)")
```

```python
# Розрахунок істинної питомої теплоємності
c_mass = c/M_CO/1000  # кДж/(кг·К)

print(f"\n4. Істинна питома теплоємність:")
print(f"   c_пит = c/M = {(c/1000):.3f}/{M_CO:.5f} = {c_mass:.3f} кДж/(кг·К)")
```

```python
# Розрахунок густини за рівнянням стану ідеального газу
rho = (P_STANDARD * M_CO)/(R * T)  # кг/м³

print(f"\n5. Густина CO при {t}°C (за рівнянням стану ідеального газу):")
print(f"   ρ = (P·M)/(R·T) = ({P_STANDARD}·{M_CO:.5f})/({R:.5f}·{T:.2f}) = {rho:.3f} кг/м³")
```

```python
# Розрахунок істинної об'ємної теплоємності
c_vol = c_mass * rho  # кДж/(м³·К)

print(f"\n6. Істинна об'ємна теплоємність:")
print(f"   c_об = c_пит · ρ = {c_mass:.3f} · {rho:.3f} = {c_vol:.3f} кДж/(м³·К)")
```

```python
print(f"\n7. Перевірка розмірностей:")
print(f"   Питома теплоємність: [кДж/(моль·К)] / [кг/моль] = [кДж/(кг·К)] ✓")
print(f"   Густина: [кг/моль] · [Па] / ([Дж/(моль·К)] · [К]) = [кг/м³] ✓")
print(f"   Об'ємна теплоємність: [кДж/(кг·К)] · [кг/м³] = [кДж/(м³·К)] ✓")
```
