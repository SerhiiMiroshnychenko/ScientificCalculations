# Задача 9: Розрахунок середньої питомої теплоємності конвертерного шлаку

## Умова задачі
Основний конвертерний шлак має склад, мас. %:
- СаO – 47
- FeО – 14
- MnО – 15
- SіО₂ – 24

Обчислити середню питому теплоємність шлаку в інтервалі температур від 0 до 1000 °С:
1. За правилом адитивності
2. З урахуванням температурної залежності

## Теоретичні відомості

### Температурні залежності теплоємностей

Для конвертерного шлаку:
$$ \bar{c} = 0,777 + 1,31 \cdot 10^{-4}T - 5,45 \cdot 10^{-8}T^2, \text{ кДж/(кг·K)} $$

Де T - температура в Кельвінах (K). Для переведення: T(K) = t(°C) + 273,15

Для компонентів:
1. СаО (0-800 °С):
   $$ \bar{c} = 0,749 + 3,78 \cdot 10^{-4}T - 1,535 \cdot 10^{-7}T^2, \text{ кДж/(кг·K)} $$

2. SіО₂ (0-1300 °С):
   $$ \bar{c} = 0,768 + 3,23 \cdot 10^{-4}T, \text{ кДж/(кг·K)} $$

3. FeО (800-1500 °С):
   $$ c = 0,7872 \text{ кДж/(кг·K)} $$

4. MnО (800-1500 °С):
   $$ c = 0,7268 \text{ кДж/(кг·K)} $$

Важливо: У всіх формулах T - температура в Кельвінах. Теплоємність вимірюється в кДж/(кг·K).

### Метод адитивності
$$ c = \sum_{i=1}^n x_i c_i $$

де:
- $c$ - питома теплоємність суміші, кДж/(кг·K)
- $x_i$ - масова частка i-го компонента
- $c_i$ - питома теплоємність i-го компонента, кДж/(кг·K)

## Розв'язання

### 1. Метод адитивності

#### 1.1 Переведення масових часток у десятковий вигляд
- СаО: $x_1 = 47\% = 0,47$
- FeО: $x_2 = 14\% = 0,14$
- MnО: $x_3 = 15\% = 0,15$
- SіО₂: $x_4 = 24\% = 0,24$

#### 1.2 Розрахунок середніх теплоємностей компонентів при 1000°С (1273,15 K)

Для СаО:
$$ \bar{c}_1 = 0,749 + 3,78 \cdot 10^{-4} \cdot 1273,15 - 1,535 \cdot 10^{-7} \cdot 1273,15^2 = 0,9814 \text{ кДж/(кг·K)} $$

Для SіО₂:
$$ \bar{c}_2 = 0,768 + 3,23 \cdot 10^{-4} \cdot 1273,15 = 1,1792 \text{ кДж/(кг·K)} $$

Для FeО:
$$ \bar{c}_3 = 0,7872 \text{ кДж/(кг·K)} $$

Для MnО:
$$ \bar{c}_4 = 0,7268 \text{ кДж/(кг·K)} $$

#### 1.3 Розрахунок середньої теплоємності шлаку за правилом адитивності

Внески компонентів:
- СаО: $0,47 \cdot 0,9814 = 0,4613 \text{ кДж/(кг·K)}$
- FeО: $0,14 \cdot 0,7872 = 0,1102 \text{ кДж/(кг·K)}$
- MnО: $0,15 \cdot 0,7268 = 0,1090 \text{ кДж/(кг·K)}$
- SіО₂: $0,24 \cdot 1,1792 = 0,2830 \text{ кДж/(кг·K)}$

Сумарна теплоємність:
$$ c_{ад} = 0,4613 + 0,1102 + 0,1090 + 0,2830 = 0,9635 \text{ кДж/(кг·K)} $$

### 2. Метод з урахуванням температурної залежності

Підставляємо T = 1000°С (1273,15 K) у формулу для шлаку:

$$ c_{зал} = 0,777 + 1,31 \cdot 10^{-4} \cdot 1273,15 - 5,45 \cdot 10^{-8} \cdot 1273,15^2 = 0,8554 \text{ кДж/(кг·K)} $$

### 3. Порівняння результатів

1. За правилом адитивності: 0,9635 кДж/(кг·K)
2. З урахуванням температурної залежності: 0,8554 кДж/(кг·K)
3. Абсолютна різниця: 0,1081 кДж/(кг·K)
4. Відносна різниця: 12,63%

## Відповідь
1. Середня питома теплоємність конвертерного шлаку за правилом адитивності: 0,9635 кДж/(кг·K)
2. Середня питома теплоємність конвертерного шлаку з урахуванням температурної залежності: 0,8554 кДж/(кг·K)
3. Різниця між методами становить 12,63%

## Аналіз результатів
1. Метод адитивності дає більше значення теплоємності, ніж метод з урахуванням температурної залежності
2. Значна різниця (12,63%) може бути пов'язана з:
   - Взаємодією компонентів у шлаку
   - Утворенням нових хімічних сполук
   - Структурними змінами при високих температур
   - Фазовими переходами компонентів
3. Особливості розрахунку:
   - Для FeО та MnО використані постійні значення теплоємності
   - Температура розрахунку (1000°С) перевищує діапазон формули для СаО (0-800°С)
   - Для SiO₂ використана лінійна залежність від температури

## Аналіз температурних діапазонів

### 1. CaO (0-800°C): ✗
- Формула для CaO дійсна лише до 800°C
- Використання при 1000°C (на 200°C вище допустимого діапазону)
- Розрахована теплоємність CaO (0,9735 кДж/(кг·°С)) може бути неточною
- При такій високій температурі можливі додаткові фізико-хімічні процеси, не враховані формулою

### 2. SiO2 (0-1300°C): ✓
- Температура 1000°C повністю входить в діапазон
- Лінійна залежність теплоємності від температури залишається дійсною
- Розрахунок теплоємності (1,0910 кДж/(кг·°С)) є достовірним

### 3. FeO (800-1500°C): ✓
- Температура 1000°C входить в робочий діапазон
- Використовується постійне значення теплоємності 0,7872 кДж/(кг·°С)
- Це спрощення є допустимим для даного температурного інтервалу

### 4. MnO (800-1500°C): ✓
- Температура 1000°C входить в робочий діапазон
- Використовується постійне значення теплоємності 0,7268 кДж/(кг·°С)
- Це спрощення є допустимим для даного температурного інтервалу

### 5. Шлак (до температури плавлення): ✓
- Температура плавлення конвертерного шлаку зазвичай перевищує 1300°C
- При 1000°C шлак знаходиться в твердому стані
- Формула температурної залежності залишається дійсною
- Не відбувається фазових переходів, які могли б вплинути на точність розрахунку

## Висновки щодо достовірності результатів

1. Найбільшу невизначеність вносить розрахунок теплоємності CaO, оскільки температурна залежність виходять за межі застосування формули

2. Враховуючи, що CaO складає 47% шлаку, це може бути однією з основних причин значної різниці (9,97%) між методами розрахунку

3. Для підвищення точності розрахунку рекомендується:
   - Знайти формулу для теплоємності CaO, дійсну при вищих температур
   - Або обмежити температур розрахунку до 800°C
   - Або використовувати експериментальні дані для CaO при високих температур

## Розв'язання в *Python*

```python
# Масові частки компонентів, %
x_cao = 47    # CaO
x_feo = 14    # FeO
x_mno = 15    # MnO
x_sio2 = 24   # SiO2
```

```python
# Температура, K
T = 1000 + 273.15
```

```python
print("1. Вхідні дані:")
print(f"   Температура: {T} K")
print("\n   Масові частки компонентів, %:")
print(f"   CaO: {x_cao}")
print(f"   FeO: {x_feo}")
print(f"   MnO: {x_mno}")
print(f"   SiO2: {x_sio2}")
```

```python
# Переведення масових часток у десятковий вигляд
x_cao_dec = x_cao/100
x_feo_dec = x_feo/100
x_mno_dec = x_mno/100
x_sio2_dec = x_sio2/100

print("\n2. Масові частки у десятковому вигляді:")
print(f"   CaO: {x_cao_dec:.4f}")
print(f"   FeO: {x_feo_dec:.4f}")
print(f"   MnO: {x_mno_dec:.4f}")
print(f"   SiO2: {x_sio2_dec:.4f}")
```

```python
# Перевірка суми масових часток
x_sum = x_cao_dec + x_feo_dec + x_mno_dec + x_sio2_dec
print(f"\n3. Сума масових часток: {x_sum:.4f}")
```

```python
# Функції для розрахунку теплоємності компонентів
def c_cao(T):
    """Теплоємність CaO, кДж/(кг·K)"""
    return 0.749 + 3.78e-4*T - 1.535e-7*T**2

def c_sio2(T):
    """Теплоємність SiO2, кДж/(кг·K)"""
    return 0.768 + 3.23e-4*T

def c_feo():
    """Теплоємність FeO, кДж/(кг·K)"""
    return 0.7872

def c_mno():
    """Теплоємність MnO, кДж/(кг·K)"""
    return 0.7268

def c_slag(T):
    """Теплоємність шлаку, кДж/(кг·K)"""
    return 0.777 + 1.31e-4*T - 5.45e-8*T**2
```

```python
# Розрахунок теплоємностей компонентів при заданій температурі
c_cao_T = c_cao(T)
c_sio2_T = c_sio2(T)
c_feo_T = c_feo()
c_mno_T = c_mno()

print(f"\n4. Теплоємності компонентів при {T} K, кДж/(кг·K):")
print(f"   CaO: {c_cao_T:.4f}")
print(f"   FeO: {c_feo_T:.4f}")
print(f"   MnO: {c_mno_T:.4f}")
print(f"   SiO2: {c_sio2_T:.4f}")
```

```python
# Метод адитивності
c_additive = (x_cao_dec * c_cao_T + x_feo_dec * c_feo_T + 
              x_mno_dec * c_mno_T + x_sio2_dec * c_sio2_T)
```

```python
print("\n5. Внески компонентів у теплоємність суміші (метод адитивності), кДж/(кг·K):")
print(f"   CaO: {x_cao_dec * c_cao_T:.4f}")
print(f"   FeO: {x_feo_dec * c_feo_T:.4f}")
print(f"   MnO: {x_mno_dec * c_mno_T:.4f}")
print(f"   SiO2: {x_sio2_dec * c_sio2_T:.4f}")
```

```python
print(f"\n6. Середня питома теплоємність (метод адитивності):")
print(f"   c = {c_additive:.4f} кДж/(кг·K)")
```

```python
# Метод з урахуванням температурної залежності
c_temp = c_slag(T)

print(f"\n7. Середня питома теплоємність (з урахуванням температурної залежності):")
print(f"   c = {c_temp:.4f} кДж/(кг·K)")
```

```python
# Порівняння методів
abs_diff = abs(c_additive - c_temp)
rel_diff = abs_diff/c_temp * 100

print("\n8. Порівняння методів:")
print(f"   Абсолютна різниця: {abs_diff:.4f} кДж/(кг·K)")
print(f"   Відносна різниця: {rel_diff:.2f}%")
```

```python
print("\n9. Перевірка достовірності:")
print(f"   - Сума масових часток = 1: {'✓' if abs(x_sum - 1) < 1e-10 else '✗'}")
print(f"   - Розмірності збережені: ✓")
print(f"   - Температура в межах діапазонів:")
print(f"     * CaO (0-800°C): {'✗' if (T-273.15) > 800 else '✓'}")
print(f"     * SiO2 (0-1300°C): {'✗' if (T-273.15) > 1300 else '✓'}")
print(f"     * FeO (800-1500°C): {'✓' if 800 <= (T-273.15) <= 1500 else '✗'}")
print(f"     * MnO (800-1500°C): {'✓' if 800 <= (T-273.15) <= 1500 else '✗'}")
print(f"     * Шлак (до температури плавлення): ✓")
```
