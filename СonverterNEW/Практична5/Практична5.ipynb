# Практична робота №5. Розрахунок охолоджувачів (брухт vs руда) для термостатування плавки

## 1. Постановка задачі

Необхідно виконати розрахунок маси охолоджувача, яку слід додати до шихти киснево-конвертерної плавки для повного погашення теплового надлишку $\Delta Q$ (з ПР №4) при збереженні цільової температури випуску сталі $t_{tap}$.

Розрахунок ведеться на базі **1000 кг початкової металевої шихти** ($m_{charge} = 1000$ кг) з результатів ПР №1 та ПР №4.

### 1.1. Цілі роботи

- обчислити **охолоджуючу здатність** сталевого брухту $\Delta h_{scrap}$ за заданої температури випуску $t_{tap}$;
- обчислити **охолоджуючу здатність** залізної руди $\Delta h_{ore}$ та зрозуміти природу кожної складової;
- розрахувати масу охолоджувача (брухт, руда або їхня суміш) для повного погашення $\Delta Q$;
- визначити оновлені маси шихти та сталі з урахуванням доданого охолоджувача;
- виконати **верифікацію** теплового балансу: переконатися, що $Q_{absorbed} \approx \Delta Q$;
- провести аналіз чутливості результатів до параметра $x_{ore}$;
- порівняти ефективність різних типів охолоджувачів та обрати раціональне рішення.

### 1.2. Вхідні дані (з попередніх робіт)

**З ПР №1:** $m_{hm}$ — маса рідкого чавуну [кг]; $m_{scr}$ — маса шихтового брухту [кг]; $m_{steel}$ — маса сталі до додавання охолоджувача [кг]; $Yield$ — базовий вихід сталі [%].

**З ПР №4:** $\Delta Q = Q_{in} - Q_{out}$ — тепловий надлишок [МДж]; $t_{tap}$ — цільова температура випуску сталі [°C].

### 1.3. Параметри варіанта (ПР №5)

- $x_{ore}$ — частка руди в сумішевому охолоджувачі ($0 \le x_{ore} \le 1$).

### 1.4. Прийняті типові параметри моделі

| Параметр | Позначення | Значення | Розмірність | Обґрунтування та діапазон |
|:---|:---|:---:|:---|:---|
| Початкова температура охолоджувача | $t_0$ | 20 | °C | Кімнатна температура складу |
| Температура плавлення заліза | $T_m$ | 1535 | °C | Стандарт укр. підручників; IUPAC: 1538 °C |
| Середня теплоємність твердого Fe | $C_{p,s}$ | 0,65 | кДж/(кг·К) | Середньоінт. за 20–1535 °C; діапазон 0,63–0,68 (Kubaschewski) |
| Теплоємність рідкої сталі | $C_{p,l}$ | 0,836 | кДж/(кг·К) | Для рідкої сталі 1500–1650 °C; діапазон 0,79–0,84 |
| Теплота плавлення Fe | $L_m$ | 272 | кДж/кг | Укр. підручники; NIST: 247 кДж/кг; прийнятий діапазон 247–272 |
| Охолоджуюча здатність руди | $\Delta h_{ore}$ | 2800 | кДж/кг | Типова руда 85% Fe₂O₃, 8% H₂O; діапазон 2500–3200 |
| Коефіцієнт виходу металу з брухту | $\eta_{scrap}$ | 0,92 | — | 8% брухту окиснюється → FeO у шлак і окалина |
| Коефіцієнт виходу металу з руди | $\eta_{ore}$ | 0,56 | — | $\approx 0{,}70 \times 0{,}80$: вміст Fe × ступінь відновлення |

> **Спрощення моделі.** Формула $\Delta h_{scrap}$ використовує властивості чистого заліза. Реальний брухт містить 0,1–0,5% C, що знижує ефективну температуру плавлення до ~1480–1515 °C. У рамках навчальної моделі ця поправка складає менше 2% від $\Delta h_{scrap}$ і нею нехтують.

## 2. Теоретичні положення

### 2.1. Фізичний зміст $\Delta Q$ і необхідність охолоджувача

Тепловий надлишок:

$$\Delta Q = Q_{in} - Q_{out}$$

**Якщо $\Delta Q > 0$:** фактична температура випуску перевищила б $t_{tap}$ на:

$$\Delta T = \frac{\Delta Q \cdot 10^3}{m_{steel} \cdot C_{p,l}} \approx 25\text{–}30 \text{ °C}$$

Це технологічно неприпустимо (прискорена ерозія футерівки, перегрів сталі, зростання вмісту азоту). **Необхідне додавання охолоджувача.**

**Якщо $\Delta Q < 0$:** у ванні існує дефіцит тепла — без коригування температура сталі не досягне $t_{tap}$. Охолоджувач у цьому випадку **не додають**. Можливі рішення: зменшення частки брухту в шихті, підвищення вмісту Si у чавуні, застосування додаткового кисневого перегріву.

**Якщо $\Delta Q = 0$:** баланс теоретично ідеальний, охолоджувача не потрібно.

У всіх варіантах цієї роботи $\Delta Q > 0$.

### 2.2. Механізм дії охолоджувачів

**Сталевий брухт** поглинає тепло виключно **фізично** — через три послідовні процеси: нагрів твердого металу від $t_0$ до $T_m$, плавлення при $T_m$ (витрата прихованої теплоти $L_m$), перегрів рідкого металу від $T_m$ до $t_{tap}$.

**Залізна руда** поглинає тепло **фізично та хімічно**: нагрів твердої руди, випаровування гігроскопічної вологи (ендотермічно), ендотермічне відновлення Fe₂O₃ вуглецем і CO ванни, розчинення відновленого заліза в розплаві. Саме хімічна складова пояснює, чому **1 кг руди поглинає в 2,0–2,5 рази більше тепла**, ніж 1 кг брухту.

### 2.3. Охолоджуюча здатність брухту

$$\boxed{\Delta h_{scrap} = C_{p,s}(T_m - t_0) + L_m + C_{p,l}(t_{tap} - T_m)} \quad [\text{кДж/кг}]$$

де:
- $C_{p,s}(T_m - t_0)$ — фізичне тепло нагріву твердого заліза від $t_0$ до $T_m$;
- $L_m$ — прихована теплота фазового переходу «тверде → рідке»;
- $C_{p,l}(t_{tap} - T_m)$ — теплота перегріву рідкої сталі до температури випуску.

**Залежність від $t_{tap}$:** $\Delta h_{scrap}$ лінійно зростає з $t_{tap}$ з нахилом $C_{p,l} = 0{,}836$ кДж/(кг·К). Для практичного діапазону 1450–1650 °C: $\Delta h_{scrap} \approx 1200$–$1490$ кДж/кг.

### 2.4. Наукове обґрунтування констант для $\Delta h_{scrap}$

**$C_{p,s} = 0{,}65$ кДж/(кг·К)** — середньоінтегральне значення для заліза за діапазоном 20–1535 °C. Реальна залежність $C_{p}(T)$ для Fe має складний вигляд: від 0,449 кДж/(кг·К) при 20 °C до аномального піку ~0,80 кДж/(кг·К) поблизу точки Кюрі (770 °C, $\alpha \to \beta$ перехід), далі ~0,63 кДж/(кг·К) вище 850 °C. Середнє за повним діапазоном — 0,63–0,68 кДж/(кг·К), значення 0,65 відповідає цьому інтервалу.

**$L_m = 272$ кДж/кг** — теплота плавлення чистого заліза. Значення NIST: 247 кДж/кг (13,81 кДж/моль); українські та деякі міжнародні металургійні підручники використовують 272 кДж/кг, що включає усереднення для технічного заліза. Розбіжність ~10% несуттєво впливає на результат, оскільки $L_m$ становить лише ~21% від $\Delta h_{scrap}$.

**$C_{p,l} = 0{,}836$ кДж/(кг·К)** — теплоємність рідкої сталі при 1500–1650 °C. Kubaschewski & Alcock наводять 0,795–0,836 кДж/(кг·К) для рідкого заліза залежно від температури; значення 0,836 відповідає верхній частині цього діапазону і є стандартом низки підручників.

### 2.5. Охолоджуюча здатність залізної руди: декомпозиція за процесами

Для типової залізної руди ($\approx$75–80% Fe₂O₃, $\approx$8% H₂O, решта — пуста порода SiO₂, Al₂O₃) значення $\Delta h_{ore} = 2800$ кДж/кг розкладається на такі фізично визначені складові:

| Процес | Теплопоглинання, кДж/кг руди |
|:---|:---:|
| 1. Нагрів твердої руди від $t_0 = 20$ до ~1600 °C ($C_{p,ore} \approx 0{,}87$ кДж/(кг·К)) | ~1350–1400 |
| 2. Випаровування вологи (8% H₂O): нагрів + пароутворення | ~200–260 |
| 3. Ендотермічне відновлення Fe₂O₃ вуглецем ванни: $\text{Fe}_2\text{O}_3 + 3\text{C} \to 2\text{Fe} + 3\text{CO}\uparrow$ | ~900–1100 |
| 4. Тепло розчинення Fe у рідкій ванні (слабко екзотермічна складова, знак «–») | ~(80–120) |
| **Сума** | **~2450–2740** |

Округлення до **2800 кДж/кг** відповідає верхній межі для руди з підвищеним вмістом Fe₂O₃ і є консервативною оцінкою в рамках навчальної моделі.

> **Примітка щодо реакції відновлення.** При 298 K реакція Fe₂O₃ + 3CO → 2Fe + 3CO₂ є слабко екзотермічною (ΔH ≈ −25 кДж/моль Fe₂O₃). Однак у конвертері за ~1600 °C відновлення відбувається переважно **вуглецем**, а не CO, і продуктом є CO, а не CO₂:
> $$\text{Fe}_2\text{O}_3 + 3\text{C} \to 2\text{Fe} + 3\text{CO}, \quad \Delta H^\circ_{298} \approx +493 \text{ кДж/моль Fe}_2\text{O}_3 \approx +3090 \text{ кДж/кг Fe}_2\text{O}_3$$
> Ця реакція **суттєво ендотермічна** і є основним джерелом охолоджуючого ефекту залізної руди. Для руди з 80% Fe₂O₃: внесок відновлення ≈ $3090 \times 0{,}80 / 2{,}79 \approx 886$ кДж/кг, де 2,79 — коефіцієнт перерахунку з молярної маси на масу Fe₂O₃ з урахуванням ступеня відновлення.

### 2.6. Обґрунтування коефіцієнтів виходу металу $\eta$

**$\eta_{scrap} = 0{,}92$** — 92% маси доданого брухту переходить у рідку сталь. Втрати 8% складаються з: окиснення поверхневого заліза → FeO у шлак (~4–5%), вигоряння C і Mn (~2%), окалина і забруднення на поверхні (~1–2%). Промисловий діапазон: $\eta_{scrap} = 0{,}90$–$0{,}95$.

**$\eta_{ore} = 0{,}56$** — визначається двоступінчасто:

| Крок | Розмірність | Значення |
|:---|:---|:---:|
| 1. Масовий вміст Fe у руді: для Fe₂O₃ теоретично $w_{Fe} = 2 \times 56{,}85 / 159{,}69$ | — | ≈ 0,70 |
| 2. Ступінь відновлення: з відновленого Fe ~80% переходить у сталь, ~20% — у шлак у вигляді FeO | — | ≈ 0,80 |
| **Результат:** $\eta_{ore} = w_{Fe} \times \eta_{recovery} = 0{,}70 \times 0{,}80$ | — | **= 0,56** |

Кожен кілограм доданої руди збільшує масу шлаку на $(1 - \eta_{ore}) \cdot m_{ore} = 0{,}44 \cdot m_{ore}$ кг (FeO + пуста порода).

### 2.7. Розрахункові формули: сумішевий охолоджувач

**Охолоджуюча здатність суміші:**
$$\Delta h_{mix} = x_{ore} \cdot \Delta h_{ore} + (1 - x_{ore}) \cdot \Delta h_{scrap} \quad [\text{кДж/кг}]$$

**Маса охолоджувача:**
$$m_{mix} = \frac{\Delta Q \cdot 10^3}{\Delta h_{mix}} \quad [\text{кг}] \qquad (\Delta Q \text{ подається в МДж})$$

**Розбивка за компонентами:**
$$m_{ore} = x_{ore} \cdot m_{mix}, \qquad m_{scrap\,add} = (1 - x_{ore}) \cdot m_{mix}$$

**Оновлені маси шихти та сталі:**
$$m_{charge}^{new} = m_{charge} + m_{mix}$$
$$m_{steel}^{new} = m_{steel} + m_{scrap\,add} \cdot \eta_{scrap} + m_{ore} \cdot \eta_{ore}$$
$$Yield^{new} = \frac{m_{steel}^{new}}{m_{charge}^{new}} \cdot 100\% $$

**Верифікаційний баланс (обов'язковий крок):**
$$Q_{absorbed} = m_{scrap\,add} \cdot \Delta h_{scrap} + m_{ore} \cdot \Delta h_{ore} \approx \Delta Q \cdot 10^3 \quad [\text{кДж}]$$
$$\varepsilon = \frac{|Q_{absorbed} - \Delta Q \cdot 10^3|}{\Delta Q \cdot 10^3} \cdot 100\% < 0{,}1\% \quad \text{(похибка тільки від округлення)}$$

### 2.8. Аналіз чутливості за $x_{ore}$

При фіксованому $\Delta Q$ збільшення частки руди $x_{ore}$:
- **зменшує** $m_{mix}$ — руда ефективніша за масою ($\Delta h_{ore} > \Delta h_{scrap}$);
- **незначно знижує** $Yield^{new}$ — оскільки $\eta_{ore} < \eta_{scrap}$, більше заліза йде в шлак.

Однак вплив на $Yield^{new}$ є **малим** (десяті частки відсоткового пункту), бо маса охолоджувача (~8–17 кг) незначна порівняно з базовою масою шихти (1000 кг).

**Таблиця аналізу чутливості (варіант 20: $\Delta Q = 22{,}63$ МДж, $t_{tap} = 1616$ °C, $m_{steel} = 920{,}60$ кг):**

| $x_{ore}$ | $\Delta h_{mix}$, кДж/кг | $m_{mix}$, кг | $m_{scrap\,add}$, кг | $m_{ore}$, кг | $Yield^{new}$, % |
|:---:|:---:|:---:|:---:|:---:|:---:|
| 0,00 (чистий брухт) | 1324,47 | 17,09 | 17,09 | 0,00 | **92,06** |
| 0,25 | 1693,35 | 13,36 | 10,02 | 3,34 | 91,94 |
| **0,50** | **2062,24** | **10,97** | **5,49** | **5,49** | **91,87** |
| 0,75 | 2431,12 | 9,31 | 2,33 | 6,98 | 91,81 |
| 1,00 (чиста руда) | 2800,00 | 8,08 | 0,00 | 8,08 | **91,77** |

> **Висновок з аналізу:** перехід від чистого брухту до чистої руди скорочує масу охолоджувача в $17{,}09 / 8{,}08 \approx \mathbf{2{,}1}$ рази. При цьому $Yield^{new}$ знижується лише на **0,29 відсоткового пункту** — ефект незначний і компенсується перевагами у масі та логістиці. Вибір $x_{ore}$ визначається технологічними та економічними пріоритетами конкретного цеху.

## 3. Покроковий аналітичний розв'язок (варіант 20)

**Вхідні дані:** $\Delta Q = 22{,}63$ МДж, $t_{tap} = 1616$ °C, $x_{ore} = 0{,}50$, $m_{steel} = 920{,}60$ кг, $m_{charge} = 1000$ кг.

### Крок 1. Прийняти $\Delta Q$ з ПР №4

$$\Delta Q = +22{,}63 \text{ МДж} \quad (>0 \Rightarrow \text{охолоджувач необхідний})$$

### Крок 2. Розрахунок $\Delta h_{scrap}$

$$\Delta h_{scrap} = C_{p,s}(T_m - t_0) + L_m + C_{p,l}(t_{tap} - T_m)$$
$$= 0{,}65 \times (1535 - 20) + 272 + 0{,}836 \times (1616 - 1535)$$
$$= 0{,}65 \times 1515 + 272 + 0{,}836 \times 81 = 984{,}75 + 272{,}00 + 67{,}72 = \mathbf{1324{,}47} \text{ кДж/кг}$$

### Крок 3. Прийняти $\Delta h_{ore}$

$$\Delta h_{ore} = 2800 \text{ кДж/кг} \quad \text{(типова залізна руда, обґрунтування — розд. 2.5)}$$

### Крок 4. Розрахунок $\Delta h_{mix}$ при $x_{ore} = 0{,}50$

$$\Delta h_{mix} = 0{,}50 \times 2800 + 0{,}50 \times 1324{,}47 = 1400{,}00 + 662{,}24 = \mathbf{2062{,}24} \text{ кДж/кг}$$

### Крок 5. Розрахунок мас охолоджувача

**Сумішевий охолоджувач (50% руди + 50% брухту):**
$$m_{mix} = \frac{\Delta Q \cdot 10^3}{\Delta h_{mix}} = \frac{22630}{2062{,}24} = \mathbf{10{,}97} \text{ кг}$$
$$m_{ore} = 10{,}97 \times 0{,}50 = \mathbf{5{,}49} \text{ кг}, \qquad m_{scrap\,add} = 10{,}97 \times 0{,}50 = \mathbf{5{,}49} \text{ кг}$$

**Альтернативи для порівняння:**

| Варіант | Формула | Маса охолоджувача, кг |
|:---|:---|:---:|
| Чистий брухт ($x_{ore} = 0$) | $22630 / 1324{,}47$ | **17,09** |
| Суміш 50/50 ($x_{ore} = 0{,}50$) | $22630 / 2062{,}24$ | **10,97** |
| Чиста руда ($x_{ore} = 1$) | $22630 / 2800{,}00$ | **8,08** |

### Крок 6. Оновлені маси шихти та сталі

$$m_{charge}^{new} = 1000 + 10{,}97 = \mathbf{1010{,}97} \text{ кг}$$
$$m_{steel}^{new} = 920{,}60 + 5{,}49 \times 0{,}92 + 5{,}49 \times 0{,}56 = 920{,}60 + 5{,}05 + 3{,}07 = \mathbf{928{,}72} \text{ кг}$$
$$Yield^{new} = \frac{928{,}72}{1010{,}97} \times 100 = \mathbf{91{,}87}\%$$

### Крок 7. Верифікація теплового балансу

**Верифікація 1 — поглинання тепла охолоджувачем:**

$$Q_{absorbed} = m_{scrap\,add} \cdot \Delta h_{scrap} + m_{ore} \cdot \Delta h_{ore}$$
$$= 5{,}49 \times 1324{,}47 + 5{,}49 \times 2800{,}00 = 7271{,}3 + 15372{,}0 = \mathbf{22643{,}3} \text{ кДж}$$

$$\varepsilon = \frac{|22643{,}3 - 22630{,}0|}{22630{,}0} \times 100\% = \mathbf{0{,}06\%} \quad \checkmark \text{ (лише похибка округлення)}$$

**Верифікація 2 — контрольні діапазони:**

| Контрольна величина | Допустимий діапазон | Результат | Статус |
|:---|:---:|:---:|:---:|
| $\Delta h_{scrap}$, кДж/кг | 1300–1350 | 1324 | ✓ |
| $\Delta h_{ore}$, кДж/кг | 2500–3200 | 2800 | ✓ |
| $m_{scrap\,pure} / m_{ore\,pure}$ | 1,5–2,5 | 2,11 | ✓ |
| $Yield^{new}$, % | 90–93 | 91,87 | ✓ |
| $\varepsilon$, % | < 0,1 | 0,06 | ✓ |

---

## 4. Python-скрипт для варіанта 20

```python
"""Практична робота №5: Розрахунок охолоджувачів — варіант 20."""

# ── ВХІДНІ ДАНІ (з ПР №1–4) ─────────────────────────────────────────────────
variant_no = 20
delta_Q  = 22.63    # МДж — тепловий надлишок (ПР №4)
t_tap    = 1616.0   # °C  — цільова температура випуску (ПР №4)
m_steel  = 920.60   # кг  — маса сталі до охолоджувача (ПР №1)
m_charge = 1000.0   # кг  — початкова маса шихти
x_ore    = 0.50     # —   — частка руди в суміші

# ── МОДЕЛЬНІ КОНСТАНТИ ───────────────────────────────────────────────────────
T0, T_MELT  = 20.0, 1535.0      # °C
CP_S, CP_L  = 0.65, 0.836       # кДж/(кг·К)
L_MELT      = 272.0             # кДж/кг
DH_ORE      = 2800.0            # кДж/кг
ETA_SCRAP   = 0.92              # —
ETA_ORE     = 0.56              # —

# ── РОЗРАХУНОК ───────────────────────────────────────────────────────────────
dh_scrap = CP_S * (T_MELT - T0) + L_MELT + CP_L * (t_tap - T_MELT)
dh_mix   = x_ore * DH_ORE + (1.0 - x_ore) * dh_scrap
m_mix    = (delta_Q * 1000.0) / dh_mix
m_ore    = m_mix * x_ore
m_scrap_add = m_mix * (1.0 - x_ore)

# Альтернативи (тільки брухт / тільки руда)
m_scrap_pure = (delta_Q * 1000.0) / dh_scrap
m_ore_pure   = (delta_Q * 1000.0) / DH_ORE

# Оновлені маси
m_charge_new = m_charge + m_mix
m_steel_new  = m_steel + m_scrap_add * ETA_SCRAP + m_ore * ETA_ORE
Yield_new    = (m_steel_new / m_charge_new) * 100.0

# Верифікація теплового балансу
Q_absorbed = m_scrap_add * dh_scrap + m_ore * DH_ORE   # кДж
eps_pct    = abs(Q_absorbed - delta_Q * 1000.0) / (delta_Q * 1000.0) * 100.0

# ── ВИВІД ────────────────────────────────────────────────────────────────────
SEP = "=" * 70
print(SEP)
print(f"  ПР №5 — ОХОЛОДЖУВАЧІ  (ВАРІАНТ {variant_no})")
print(SEP)
print(f"  Вхідні дані: ΔQ = {delta_Q:.2f} МДж | t_tap = {t_tap:.0f} °C | x_ore = {x_ore:.2f}")
print(f"  Δh_scrap = {dh_scrap:.2f} кДж/кг | Δh_ore = {DH_ORE:.0f} | Δh_mix = {dh_mix:.2f}")
print()
print(f"  Суміш:        m_mix = {m_mix:.2f} кг  (руда {m_ore:.2f} кг + брухт {m_scrap_add:.2f} кг)")
print(f"  Тільки брухт: m = {m_scrap_pure:.2f} кг")
print(f"  Тільки руда:  m = {m_ore_pure:.2f} кг")
print()
print(f"  m_charge_new = {m_charge_new:.2f} кг | m_steel_new = {m_steel_new:.2f} кг")
print(f"  Yield_new    = {Yield_new:.2f} %")
print()
print(f"  Верифікація: Q_absorbed = {Q_absorbed:.1f} кДж | ε = {eps_pct:.2f}%")
print(SEP)
```

## 5. Висновки

1. **Тепловий надлишок $\Delta Q > 0$** є невід'ємним атрибутом автогенного конвертерного процесу: він виникає через технологічний запас тепла та варіабельність складу чавуну. Для варіанта 20 $\Delta Q = 22{,}63$ МДж відповідає потенційному перегріву ванни на ~27 °C без охолоджувача, що є технологічно неприпустимим.

2. **Охолоджуюча здатність брухту** $\Delta h_{scrap}$ лінійно залежить від $t_{tap}$ (нахил $C_{p,l} = 0{,}836$ кДж/(кг·К)) і становить 1200–1490 кДж/кг у промисловому діапазоні 1450–1650 °C. Провідний внесок — нагрів твердої фази (~74%), теплота плавлення (~21%) і перегрів рідини (~5%).

3. **Руда в 2,1 рази ефективніша за масою**, ніж чистий брухт (2800 vs 1324 кДж/кг). Ключова причина — ендотермічна реакція відновлення: Fe₂O₃ + 3C → 2Fe + 3CO ($\Delta H > 0$), що відбирає хімічну енергію безпосередньо у ванни. При цьому відновлене залізо переходить у сталь, тобто руда виконує роль і охолоджувача, і джерела металу.

4. **Вплив $x_{ore}$ на вихід сталі є малим.** Збільшення частки руди від 0 до 1 знижує $Yield^{new}$ лише на **0,29 відсоткового пункту** (92,06% → 91,77% для варіанта 20), оскільки маса охолоджувача (~8–17 кг) незначна відносно базової шихти (1000 кг). Вибір $x_{ore}$ визначається переважно технологічними та логістичними чинниками, а не виходом металу.

5. **Сумішевий підхід** ($x_{ore} = 0{,}30$–$0{,}70$) дозволяє збалансувати масу охолоджувача та складнощі його завантаження у конвертер, що широко застосовується на промислових підприємствах (ПАТ «АрселорМіттал Кривий Ріг», Запоріжсталь тощо).

6. **Верифікаційний крок ($Q_{absorbed} \approx \Delta Q$)** є обов'язковим елементом розрахунку. Похибка 0,06% підтверджує коректність усіх проміжних розрахунків і математичну замкнутість моделі.

7. **Зв'язок з ПР №6.** Уточнені маси $m_{charge}^{new}$, $m_{steel}^{new}$ і додатковий шлак від руди ($m_{ore} \times (1 - \eta_{ore}) = 5{,}49 \times 0{,}44 = 2{,}42$ кг FeO + пустої породи) є вхідними даними для розрахунку шлакового режиму та MgO-насичення.
