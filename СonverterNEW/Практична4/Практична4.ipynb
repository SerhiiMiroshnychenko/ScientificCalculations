# Практична робота №4. Видаткова частина теплового балансу та теплові втрати

## 1) Постановка задачі

Необхідно виконати розрахунок видаткової частини теплового балансу киснево-конвертерної плавки та визначити нерозподілений тепловий надлишок $\Delta Q$. Розрахунок ведеться на базі **1000 кг металевої шихти** ($m_{charge} = 1000$ кг).

Цілями роботи є:
- обчислити фізичну ентальпію рідкої сталі за заданої цільової температури випуску;
- обчислити фізичну ентальпію шлаку;
- розрахувати тепло, що виноситься відхідними газами та механічним пилом (виносами);
- виконати інженерне оцінювання теплових втрат через футерівку та випромінювання;
- визначити сумарний видаток тепла $Q_{out}$ та величину теплового надлишку $\Delta Q = Q_{in} - Q_{out}$.

### Вхідні дані (з результатів ПР №1, 2, 3 для Варіанта 20)

- $m_{steel} = 920{,}60$ кг — маса рідкої сталі;
- $m_{slag} = 142{,}63$ кг — маса шлаку;
- $m_{CO} = 66{,}91$ кг, $m_{CO_2} = 11{,}68$ кг — маси газів;
- $m_{dust} = 15{,}0$ кг — маса пилу;
- $Q_{in} = 1721{,}48$ МДж — загальний прихід тепла.

### Умова Варіанта 20 (ПР №4)

- $t_{tap} = 1620\ ^\circ C$ — цільова температура сталі при випуску.

## 2) Теоретична частина

### 2.1. Термодинамічна основа теплового балансу конвертерної плавки

Тепловий баланс плавки є безпосереднім застосуванням **першого закону термодинаміки** до металургійного агрегату. Для стаціонарного відкритого термодинамічного процесу при постійному тиску (умови, що апроксимують конвертерний процес) перший закон записується у формі балансу ентальпії:

$$\sum H_{in} = \sum H_{out} + \Delta Q$$

де $\sum H_{in}$ — сумарна ентальпія матеріальних потоків і хімічних реакцій (прихідна частина), $\sum H_{out}$ — сумарна ентальпія вихідних продуктів (видаткова частина), $\Delta Q$ — нерозподілений тепловий надлишок.

Фізичний зміст цього рівняння для конвертерного процесу такий: хімічна та фізична енергія, що вноситься рідким чавуном, твердими компонентами шихти та вивільняється під час реакцій окиснення домішок (ПР №3), повністю розподіляється між продуктами плавки. Під «продуктами» розуміються рідка сталь, шлак, відхідні гази, механічний пил, а також неминучі теплові втрати у навколишнє середовище. Якщо після урахування всіх статей видатку залишається ненульова різниця — це тепловий надлишок $\Delta Q$, що безпосередньо впливає на технологічні параметри плавки.

### 2.2. Ентальпія як розрахункова функція стану

При постійному тиску зміна внутрішньої енергії системи дорівнює теплообміну з оточенням, тому зручною функцією стану є **ентальпія** $H = U + pV$. Саме ентальпія, а не внутрішня енергія, є правильною термодинамічною величиною для розрахунку теплових ефектів у металургійних процесах, що відбуваються при атмосферному або близькому до нього тиску.

Питома ентальпія $h(T)$ речовини відносно довільної референтної температури $T_0$ визначається як:

$$h(T) = \int_{T_0}^{T_{sol}} C_{p,s}(T)\, dT + L_m + \int_{T_m}^{T} C_{p,l}(T)\, dT$$

де $C_{p,s}(T)$ і $C_{p,l}(T)$ — теплоємності твердої та рідкої фаз відповідно, $T_{sol}$ — температура плавлення, $L_m$ — питома теплота плавлення (Дж/кг), а остання інтегральна складова — перегрів розплаву вище $T_m$ до технологічно необхідної температури $T$.

**Довідка про референтну температуру.** В інженерних теплових балансах металургійних процесів традиційно обирають $t_0 = 0\ ^\circ C$ як нульову точку відліку (на противагу до стандартної термохімічної температури $25\ ^\circ C$, що використовується в NIST-JANAF). Це узгоджено у всіх наведених формулах. Зміна $t_0$ не впливає на результат $\Delta Q$ за умови єдиного відліку в обох частинах балансу, однак неузгодженість відліку в різних складових балансу є типовою помилкою, що призводить до систематичних похибок.

### 2.3. Структура та статті видаткової частини балансу

Видаткова частина теплового балансу конвертерної плавки включає п'ять основних статей:

#### 2.3.1. Фізична ентальпія рідкої сталі

Це найбільша за величиною стаття витрат — у добре збалансованій промисловій плавці вона становить **70–80%** від загального видатку тепла. Фізична ентальпія металу включає три складові:

1. **Нагрів твердої фази** від $t_0 = 0\ ^\circ C$ до температури солідусу $t_{sol} \approx 1490\ ^\circ C$ — близько 680–700 кДж/кг;
2. **Прихована теплота плавлення** $L_m \approx 250\text{–}272$ кДж/кг (для маловуглецевих сталей);
3. **Перегрів рідкої сталі** від $t_m = 1535\ ^\circ C$ (умовна температура плавлення чистого заліза) до $t_{tap}$ зі швидкістю $C_{p,l} \approx 0{,}836$ кДж/(кг·К).

Лінеаризована інженерна формула, що об'єднує всі три складові відносно $0\ ^\circ C$:

$$h_{steel}(t_{tap}) = 1250 + 0{,}836 \cdot (t_{tap} - 1535) \quad [\text{кДж/кг}]$$

де константа $1250$ кДж/кг є питомою ентальпією рідкої сталі при $t_m = 1535\ ^\circ C$ відносно $0\ ^\circ C$, що акумулює нагрів твердої фази та теплоту плавлення. Коефіцієнт $0{,}836$ відповідає теплоємності рідкого заліза та є добре підтвердженою калориметричною константою [Кудрін, 1989; Kubaschewski & Alcock, 1979].

#### 2.3.2. Фізична ентальпія шлаку

Конвертерний основний шлак є складною гетерогенною системою у псевдобінарному апроксимуванні: $\text{CaO–FeO–SiO}_2$ з домішками $\text{MgO}$, $\text{MnO}$, $\text{Al}_2\text{O}_3$. Ентальпія такої системи включає нагрів твердих оксидних фаз, перетворення кристалічних поліморфних модифікацій (наприклад, $\beta \to \alpha$ кварц у $573\ ^\circ C$), та теплоту шлакоутворення (розчинення компонентів у розплаві, $L_{slag} \approx 300\text{–}400$ кДж/кг).

Відмінністю від металу є відсутність чіткої температури плавлення: шлак переходить в рідкий стан у діапазоні $1250\text{–}1450\ ^\circ C$ залежно від основності $B = \text{CaO}/\text{SiO}_2$ та вмісту $\text{FeO}$. Ефективна теплоємність рідкого шлаку складає $C_{p,l} = 0{,}90\text{–}1{,}00$ кДж/(кг·К). Лінеаризована формула ентальпії шлаку від $0\ ^\circ C$:

$$h_{slag}(t_{slag}) = 209 + 0{,}90 \cdot t_{slag} \quad [\text{кДж/кг}]$$

де коефіцієнт $0{,}90$ — ефективна питома теплоємність рідкої шлакової системи, а константа $209$ кДж/кг — інтегральний внесок нагріву твердих оксидів та теплоти шлакоутворення у діапазоні від $0\ ^\circ C$ до початку рідкого стану (відтворює довідникові дані з похибкою $< 5\%$ у діапазоні $1600\text{–}1700\ ^\circ C$).

Температура шлаку перевищує температуру металу на $30\text{–}60\ ^\circ C$, оскільки найбільш екзотермічні реакції окиснення (наприклад, $\text{Si} + \text{O}_2 \to \text{SiO}_2$, $\Delta H = -910$ кДж/моль) відбуваються на межі поділу метал–шлак, локально нагріваючи шлакову фазу. У розрахунку прийнято $t_{slag} = t_{tap} + 50\ ^\circ C$.

#### 2.3.3. Фізична ентальпія відхідних газів

Під час конвертерного дуття основними газоподібними продуктами є $CO$ та $CO_2$ — продукти окиснення вуглецю розплаву. Гази залишають конвертер через горловину при температурах $1400\text{–}1550\ ^\circ C$.

Теплоємність газів є функцією температури, тому для коректного розрахунку слід застосовувати **середні** (інтегральні) питомі теплоємності від $t_0 = 0\ ^\circ C$ до $t_{gas}$:

$$\bar{C}_p = \frac{1}{T - T_0} \int_{T_0}^{T} C_p(T)\, dT$$

Значення $\bar{C}_{p,CO} = 1{,}17$ кДж/(кг·К) та $\bar{C}_{p,CO_2} = 1{,}15$ кДж/(кг·К) при $t_{gas} = 1500\ ^\circ C$ узгоджуються з даними NIST-JANAF та класичними термохімічними довідниками. Ці значення є **середніми від $0\ ^\circ C$**, а не миттєвими при $1500\ ^\circ C$ (миттєві значення при $1500\ ^\circ C$ для $CO$ становлять $\approx 1{,}18$, для $CO_2$ — $\approx 1{,}31$ кДж/(кг·К) за NIST). Плутанина між середньою та миттєвою теплоємністю є типовою і призводить до систематичного завищення $Q_{gas}$.

#### 2.3.4. Фізична ентальпія пилу (виносів)

Конвертерний пил утворюється внаслідок випаровування та механічного виносу дрібних частинок металу та оксидів газовим потоком з горловини. Хімічний склад пилу близький до суміші заліза, $\text{FeO}$ та $\text{Fe}_2\text{O}_3$. Питома теплоємність такої суміші від $0\ ^\circ C$ до $1500\ ^\circ C$ становить $\bar{C}_{p,dust} \approx 0{,}82$ кДж/(кг·К). Температура пилу приймається рівною температурі газів: $t_{dust} = t_{gas} = 1500\ ^\circ C$.

#### 2.3.5. Теплові втрати в навколишнє середовище

Теплові втрати конвертера складаються з двох складових:

- **Втрати через футерівку** (конвективно-кондуктивний механізм): теплопровідність через шари $\text{MgO-C}$ або периклазохромітових цеглин, вогнетривкого бетону та металевого кожуха. Для типових режимів плавки та сучасних вогнетривких матеріалів ці втрати складають $1{,}5\text{–}2{,}5\%$ від $Q_{in}$.
- **Втрати через випромінювання через горловину** (радіаційний механізм): відповідно до закону Стефана–Больцмана $q = \varepsilon \sigma T^4$, при температурах $> 1600\ ^\circ C$ радіаційні втрати через відкриту горловину можуть становити $1{,}5\text{–}2{,}5\%$ від $Q_{in}$.

Сумарні інтегральні теплові втрати для промислових конвертерів складають **3–5%** від $Q_{in}$. У розрахунку прийнято $k_{loss} = 0{,}045$, що відповідає середнині діапазону та є консервативною (обережною) оцінкою.

### 2.4. Фізичний зміст теплового надлишку $\Delta Q$

Величина $\Delta Q = Q_{in} - Q_{out}$ є ключовим технологічним параметром, що характеризує **теплоенергетичний потенціал плавки**.

**Якщо $\Delta Q > 0$**: система генерує надлишкову ентальпію. Ця енергія може бути використана для розплавлення додаткових холодних матеріалів — насамперед сталевого брухту або залізорудних охолоджувачів (залізної руди, агломерату). Саме величина $\Delta Q$ є вхідним параметром для ПР №5, де розраховується максимально допустима маса охолоджувача при збереженні $t_{tap}$.

**Якщо $\Delta Q < 0$**: тепловий дефіцит свідчить про некоректність або розрахункової моделі (помилки в коефіцієнтах, незгоджений відлік ентальпії), або про те, що задана $t_{tap}$ недосяжна при даному складі шихти. В реальній виробничій практиці ситуація $\Delta Q < 0$ неможлива без зовнішнього підведення тепла, оскільки конвертерний процес є автотермічним.

**Технологічна інтерпретація $\Delta Q$:** у типових умовах промислової плавки з чавуном зі стандартним вмістом кремнію (Si $\approx 0{,}6\text{–}1{,}0\%$) тепловий надлишок складає $50\text{–}150$ МДж/т шихти, що дозволяє засвоювати $10\text{–}25\%$ брухту від маси шихти. Менший $\Delta Q$ (як у нашому прикладі) характерний для нижчих значень Si або вищих цільових температур $t_{tap}$.

## 3) Покроковий аналітичний розв'язок

Відлік ентальпій ведеться від $t_0 = 0\ ^\circ C$.

### Крок 1. Фізична ентальпія рідкої сталі

Використовуємо лінеаризовану формулу ентальпії для рідкої сталі (кДж/кг), що враховує нагрів від $0\ ^\circ C$, плавлення та перегрів розплаву:

$$h_{steel} = 1250 + 0{,}836 \cdot (t_{tap} - 1535) \quad [\text{кДж/кг}]$$

де:
- $1250$ — питома ентальпія рідкої сталі при температурі плавлення ($1535\ ^\circ C$) відносно $0\ ^\circ C$, кДж/кг;
- $0{,}836$ — питома теплоємність рідкої сталі, кДж/(кг·К);
- $t_{tap}$ — цільова температура сталі при випуску, $^\circ C$;
- $1535$ — температура плавлення чистого заліза (прийнята для сталі), $^\circ C$.

Загальне тепло сталі:

$$Q_{steel} = \frac{m_{steel} \cdot h_{steel}}{1000} \quad [\text{МДж}]$$

де:
- $m_{steel}$ — маса рідкої сталі, кг (з ПР №1);
- $h_{steel}$ — питома ентальпія сталі, кДж/кг;
- $1000$ — коефіцієнт перерахунку кДж у МДж.

### Крок 2. Фізична ентальпія шлаку

Приймаємо температуру шлаку на $50\ ^\circ C$ вищою за температуру металу: $t_{slag} = t_{tap} + 50$.

Питома ентальпія конвертерного основного шлаку (кДж/кг), лінеаризована від $0\ ^\circ C$:

$$h_{slag} = 209 + 0{,}90 \cdot t_{slag} \quad [\text{кДж/кг}]$$

де:
- $0{,}90$ — ефективна питома теплоємність шлакової системи $\text{CaO–FeO–SiO}_2$, кДж/(кг·К);
- $209$ — інтегральний коефіцієнт (теплота нагріву та плавлення оксидів), кДж/кг;
- $t_{slag}$ — температура шлаку (прийнята як $t_{tap} + 50$), $^\circ C$.

Загальне тепло шлаку:

$$Q_{slag} = \frac{m_{slag} \cdot h_{slag}}{1000} \quad [\text{МДж}]$$

де:
- $m_{slag}$ — маса шлаку, кг (з ПР №1);
- $h_{slag}$ — питома ентальпія шлаку, кДж/кг.

> **Обґрунтування формули**: для конвертерного основного шлаку з температурою розм'якшення ~$1300\text{–}1400\ ^\circ C$ та теплотою плавлення $L_{slag} \approx 300\text{–}400$ кДж/кг, наведена лінеаризація відтворює довідникові дані [Кудрін, 1989] у діапазоні $1600\text{–}1700\ ^\circ C$ з похибкою менше 5%.

### Крок 3. Витрати тепла з газами

Приймаємо температуру газів на горловині $t_{gas} = 1500\ ^\circ C$.

Використовуємо **середні** питомі теплоємності газів від $t_0 = 0\ ^\circ C$ до $t_{gas} = 1500\ ^\circ C$ (за термохімічними довідниками):

$$\bar{C}_{p,CO} = 1{,}17\ \text{кДж/(кг·К)}, \quad \bar{C}_{p,CO_2} = 1{,}15\ \text{кДж/(кг·К)}$$

$$Q_{gas} = \frac{\left(m_{CO} \cdot \bar{C}_{p,CO} + m_{CO_2} \cdot \bar{C}_{p,CO_2}\right) \cdot t_{gas}}{1000} \quad [\text{МДж}]$$

де:
- $m_{CO}, m_{CO_2}$ — маси компонентів відхідного газу, кг (з ПР №1);
- $\bar{C}_{p,CO}, \bar{C}_{p,CO_2}$ — середні питомі теплоємності газів, кДж/(кг·К);
- $t_{gas}$ — температура газів на горловині (прийнято $1500\ ^\circ C$);
- $1000$ — коефіцієнт перерахунку кДж у МДж.

> Коректні середні значення від $0\ ^\circ C$ до $1500\ ^\circ C$: для $CO$ — $1{,}13\text{–}1{,}17$, для $CO_2$ — $1{,}10\text{–}1{,}15$ кДж/(кг·К).

### Крок 4. Фізична ентальпія пилу (виносів)

Приймаємо температуру пилу $t_{dust} = 1500\ ^\circ C$ та середню теплоємність $\bar{C}_{p,dust} = 0{,}82$ кДж/(кг·К):

$$Q_{dust} = \frac{m_{dust} \cdot \bar{C}_{p,dust} \cdot t_{dust}}{1000} \quad [\text{МДж}]$$

де:
- $m_{dust}$ — маса пилу (виносів), кг (з ПР №1);
- $\bar{C}_{p,dust}$ — середня теплоємність пилу, кДж/(кг·К);
- $t_{dust}$ — температура пилу, $^\circ C$ (прийнято $1500\ ^\circ C$).

### Крок 5. Теплові втрати конвертера

Приймаємо інтегральні теплові втрати (футерівка + випромінювання) на рівні 4,5% від загального приходу енергії:

$$Q_{loss} = k_{loss} \cdot Q_{in} \quad [\text{МДж}]$$

де:
- $k_{loss}$ — коефіцієнт теплових втрат (прийнято $0{,}045$ або $4{,}5\%$);
- $Q_{in}$ — загальний прихід тепла, МДж (з ПР №3).

### Крок 6. Сумарний видаток та тепловий надлишок

$$Q_{out} = Q_{steel} + Q_{slag} + Q_{gas} + Q_{dust} + Q_{loss} \quad [\text{МДж}]$$

$$\Delta Q = Q_{in} - Q_{out} \quad [\text{МДж}]$$

де:
- $Q_{out}$ — сумарний видаток тепла, МДж;
- $Q_{in}$ — сумарний прихід тепла, МДж;
- $\Delta Q$ — тепловий надлишок (позитивний) або дефіцит (від’ємний).

**Критерій адекватності**: частка ентальпії сталі в сумарному видатку:

$$\frac{Q_{steel}}{Q_{out}} \in [0{,}70;\, 0{,}80]$$

## Числовий приклад розрахунку (Варіант 20)

Нижче наведено покроковий розрахунок для Варіанта 20 ($t_{tap} = 1620\ ^\circ C$):

**1. Фізична ентальпія сталі:**

$$h_{steel} = 1250 + 0{,}836 \cdot (t_{tap} - 1535)$$
$$h_{steel} = 1250 + 0{,}836 \cdot (1620 - 1535) = 1250 + 71{,}06 = 1321{,}06\ \text{кДж/кг}$$

$$Q_{steel} = \frac{m_{steel} \cdot h_{steel}}{1000}$$
$$Q_{steel} = \frac{920{,}60 \cdot 1321{,}06}{1000} = 1216{,}17\ \text{МДж}$$

**2. Фізична ентальпія шлаку** ($t_{slag} = 1620 + 50 = 1670\ ^\circ C$):

$$h_{slag} = 209 + 0{,}90 \cdot t_{slag}$$
$$h_{slag} = 209 + 0{,}90 \cdot 1670 = 209 + 1503 = 1712{,}0\ \text{кДж/кг}$$

$$Q_{slag} = \frac{m_{slag} \cdot h_{slag}}{1000}$$
$$Q_{slag} = \frac{142{,}63 \cdot 1712{,}0}{1000} = 244{,}18\ \text{МДж}$$

**3. Витрати тепла з газами** ($t_{gas} = 1500\ ^\circ C$):

$$Q_{gas} = \frac{(m_{CO} \cdot \bar{C}_{p,CO} + m_{CO_2} \cdot \bar{C}_{p,CO_2}) \cdot t_{gas}}{1000}$$
$$Q_{gas} = \frac{(66{,}91 \cdot 1{,}17 + 11{,}68 \cdot 1{,}15) \cdot 1500}{1000} = 137{,}57\ \text{МДж}$$

**4. Ентальпія пилу** ($t_{dust} = 1500\ ^\circ C$):

$$Q_{dust} = \frac{m_{dust} \cdot \bar{C}_{p,dust} \cdot t_{dust}}{1000}$$
$$Q_{dust} = \frac{15{,}0 \cdot 0{,}82 \cdot 1500}{1000} = 18{,}45\ \text{МДж}$$

**5. Теплові втрати** ($k_{loss} = 0{,}045$):

$$Q_{loss} = k_{loss} \cdot Q_{in}$$
$$Q_{loss} = 0{,}045 \cdot 1721{,}48 = 77{,}47\ \text{МДж}$$

**6. Підсумок видаткової частини та надлишок:**

$$Q_{out} = Q_{steel} + Q_{slag} + Q_{gas} + Q_{dust} + Q_{loss}$$
$$Q_{out} = 1216{,}17 + 244{,}18 + 137{,}57 + 18{,}45 + 77{,}47 = 1693{,}84\ \text{МДж}$$

$$\Delta Q = Q_{in} - Q_{out}$$
$$\Delta Q = 1721{,}48 - 1693{,}84 = \mathbf{+27{,}64\ \text{МДж}}$$

**Перевірка критерію адекватності:**

$$\frac{Q_{steel}}{Q_{out}} = \frac{1216{,}17}{1693{,}84} = 71{,}8\%\ \in [70\%;\,80\%] \quad \checkmark$$

> **Висновок з перевірки**: тепловий надлишок є додатнім, критерій частки сталі виконано. Модель фізично коректна.

## 4) Розв'язок за допомогою Python-скрипта

```python
"""Практична робота №4 (Python).
Видаткова частина теплового балансу та теплові втрати.
Варіант 20.
"""

# =========================================================================
# 0) Вхідні дані (Варіант 20)
# =========================================================================
variant_no = 20

# Маси продуктів з ПР №1, кг
m_steel = 920.60
m_slag  = 142.63
m_CO    = 66.91
m_CO2   = 11.68
m_dust  = 15.0

# Загальний прихід тепла з ПР №3, МДж
Q_in = 1721.48

# Цільова температура сталі при випуску, °C
t_tap = 1620.0

# =========================================================================
# 1) Крок 1: Ентальпія рідкої сталі
# =========================================================================
# Лінеаризована формула: враховує нагрів твердої фази, плавлення
# та перегрів розплаву від t0 = 0 °C
h_steel = 1250.0 + 0.836 * (t_tap - 1535.0)   # кДж/кг
Q_steel = m_steel * h_steel / 1000.0            # МДж

# =========================================================================
# 2) Крок 2: Ентальпія конвертерного шлаку
# =========================================================================
# Температура шлаку вища за метал на 50 °C
t_slag = t_tap + 50.0

# Лінеаризована формула для основного шлаку CaO-FeO-SiO2 (від 0 °C):
# h = 209 + 0.90*t [кДж/кг]  (Cp_eff = 0.90 кДж/(кг·К); 209 — внесок
# нагріву та плавлення твердих оксидів)
h_slag = 209.0 + 0.90 * t_slag                 # кДж/кг
Q_slag = m_slag * h_slag / 1000.0              # МДж

# =========================================================================
# 3) Крок 3: Тепло відхідних газів (на горловині, t_gas = 1500 °C)
# =========================================================================
t_gas = 1500.0

# Середні питомі теплоємності від 0 °C до 1500 °C (термохімічні довідники):
# CO   -> 1.17 кДж/(кг·К)
# CO2  -> 1.15 кДж/(кг·К)
cp_CO  = 1.17
cp_CO2 = 1.15

Q_gas = (m_CO * cp_CO + m_CO2 * cp_CO2) * t_gas / 1000.0  # МДж

# =========================================================================
# 4) Крок 4: Ентальпія пилу (виносів)
# =========================================================================
t_dust   = 1500.0
cp_dust  = 0.82   # кДж/(кг·К), середня для металевого пилу + оксидів

Q_dust = m_dust * cp_dust * t_dust / 1000.0    # МДж

# =========================================================================
# 5) Крок 5: Теплові втрати (футерівка + випромінювання)
# =========================================================================
k_loss = 0.045   # 4.5% від Q_in

Q_loss = k_loss * Q_in                          # МДж

# =========================================================================
# 6) Крок 6: Сумарний видаток та тепловий надлишок
# =========================================================================
Q_out   = Q_steel + Q_slag + Q_gas + Q_dust + Q_loss
delta_Q = Q_in - Q_out

# Частка сталі в сумарному видатку (критерій адекватності: 70–80%)
share_steel_pct = Q_steel / Q_out * 100.0

# =========================================================================
# 7) Вивід результатів
# =========================================================================
print("=" * 65)
print(f"  ПРАКТИЧНА РОБОТА №4: ВИДАТОК ТЕПЛА (ВАРІАНТ №{variant_no})")
print("=" * 65)
print(f"  1) Ентальпія сталі   h = {h_steel:.2f} кДж/кг  ->  {Q_steel:.2f} МДж")
print(f"  2) Ентальпія шлаку   h = {h_slag:.2f} кДж/кг  ->  {Q_slag:.2f} МДж")
print(f"  3) Тепло газів       (t_gas = {t_gas:.0f} °C)       {Q_gas:.2f} МДж")
print(f"  4) Тепло пилу        (t_dust = {t_dust:.0f} °C)      {Q_dust:.2f} МДж")
print(f"  5) Теплові втрати    (k = {k_loss})             {Q_loss:.2f} МДж")
print("-" * 65)
print(f"  СУМАРНИЙ ВИДАТОК  Q_out  =  {Q_out:.2f} МДж")
print(f"  ЗАГАЛЬНИЙ ПРИХІД  Q_in   =  {Q_in:.2f} МДж")
print(f"  ТЕПЛОВИЙ НАДЛИШОК dQ     =  {delta_Q:+.2f} МДж")
print("=" * 65)

# Критерій адекватності
print(f"\n  Частка сталі у видатку: {share_steel_pct:.1f}%", end="  ")
if 70.0 <= share_steel_pct <= 80.0:
    print("[OK: 70–80%]")
else:
    print("[УВАГА: поза діапазоном 70–80% — перевірте формули]")

# Висновок щодо балансу
print()
if delta_Q > 0:
    print(f"  РЕЗУЛЬТАТ: Надлишок {delta_Q:.2f} МДж передається до ПР №5.")
    print("  Розрахунок маси охолоджувача (брухту/руди) — наступний крок.")
elif delta_Q == 0:
    print("  РЕЗУЛЬТАТ: Баланс точно замкнений. dQ = 0.")
else:
    print("  УВАГА: Тепловий дефіцит! Перевірте вхідні дані та формули.")
print("=" * 65)
```

$$m_{scrap,max} = \frac{\Delta Q}{\Delta h_{scrap}} \approx \frac{27{,}64 \cdot 10^3}{1225} \approx 22{,}6\ \text{кг на 1000 кг шихти}$$

Сумарна оцінна похибка моделі становить $\pm 5\text{–}8\%$ за значенням $\Delta Q$, що є допустимим для задач технологічного проєктування та статичного розрахунку.

```python
"""
Практична робота №4.
Алгоритм розрахунку для всіх 20 варіантів на мові Python.
"""

def compute_all_variants():
    # Константи методології ПР №4
    H_ST_REF = 1250.0  # Ентальпія рідкої сталі
    CP_ST    = 0.836   # Теплоємність рідкої сталі
    T_M      = 1535.0  # Температура плавлення сталі
    
    H_SL_REF = 209.0   # Початковий член ентальпії шлаку
    CP_SL    = 0.90    # Теплоємність рідкого шлаку
    
    T_GAS    = 1500.0  # Температура газів
    CP_CO    = 1.17    # Теплоємність CO
    CP_CO2   = 1.15    # Теплоємність CO2
    
    T_DUST   = 1500.0  # Температура пилу
    CP_DUST  = 0.82    # Теплоємність пилу
    M_DUST   = 15.0    # Маса пилу
    
    K_LOSS   = 0.045   # Коефіцієнт втрат (4.5%)

    # Вхідні дані: (Вар, m_st, m_sl, m_CO, m_CO2, Q_in, t_tap)
    # m_i — з ПР №1, Q_in — з ПР №3, t_tap — з ПР №4
    data = [
        (1, 939.94, 54.79, 66.13, 11.54, 1419.10, 1580),
        (2, 938.45, 62.52, 66.23, 11.56, 1426.30, 1584),
        (3, 938.09, 61.28, 66.32, 11.58, 1413.50, 1588),
        (4, 936.61, 69.10, 66.41, 11.59, 1436.40, 1592),
        (5, 935.08, 77.35, 66.49, 11.61, 1459.20, 1596),
        (6, 934.94, 74.52, 66.56, 11.62, 1473.70, 1600),
        (7, 933.43, 82.84, 66.62, 11.63, 1496.00, 1604),
        (8, 931.87, 91.60, 66.68, 11.64, 1518.30, 1608),
        (9, 931.94, 87.22, 66.74, 11.65, 1531.10, 1612),
        (10, 930.40, 96.02, 66.78, 11.66, 1552.80, 1616),
        (11, 928.82, 105.25, 66.83, 11.67, 1574.50, 1624),
        (12, 929.09, 99.38, 66.86, 11.67, 1585.70, 1628),
        (13, 927.53, 108.64, 66.89, 11.68, 1606.80, 1632),
        (14, 925.93, 118.30, 66.91, 11.68, 1627.90, 1636),
        (15, 926.39, 111.00, 66.92, 11.68, 1637.40, 1640),
        (16, 924.81, 120.69, 66.93, 11.69, 1657.90, 1644),
        (17, 923.19, 130.76, 66.94, 11.69, 1678.40, 1648),
        (18, 923.84, 122.09, 66.93, 11.68, 1685.70, 1652),
        (19, 922.24, 132.17, 66.92, 11.68, 1703.60, 1656),
        (20, 920.60, 142.63, 66.91, 11.68, 1721.48, 1620),
    ]

    print(f" {'Вар':>3} | {'t_tap':>5} | {'Q_in':>8} | {'Q_out':>8} | {'dQ':>8} | {'%_st':>5}")
    print("-" * 55)
    for v in data:
        v_no, m_st, m_sl, m_CO, m_CO2, Q_in, t_tap = v
        
        # 1. СталЬ
        h_st = H_ST_REF + CP_ST * (t_tap - T_M)
        Q_st = m_st * h_st / 1000.0
        
        # 2. Шлак
        h_sl = H_SL_REF + CP_SL * (t_tap + 50.0)
        Q_sl = m_sl * h_sl / 1000.0
        
        # 3. Гази
        Q_gas = (m_CO * CP_CO + m_CO2 * CP_CO2) * T_GAS / 1000.0
        
        # 4. ПилЬ
        Q_dust = M_DUST * CP_DUST * T_DUST / 1000.0
        
        # 5. Втрати
        Q_loss = K_LOSS * Q_in
        
        # 6. Сума
        Q_out = Q_st + Q_sl + Q_gas + Q_dust + Q_loss
        dQ = Q_in - Q_out
        
        print(f" {v_no:3} | {t_tap:5} | {Q_in:8.2f} | {Q_out:8.2f} | {dQ:+8.2f} | {Q_st/Q_out*100:4.1f}%")

if __name__ == "__main__":
    compute_all_variants()
```
