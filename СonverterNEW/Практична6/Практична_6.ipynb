## 1. Постановка задачі

Необхідно виконати розрахунок параметрів шлакового режиму для забезпечення експлуатаційної стійкості периклазовуглецевої (MgO-C) футерівки кисневого конвертера. Розрахунок ведеться на базі **1000 кг початкової металевої шихти** з використанням результатів матеріального та теплового балансів (ПР №1, ПР №4, ПР №5).

### 1.1. Цілі роботи

- розрахувати рівень термодинамічного насичення конвертерного шлаку оксидом магнію ($MgO_{sat}$) залежно від температури випуску та основності;
- визначити необхідну масу магнезіального флюсу (доломітизованого вапна) для досягнення цільової концентрації $MgO$;
- розрахувати теплові витрати на нагрів і розчинення флюсу;
- виконати перерахунок теплового балансу та визначити скориговану масу сталевого брухту з урахуванням витрат тепла на захист футерівки;
- кількісно оцінити технологічний компроміс між стійкістю футерівки та енергетичною ефективністю плавки.

## 2. Теоретичні положення

### 2.1. Механізм зносу периклазовуглецевої футерівки

Сучасні кисневі конвертери футеруються периклазовуглецевими вогнетривами (MgO-C), які поєднують високу вогнетривкість периклазу ($MgO$) та шлакостійкість графіту. Головним механізмом хімічного зносу футерівки є розчинення $MgO$ в агресивному залізисто-силікатному шлаку.

Швидкість розчинення пропорційна рушійній силі:
$$\Delta C = MgO_{sat} - \%MgO_{fact}$$
де $MgO_{sat}$ — межа розчинності оксиду магнію у шлаку даного складу при заданій температурі. Якщо шлак ненасичений магнезією ($\%MgO_{fact} < MgO_{sat}$), він інтенсивно екстрагує $MgO$ з робочого шару футерівки, що призводить до прискореної ерозії. При досягненні насичення ($\%MgO_{fact} = MgO_{sat}$) термодинамічна рушійна сила зносу зникає і знос практично зупиняється.

### 2.2. Термодинаміка розчинності MgO

Межа насичення $MgO_{sat}$ є функцією кількох параметрів:

1. **Температура ($t_{tap}$):** розчинення $MgO$ у силікатних розплавах — ендотермічний процес. Зі зростанням температури $MgO_{sat}$ збільшується.
2. **Основність ($B = CaO/SiO_2$):** підвищення вмісту $CaO$ знижує термодинамічну активність $SiO_2$, що зменшує спорідненість шлаку до основного оксиду $MgO$. Високоосновні шлаки потребують меншої концентрації магнезії для насичення.
3. **Окиснюваність ($w_{FeO}$):** оксиди заліза розширюють гомогенну рідку область і збільшують розчинність $MgO$, однак у типових конвертерних шлаках вміст $FeO$ в кінці продувки мало варіює між варіантами цієї роботи (15–20%) і поглинається в константний член рівняння.

Для інженерних розрахунків у діапазоні $1440\text{–}1650\ ^\circ C$ застосовується лінійна емпірична залежність, отримана регресією за табличними даними Slag Atlas (VDEh, 1995) для четверної системи $CaO\text{–}SiO_2\text{–}MgO\text{–}FeO$ при типовому вмісті $FeO \approx 15\text{–}20\%$:

$$\boxed{MgO_{sat} = 6{,}0 + 0{,}015 \cdot (t_{tap} - 1550) - 0{,}5 \cdot (B - 3{,}0)} \quad [\%] \tag{1}$$

Коефіцієнти формули мають фізичний зміст: $+0{,}015\ \%/^\circ C$ відображає ендотермічну природу розчинення $MgO$, а $-0{,}5$ на одиницю основності — зменшення термодинамічної спорідненості шлаку до $MgO$ при зростанні активності $CaO$. Перевірка за даними Slag Atlas підтверджує прийнятну точність формули по всьому робочому діапазону: при $t = 1450\ ^\circ C$, $B = 3{,}0$ формула дає $4{,}50\%$ (довідникові значення $4{,}0\text{–}5{,}0\%$); при $t = 1600\ ^\circ C$ — $6{,}75\%$ (довідник: $6{,}5\text{–}7{,}5\%$). Властива будь-якій лінійній регресії похибка не перевищує $\pm 0{,}5\%$.

### 2.3. Вивід формули для маси флюсу

Нехай $m_{MgO}^{req}$ — маса чистого $MgO$, яку необхідно розчинити у шлаку. Запишемо масовий баланс кінцевого шлаку:

- **Маса шлаку після присадки:** $m_{slag}^{base} + m_{MgO}^{req}$
- **Маса $MgO$ у кінцевому шлаку:** $m_{slag}^{base} \cdot \dfrac{[MgO]_{init}}{100} + m_{MgO}^{req}$
- **Умова досягнення цільового вмісту:**

$$\frac{m_{slag}^{base} \cdot \dfrac{[MgO]_{init}}{100} + m_{MgO}^{req}}{m_{slag}^{base} + m_{MgO}^{req}} = \frac{\%MgO_{target}}{100}$$

Розв'яжемо відносно $m_{MgO}^{req}$:

$$m_{slag}^{base} \cdot [MgO]_{init} + 100 \cdot m_{MgO}^{req} = \%MgO_{target} \cdot m_{slag}^{base} + \%MgO_{target} \cdot m_{MgO}^{req}$$

$$m_{MgO}^{req} \cdot (100 - \%MgO_{target}) = m_{slag}^{base} \cdot (\%MgO_{target} - [MgO]_{init})$$

$$\boxed{m_{MgO}^{req} = m_{slag}^{base} \cdot \frac{\Delta\%MgO}{100 - \%MgO_{target}}} \tag{2}$$

де $\Delta\%MgO = \%MgO_{target} - [MgO]_{init}$.

### 2.4. Технологія ошлакування та тепловий компроміс

Для запобігання зносу футерівки в конвертер присаджують доломітизоване вапно (обпалений доломіт), щоб штучно підвищити концентрацію $MgO$ у шлаку до рівня насичення. Коефіцієнт $\beta_{sat} \approx 0{,}9\text{–}1{,}1$ регулює ступінь насичення:
- $\beta_{sat} < 1{,}0$ — часткове насичення (економія флюсу, певний знос допускається);
- $\beta_{sat} = 1{,}0$ — повне насичення (нульова термодинамічна рушійна сила зносу);
- $\beta_{sat} > 1{,}0$ — пересичення (надійний захист, але вища витрата тепла та флюсу).

Введення магнезіальних флюсів супроводжується значними витратами тепла ($\Delta h_{dl} \approx 800$ кДж/кг), що безпосередньо зменшує тепловий надлишок системи $\Delta Q$. Це формує **прямий технологічний компроміс**: кожен доданий кілограм флюсу для захисту футерівки скорочує кількість металобрухту, яку здатен переплавити конвертер.

## 3. Покроковий аналітичний розв'язок (варіант 20)

**Вхідні дані:** $t_{tap} = 1616\ ^\circ C$, $B = 3{,}4$, $m_{slag}^{base} = 142{,}63$ кг, $\Delta Q = 22{,}63$ МДж, $\Delta h_{scrap} = 1324{,}47$ кДж/кг, $\beta_{sat} = 1{,}00$, $[MgO]_{init} = 1{,}50\%$.

### Крок 1. Розрахунок межі насичення шлаку $MgO_{sat}$

$$MgO_{sat} = 6{,}0 + 0{,}015 \cdot (t_{tap} - 1550) - 0{,}5 \cdot (B - 3{,}0)$$

$$MgO_{sat} = 6{,}0 + 0{,}015 \cdot (1616 - 1550) - 0{,}5 \cdot (3{,}4 - 3{,}0)$$

$$MgO_{sat} = 6{,}0 + 0{,}015 \cdot 66 - 0{,}5 \cdot 0{,}4 = 6{,}0 + 0{,}990 - 0{,}200 = \mathbf{6{,}79\%}$$

### Крок 2. Визначення цільового вмісту та дефіциту MgO

Цільова концентрація з урахуванням коефіцієнта $\beta_{sat}$:

$$\%MgO_{target} = MgO_{sat} \cdot \beta_{sat} = 6{,}79 \cdot 1{,}00 = \mathbf{6{,}79\%}$$

Необхідний приріст концентрації з урахуванням фонового $[MgO]_{init}$:

$$\Delta\%MgO = \%MgO_{target} - [MgO]_{init} = 6{,}79 - 1{,}50 = \mathbf{5{,}29\%}$$

### Крок 3. Розрахунок маси магнезіальних флюсів

Маса чистого $MgO$, яку необхідно розчинити у шлаку (формула (2), виведена у розд. 2.3):

$$m_{MgO}^{req} = m_{slag}^{base} \cdot \frac{\Delta\%MgO}{100 - \%MgO_{target}}$$

$$m_{MgO}^{req} = 142{,}63 \cdot \frac{5{,}29}{100 - 6{,}79} = 142{,}63 \cdot \frac{5{,}29}{93{,}21} = 142{,}63 \cdot 0{,}05676 = \mathbf{8{,}095 \text{ кг}}$$

Маса комерційного доломітизованого вапна при $w_{MgO}^{dl} = 0{,}40$:

$$m_{dl} = \frac{m_{MgO}^{req}}{w_{MgO}^{dl}} = \frac{8{,}095}{0{,}40} = \mathbf{20{,}24 \text{ кг}}$$

### Крок 4. Оцінка енергетичних витрат на захист футерівки

Теплова енергія, що витрачається на нагрів і розчинення флюсу:

$$Q_{dl} = \frac{m_{dl} \cdot \Delta h_{dl}}{1000} = \frac{20{,}24 \cdot 800}{1000} = \mathbf{16{,}19 \text{ МДж}}$$

### Крок 5. Перерахунок маси брухту (зв'язок з ПР №5)

Залишковий тепловий надлишок після задоволення потреб шлакового режиму:

$$\Delta Q_{rem} = \Delta Q - Q_{dl} = 22{,}63 - 16{,}19 = \mathbf{6{,}44 \text{ МДж}}$$

Скоригована маса сталевого брухту, яку здатна розплавити плавка:

$$m_{scrap}^{adj} = \frac{\Delta Q_{rem} \cdot 1000}{\Delta h_{scrap}} = \frac{6{,}44 \cdot 1000}{1324{,}47} = \mathbf{4{,}86 \text{ кг}}$$

*(Для порівняння: без додавання доломіту конвертер міг розплавити $17{,}09$ кг брухту (з ПР №5). Плата за стійкість футерівки склала майже 12 кг брухту на тонну шихти.)*

## 4. Python-скрипт для варіанта 20

```python
"""Практична робота №6: Розрахунок шлакового режиму (варіант 20)."""

# ── ВХІДНІ ДАНІ ──────────────────────────────────────────────────────────────
variant_no   = 20
t_tap        = 1616.0    # °C
B            = 3.4       # Основність CaO/SiO2
m_slag_base  = 142.63    # кг (з ПР №1)
delta_Q      = 22.63     # МДж (з ПР №4)
dh_scrap     = 1324.47   # кДж/кг (з ПР №5)
beta_sat     = 1.00      # Цільовий рівень насичення (за варіантом)

# ── МОДЕЛЬНІ КОНСТАНТИ ───────────────────────────────────────────────────────
MgO_init = 1.50          # % — фоновий вміст MgO у шлаку
w_MgO_dl = 0.40          # частка — вміст MgO у доломітизованому вапні
dh_dl    = 800.0         # кДж/кг — ефективна охолоджуюча здатність флюсу

# ── РОЗРАХУНОК ───────────────────────────────────────────────────────────────
# Крок 1: межа насичення за формулою (1), діапазон 1440–1650 °C
MgO_sat    = 6.0 + 0.015 * (t_tap - 1550.0) - 0.5 * (B - 3.0)

# Крок 2: цільовий вміст і дефіцит
MgO_target = MgO_sat * beta_sat
delta_MgO  = MgO_target - MgO_init

# Крок 3: маса флюсу (формула (2), вивід — розд. 2.3)
if delta_MgO > 0:
    m_MgO_req = m_slag_base * (delta_MgO / (100.0 - MgO_target))
    m_dl      = m_MgO_req / w_MgO_dl
    Q_dl      = (m_dl * dh_dl) / 1000.0
else:
    # Шлак вже насичений — присадка флюсу не потрібна
    m_MgO_req, m_dl, Q_dl = 0.0, 0.0, 0.0

# Крок 4–5: тепловий баланс
dQ_rem = delta_Q - Q_dl

# Захист від від'ємного залишку (якщо флюс «з'їдає» весь надлишок)
m_scrap_adj = (dQ_rem * 1000.0) / dh_scrap if dQ_rem > 0 else 0.0

# ── ВИВІД ────────────────────────────────────────────────────────────────────
print("=" * 65)
print(f"  ПР №6 — ШЛАКОВИЙ РЕЖИМ ТА MgO-НАСИЧЕННЯ (ВАРІАНТ {variant_no})")
print("=" * 65)
print(f"  Межа насичення MgO_sat          = {MgO_sat:.2f} %")
print(f"  Цільовий вміст MgO_target       = {MgO_target:.2f} %")
print(f"  Дефіцит MgO                     = {delta_MgO:.2f} %")
print(f"  Необхідна маса чистого MgO      = {m_MgO_req:.3f} кг")
print(f"  Маса доломітизованого вапна     = {m_dl:.2f} кг")
print("-" * 65)
print(f"  Тепло на розчинення флюсу       = {Q_dl:.2f} МДж")
print(f"  Залишковий тепловий надлишок    = {dQ_rem:.2f} МДж")
print(f"  Скоригована маса брухту         = {m_scrap_adj:.2f} кг")
print("=" * 65)
```

## 5. Висновки

1. **Термодинаміка шлаку:** Межа насичення $MgO_{sat}$ чутлива до технологічних параметрів. Висока температура випуску варіанта 20 ($1616\ ^\circ C$) є головним чинником підвищеної розчинності ($6{,}79\%$). Підтримка високої основності ($B = 3{,}4$) частково компенсує цей ефект, знижуючи $MgO_{sat}$ на $0{,}20\%$ порівняно з базовим значенням при $B = 3{,}0$.
2. **Енергетична ціна стійкості:** Присадка $20{,}24$ кг доломітизованого вапна потребує $16{,}19$ МДж теплової енергії — це поглинає $\approx 72\%$ вільного теплового надлишку системи.
3. **Обмеження брухтового режиму:** Здатність плавити холодний брухт знизилася з $17{,}09$ кг до $4{,}86$ кг на тонну шихти. Це доводить, що формування захисного насиченого шлаку є значним обмеженням для використання металобрухту замість рідкого чавуну.

7. Порівняти $m_{scrap}^{adj}$ з початковою розрахованою масою брухту (з ПР №5) і зробити висновок про вплив шлакового режиму.

## 7. Пакетний розрахунок усіх 20 варіантів (Python)

```python
"""Практична робота №6: Пакетний розрахунок для 20 варіантів."""

# ── КОНСТАНТИ ─────────────────────────────────────────────────────────────────
MgO_init = 1.50   # % — фоновий вміст MgO
w_MgO_dl = 0.40   # частка MgO у доломітизованому вапні
dh_dl    = 800.0  # кДж/кг — охолоджуюча здатність флюсу

# ── ВАРІАНТИ: (№, t_tap, B, m_slag, dQ, dh_scrap, beta) ─────────────────────
variants = [
    (1,  1460, 3.2,  54.79, 20.21, 1194.05, 0.90),
    (2,  1452, 3.4,  62.52, 21.79, 1187.36, 0.95),
    (3,  1440, 3.0,  61.28, 20.22, 1177.33, 1.00),
    (4,  1452, 3.2,  69.10, 20.09, 1187.36, 1.05),
    (5,  1460, 3.4,  77.35, 22.62, 1194.05, 1.10),
    (6,  1480, 3.0,  74.52, 22.50, 1210.77, 0.90),
    (7,  1492, 3.2,  82.84, 20.75, 1220.80, 0.95),
    (8,  1500, 3.4,  91.60, 21.47, 1227.49, 1.00),
    (9,  1520, 3.0,  87.22, 21.92, 1244.21, 1.05),
    (10, 1528, 3.2,  96.02, 21.71, 1250.90, 1.10),
    (11, 1536, 3.4, 105.25, 20.83, 1257.59, 0.90),
    (12, 1556, 3.0,  99.38, 21.73, 1274.31, 0.95),
    (13, 1560, 3.2, 108.64, 23.44, 1277.65, 1.00),
    (14, 1568, 3.4, 118.30, 20.79, 1284.34, 1.05),
    (15, 1588, 3.0, 111.00, 22.30, 1301.06, 1.10),
    (16, 1592, 3.2, 120.69, 22.23, 1304.40, 0.90),
    (17, 1596, 3.4, 130.76, 21.65, 1307.75, 0.95),
    (18, 1616, 3.0, 122.09, 22.96, 1324.47, 1.00),
    (19, 1616, 3.2, 132.17, 23.20, 1324.47, 1.05),
    (20, 1616, 3.4, 142.63, 22.63, 1324.47, 1.00),
]

HDR = (f"{'Вар':>3}|{'t_tap':>6}|{'MgOsat%':>7}|{'targ%':>6}|"
       f"{'Δ%':>5}|{'m_dl':>7}|{'Q_dl':>6}|{'dQ_rem':>7}|{'m_scr':>6}")
SEP = "-" * len(HDR)
print(HDR)
print(SEP)

for vno, t_tap, B, m_slag, dQ, dh_scrap, beta in variants:
    # Крок 1
    MgO_sat = 6.0 + 0.015 * (t_tap - 1550) - 0.5 * (B - 3.0)
    # Крок 2
    target  = MgO_sat * beta
    delta   = target - MgO_init

    # Крок 3 — правильна формула: знаменник (100 - target)
    if delta > 0:
        req  = m_slag * (delta / (100.0 - target))
        m_dl = req / w_MgO_dl
        Q_dl = m_dl * dh_dl / 1000.0
    else:
        m_dl, Q_dl = 0.0, 0.0

    # Кроки 4–5
    dQ_rem    = dQ - Q_dl
    m_scr_adj = (dQ_rem * 1000.0) / dh_scrap if dQ_rem > 0 else 0.0

    print(f"{vno:3}|{t_tap:6.0f}|{MgO_sat:7.2f}|{target:6.2f}|"
          f"{delta:5.2f}|{m_dl:7.2f}|{Q_dl:6.2f}|{dQ_rem:7.2f}|{m_scr_adj:6.2f}")

print(SEP)
```
