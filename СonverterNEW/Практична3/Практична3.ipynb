# Практична робота №3. Прихідна частина теплового балансу конвертера

## 1. Постановка задачі

Необхідно виконати розрахунок прихідної частини теплового балансу киснево-конвертерної плавки на основі еталонної бази **1000 кг металевої шихти**.

Обчисленню підлягають:
- фізично акумульоване тепло рідкого чавуну (сенситивна ентальпія);
- хімічне тепло, яке виділяється внаслідок екзотермічних реакцій окиснення домішок ($Si$, $Mn$, $P$, $C$) та частини заліза ($Fe$);
- вплив ступеня допалювання монооксиду вуглецю ($PCR$) на загальний хімічний прихід тепла;
- загальний прихід тепла $Q_{in}$ та масові частки внеску кожного компонента.

### 1.1. Вхідні дані

**З Практичної роботи №1 (результати матеріального балансу):**
- $\Delta m_{Si},\, \Delta m_{Mn},\, \Delta m_P,\, \Delta m_C$, кг — маси елементів, що окиснилися;
- $m_{Fe \to slag}$, кг — маса заліза, що перейшла у шлакову фазу;
- $m_{hm}$, кг — маса рідкого чавуну в завалці.

**З Практичної роботи №2:**
- $PCR$ — ступінь допалювання монооксиду до діоксиду вуглецю над розплавом (частка одиниці).

**Додатковий параметр (варіант ПР №3):**
- $t_{hm}$, °C — початкова температура заливання рідкого чавуну.

### 1.2. Прийняті термодинамічні константи

Інтегральні теплові ефекти окиснення (наведені до умов сталеплавильного виробництва — від розчиненого в рідкому металі елемента до оксиду в шлаку або газовій фазі):

| Елемент | Реакція окиснення | $\Delta H°$, кДж/моль | $q_i$, кДж/кг | Продукт | Фаза |
|:---:|:---|:---:|:---:|:---:|:---:|
| Si | $[Si] + \{O_2\} \to (SiO_2)$ | −910 | 32 157 | SiO₂ | Шлак |
| P | $2[P] + \tfrac{5}{2}\{O_2\} \to (P_2O_5)$ | −1494 | 24 052 | P₂O₅ | Шлак |
| C → CO₂ | $[C] + \{O_2\} \to \{CO_2\}$ | −394 | 32 803 | CO₂ | Газ |
| C → CO | $[C] + \tfrac{1}{2}\{O_2\} \to \{CO\}$ | −110 | 9 158 | CO | Газ |
| Mn | $[Mn] + \tfrac{1}{2}\{O_2\} \to (MnO)$ | −362 | 6 589 | MnO | Шлак |

Вони враховують стандартні ентальпії утворення, енергію розчинення елементів у рідкому залізі та зміну ентальпії продуктів реакції. У промислових розрахунках іноді використовують округлені емпіричні значення (наприклад, 7000 для Mn), але в даній роботі ми дотримуємося термодинамічної узгодженості.

## 2. Теоретичні засади та термодинамічне обґрунтування

### 2.1. Кисневий конвертер як автогенний сталеплавильний агрегат

Кисневий конвертер (процес **LD/BOS** — від англ. *Basic Oxygen Steelmaking*) є провідним первинним агрегатом сталеплавильного виробництва. Станом на сьогодні **понад 67–70%** світового виробництва сталі-сирцю здійснюється саме у конвертерах з верхнім або комбінованим кисневим дуттям. Процес був винайдений у 1948 р. швейцарським інженером Робертом Дюррером та вперше промислово реалізований у 1952–1953 рр. на заводах VOEST і ÖAMG у Лінці та Донавіці (Австрія).

Принципова відмінність конвертера від електродугових печей (EAF) — **відсутність зовнішнього підводу енергії**. Процес є **автогенним**: уся теплова енергія, необхідна для розплавлення брухту, нагрівання шлаку та продуктів реакцій, формується всередині агрегату за рахунок двох джерел:

- **фізичного тепла рідкого чавуну** — сенситивної теплоємності, акумульованої в гарячому металі з доменного процесу;
- **хімічного тепла екзотермічних реакцій окиснення** домішок (C, Si, Mn, P) та частини заліза технічним киснем чистотою ≥99.5%.

Типова місткість сучасного конвертера — 150–400 т металевої шихти. Тривалість продувки — 15–25 хвилин, що робить його у 20–30 разів продуктивнішим за мартенівську піч.

### 2.2. Закон збереження енергії та структура теплового балансу

Тепловий баланс конвертера — це кількісне відображення **першого закону термодинаміки** (закону Гесса) стосовно всіх теплових потоків у системі:

$$Q_{in} = Q_{out}$$

де $Q_{in}$ — сума всіх джерел теплового приходу (МДж), $Q_{out}$ — сума всіх статей витрат тепла (МДж). Прихідна частина:

$$Q_{in} = Q_{phys} + Q_{chem}$$

Витратна частина (тема окремої роботи) включає: теплоємність рідкої сталі, тепловміст шлаку, фізичне та хімічне тепло відхідних газів, теплові втрати через водоохолоджувані елементи та футерівку.

### 2.3. Фізичне тепло рідкого чавуну

#### 2.3.1. Термодинамічна природа та ентальпія

Рідкий чавун, що надходить із доменної печі при температурі 1250–1450 °C, є основним носієм теплової енергії в конвертерному процесі. У строгому термодинамічному формулюванні для багатокомпонентного розплаву (Fe–C–Si–Mn–P–S):

$$h_{hm} = \int_{T_{ref}}^{T_{hm}} C_p(T)\, dT + L_f$$

де $C_p(T)$ — ізобарна теплоємність (залежна від температури та складу, кДж/(кг·К)), $L_f$ — прихована теплота плавлення заліза (≈272 кДж/кг). Базовою температурою відліку ($T_{ref}$) у металургійних розрахунках зазвичай приймають **0 °C**.

Для типових складів доменного чавуну ([C]≈4%, [Si]≈0.5–1.4%) питома теплоємність $C_p$ не є постійною. Вона зростає зі збільшенням вмісту кремнію, що можна описати емпіричною залежністю:

$$\boxed{C_p = 0.80 + 0.04 \cdot [Si]_{hm} \quad [\text{кДж/(кг·К)}]}$$

Тоді питома ентальпія чавуну:

$$\boxed{h_{hm} = C_p \cdot t_{hm} + 20 \quad [\text{кДж/кг}]}$$

Константа 20 кДж/кг є емпіричною поправкою, що враховує теплоту плавлення та нелінійність ентальпії при базовій температурі відліку $T_{ref} = 0$ °C. Формула дійсна в діапазоні 1200–1500 °C і є **інженерною апроксимацією**, а не строгою термодинамічною залежністю.

#### 2.3.2. Вплив температури чавуну на баланс

Зі зростанням $t_{hm}$ на 10 °C питома ентальпія збільшується на ≈8.5–9.0 кДж/кг, а загальне фізичне тепло зростає на 6–7 МДж/т шихти. Зниження температури рідкого чавуну під час транспортування від доменної печі до конвертера (1–3 °C/хв у відкритих ковшах) є одним з основних джерел прихованих теплових втрат. Тому на передових підприємствах впроваджують **теплоізоляційні кришки торпедних ковшів** та скорочують час транспортування.

На частку фізичного тепла чавуну припадає **50–65%** від загального приходу $Q_{in}$, що підтверджують результати розрахунку в даній роботі (для варіантів 1–20 ця частка становить 50.2–60.6%). Фізичне тепло формує *тепловий буфер* плавки.

### 2.4. Хімічне тепло екзотермічних реакцій окиснення

#### 2.4.1. Термохімія окиснення кремнію

Окиснення кремнію є **найпотужнішою** одиничною екзотермічною реакцією в конвертері. Реакція відбувається практично повністю в перші 2–4 хвилини продувки:

$$[Si] + \{O_2\} \to (SiO_2), \quad \Delta H° \approx -910 \text{ кДж/моль} = -32\,157 \text{ кДж/кг}$$

Висока від'ємна зміна вільної енергії Гіббса ($\Delta G°$ при 1600 °C ≈ −700 кДж/моль) забезпечує практично повний ступінь перетворення: кінцевий вміст Si в металі ≤0.01–0.03%.

Підвищення $[Si]_{hm}$ з 0.45% до 1.40% збільшує $Q_{Si}$ вдвічі (з 130 до 338 МДж/т). Підприємства з висококременистими чавунами отримують значну перевагу за рахунок збільшення частки дешевого металобрухту в шихті.

#### 2.4.2. Термохімія окиснення вуглецю та параметр PCR

Вуглець може утворювати два оксиди, термодинамічна рівновага між якими залежить від температури та $p_{O_2}$:

$$[C] + \tfrac{1}{2}\{O_2\} \to \{CO\}, \quad \Delta H° = -110 \text{ кДж/моль} = -9\,192 \text{ кДж/кг}$$

$$[C] + \{O_2\} \to \{CO_2\}, \quad \Delta H° = -394 \text{ кДж/моль} = -32\,730 \text{ кДж/кг}$$

$$\{CO\} + \tfrac{1}{2}\{O_2\} \to \{CO_2\}, \quad \Delta H° = -283 \text{ кДж/моль} \quad \text{(реакція допалювання)}$$

У ванні конвертера (T ≈ 1550–1700 °C, $p_{O_2} \ll 1$ атм) вуглець окиснюється переважно до CO. Однак газ CO піднімається у надмундштуковий простір, де може контактувати з **вторинним дуттям** і допалюватися до CO₂.

**Ступінь допалювання PCR** (*Post-Combustion Ratio*) визначається як:

$$PCR = \frac{[\%CO_2]}{[\%CO_2] + [\%CO]}$$

Промислові значення: **PCR = 0.07–0.22** для традиційних конвертерів. Ефективний тепловий ефект вуглецю розраховується з урахуванням **ефективності теплопередачі ($\eta_{PCR}$)**, яка показує, яка частина тепла від допалювання газів повертається у ванну (реально 30–90%, у цій роботі прийнято **85%**):

У реальних промислових умовах значення $\eta_{PCR}$ сильно залежить від конструкції фурми (наявність вторинних сопел), висоти занурення та технології дуття, коливаючись у межах **0.30–0.90**. У цій роботі ми використовуємо спрощену модель з постійним $\eta_{PCR} = 0.85$, що характерно для оптимізованих сучасних систем.

$$\boxed{q_C^{avg} = q_{C \to CO} + PCR \cdot \eta_{PCR} \cdot (q_{C \to CO_2} - q_{C \to CO})}$$

При $\eta_{PCR} = 0.85$ зміна PCR від 0.07 до 0.20 збільшує ефективний тепловий ефект вуглецю з ~10 560 до ~13 180 кДж/кг (**+25%**). Це робить PCR найважливішим технологічним важелем управління тепловим балансом.

> **Обмеження.** Коефіцієнт теплопередачі від допалених газів (CO₂) назад до ванни становить реально 30–90% залежно від конструкції фурми. У цій роботі ми використовуємо спрощену модель з прийнятим значенням $\eta_{PCR} = 0.85$. Слід розуміти, що це **оптимістичний сценарій** для сучасних систем допалювання; для звичайних конвертерів без спеціальних фурм реальне значення часто становить **0.50–0.65**.

#### 2.4.3. Термохімія окиснення марганцю та фосфору

**Марганець** окиснюється за реакцією $[Mn] + \tfrac{1}{2}\{O_2\} \to (MnO)$ з тепловим ефектом ≈6 589 кДж/кг. Процес є зворотним: на початку продувки Mn активно переходить у шлак, але при підвищенні температури частина MnO реверсує назад у метал. Ступінь вилучення Mn у шлак — зазвичай 55–70%.

**Фосфор** є найшкідливішою домішкою (спричиняє холодноламкість при $[P] > 0.03\%$). Реакція $2[P] + \tfrac{5}{2}\{O_2\} \to (P_2O_5)$ з ΔH ≈ −25 000 кДж/кг ефективна при низьких температурах (до 1600 °C) та основних шлаках ($V = CaO/SiO_2 > 2.5$). Незважаючи на значний питомий тепловий ефект, внесок $Q_P$ у загальний баланс малий через незначну масу окисненого фосфору.

#### 2.4.4. Окиснення заліза до вюституту FeO

Реакція $[Fe] + \tfrac{1}{2}\{O_2\} \to (FeO)$ з тепловим ефектом ≈4 799 кДж/кг (ефективне значення) є **супутнім процесом**, який відображає окислювальний потенціал шлаку, а не цільову реакцію отримання тепла. Внесок $Q_{Fe}$ у баланс супроводжується безпосередніми втратами виходу придатного заліза у шлак, тому технологічно його намагаються мінімізувати. Підвищений вміст FeO прискорює дефосфорацію, але знижує економічну ефективність плавки.

### 2.5. Порядок та кінетика реакцій окиснення

Порядок термодинамічної пріоритетності визначається від'ємністю $\Delta G°$ кожної реакції при типових умовах початку продувки (T ≈ 1400 °C):

| Пріоритет | Елемент | $\Delta G°$, кДж/моль | Коментар |
|:---:|:---:|:---:|:---|
| 1-й | Si | ≈ −630 | Окиснюється майже повністю за перші 2–4 хв |
| 2-й | P | ≈ −430 | Ефективно при T < 1600 °C та основному шлаку |
| 3-й | C | ≈ −250 | Домінує через велику концентрацію в металі |
| 4-й | Mn | ≈ −200 | Зворотна реакція при підвищенні T |
| 5-й | Fe | ≈ −125 | Формує окислювальний потенціал шлаку |

На практиці декарбурація є домінантним джерелом тепла протягом усієї основної фази продувки, незважаючи на нижчий термодинамічний пріоритет, — завдяки величезній концентрації вуглецю в металі (~4%) та газовій бульбашковій турбулізації ванни.

### 2.6. Контроль адекватності результатів

Добуток маси окисненого елемента (кг) на питомий тепловий ефект (кДж/кг) дає **кДж**. Ділення на 1000 переводить результат у **МДж** — стандартну одиницю для теплового балансу плавки.

**Еталонні діапазони для самоконтролю (на 1000 кг металевої шихти):**

| Величина | Еталонний діапазон |
|:---|:---:|
| $Q_{phys}$ — фізичне тепло чавуну | 800–1000 МДж |
| $Q_{chem}$ — хімічне тепло | 550–900 МДж |
| $Q_{in}$ — загальний прихід | 1400–1900 МДж |
| Частка $Q_{phys}$ від $Q_{in}$ | 50–65% |
| Частка $Q_C$ від $Q_{chem}$ | 45–70% |
| Частка $Q_{Si}$ від $Q_{chem}$ | 20–40% (залежить від $[Si]_{hm}$) |

## 3. Покрокова методика розрахунку

Усі рівняння подаються з відповідним масштабуванням до розмірності МДж.

### Крок 1. Фізичне тепло чавуну

**Питома теплоємність чавуну (залежно від складу):**

$$C_p = 0.80 + 0.04 \cdot [Si]_{hm} \quad [\text{кДж/(кг·К)}]$$

де $[Si]_{hm}$ — початковий вміст кремнію в рідкому чавуні, % (згідно з варіантом ПР №1).

**Питома ентальпія рідкого чавуну (від 0 °C):**

$$h_{hm} = C_p \cdot t_{hm} + 20 \quad [\text{кДж/кг}]$$

де:
- $h_{hm}$ — питома ентальпія рідкого чавуну, кДж/кг;
- $t_{hm}$ — температура заливання чавуну, °C;
- $20$ — емпірична константа, що враховує теплоту плавлення та нелінійність.

**Загальне фізичне тепло:**

$$Q_{phys} = \frac{m_{hm} \cdot h_{hm}}{1000} \quad [\text{МДж}]$$

де:
- $Q_{phys}$ — загальне фізичне тепло чавуну в завалці, МДж;
- $m_{hm}$ — маса рідкого чавуну (результат ПР №1), кг;
- $1000$ — коефіцієнт перерахунку кДж у МДж.

### Крок 2. Хімічне тепло окиснення елементів (Si, Mn, P, Fe)

Для кожного елемента $i \in \{Si,\, Mn,\, P,\, Fe\}$:

$$Q_i = \frac{\Delta m_i \cdot q_i}{1000} \quad [\text{МДж}]$$

де:
- $Q_i$ — теплова енергія від окиснення елемента $i$, МДж;
- $\Delta m_i$ — маса окисненого елемента (результат ПР №1), кг;
- $q_i$ — питомий тепловий ефект реакції (згідно з таблицею 1.2), кДж/кг.

### Крок 3. Хімічне тепло вуглецю (з урахуванням PCR та η_PCR)

**Ефективний середній тепловий ефект вуглецю:**

$$q_C^{avg} = q_{C \to CO} + PCR \cdot \eta_{PCR} \cdot (q_{C \to CO_2} - q_{C \to CO}) \quad [\text{кДж/кг}]$$

де:
- $q_C^{avg}$ — середній тепловий ефект окиснення 1 кг вуглецю, кДж/кг;
- $PCR$ — ступінь допалювання (результат ПР №2), частка одиниці;
- $\eta_{PCR}$ — ефективність теплопередачі допалювання (прийнято 0.85);
- $q_{C \to CO}, q_{C \to CO_2}$ — питомі теплові ефекти неповного та повного окиснення (згідно з табл. 1.2), кДж/кг.

**Теплова енергія від окиснення вуглецю:**

$$Q_C = \frac{\Delta m_C \cdot q_C^{avg}}{1000} \quad [\text{МДж}]$$

де:
- $Q_C$ — загальне тепло від окиснення вуглецю з урахуванням допалювання, МДж;
- $\Delta m_C$ — загальна маса окисненого вуглецю (результат ПР №1), кг.

### Крок 4. Загальні показники

$$Q_{chem} = Q_{Si} + Q_{Mn} + Q_P + Q_C + Q_{Fe} \quad [\text{МДж}]$$

де:
- $Q_{chem}$ — сумарна генерація хімічного тепла під час продувки, МДж.

$$Q_{in} = Q_{phys} + Q_{chem} \quad [\text{МДж}]$$

де:
- $Q_{in}$ — сумарна прихідна частина теплового балансу конвертера, МДж.

### 5.1. Розрахунок для варіанта №20 

```python
"""Практична робота №3: Прихідна частина теплового балансу.
Покроковий розрахунок для варіанта №20.
"""

# 1) Вхідні дані (варіант 20)
variant_no = 20

# Початкові умови та результати з попередніх робіт
m_hm = 700.0          # маса рідкого чавуну, кг
Si_hm = 1.40          # вміст кремнію у чавуні, % (з ПР №1)
t_hm = 1440.0         # температура чавуну, °C
PCR = 0.150           # ступінь допалювання, частка од.

# Маси окиснених елементів (з ПР №1, кг)
dSi = 10.504          
dMn = 5.179           
dP = 0.202            
dC = 31.879           
m_Fe_to_slag = 16.631 

# 2) Термодинамічні константи (кДж/кг)
q_Si = 32157.0
q_Mn = 6589.0
q_P = 24052.0
q_Fe = 4799.0
q_C_CO = 9158.0
q_C_CO2 = 32803.0
eta_PCR = 0.85        # ефективність теплопередачі допалювання

print("-" * 60)
print(f"РОЗРАХУНОК ТЕПЛОВОГО БАЛАНСУ (ПРИХІД) — ВАРІАНТ №{variant_no}")
print("-" * 60)

# 3) Крок 1. Фізичне тепло чавуну
cp_hm = 0.80 + 0.04 * Si_hm
h_hm = cp_hm * t_hm + 20.0
Q_phys = (m_hm * h_hm) / 1000.0

print(f"1. Питома теплоємність Cp: {cp_hm:.3f} кДж/(кг·К)")
print(f"2. Ентальпія чавуну h_hm: {h_hm:.1f} кДж/кг")
print(f"3. Фізичне тепло чавуну Q_phys: {Q_phys:.2f} МДж")

# 4) Крок 2. Хімічне тепло елементів (Si, Mn, P, Fe)
Q_Si = (dSi * q_Si) / 1000.0
Q_Mn = (dMn * q_Mn) / 1000.0
Q_P  = (dP * q_P) / 1000.0
Q_Fe = (m_Fe_to_slag * q_Fe) / 1000.0

print(f"5. Тепло окиснення Si: {Q_Si:.2f} МДж")
print(f"6. Тепло окиснення Mn: {Q_Mn:.2f} МДж")
print(f"7. Тепло окиснення P:  {Q_P:.2f} МДж")
print(f"8. Тепло окиснення Fe (в FeO): {Q_Fe:.2f} МДж")

# 5) Крок 3. Тепло вуглецю з урахуванням PCR та η_PCR
q_C_avg = q_C_CO + PCR * eta_PCR * (q_C_CO2 - q_C_CO)
Q_C = (dC * q_C_avg) / 1000.0

print(f"9. Ефективний тепловий ефект вуглецю q_C_avg: {q_C_avg:.1f} кДж/кг")
print(f"10. Тепло окиснення C: {Q_C:.2f} МДж")

# 6) Крок 4. Загальні показники
Q_chem = Q_Si + Q_Mn + Q_P + Q_C + Q_Fe
Q_in = Q_phys + Q_chem

print("-" * 60)
print(f"11. СУМАРНЕ ХІМІЧНЕ ТЕПЛО Q_chem: {Q_chem:.2f} МДж")
print(f"12. ЗАГАЛЬНИЙ ПРИХІД ТЕПЛА Q_in: {Q_in:.2f} МДж")
print("-" * 60)

# Контрольні частки
p_phys = (Q_phys / Q_in) * 100
p_C = (Q_C / Q_chem) * 100
p_Si = (Q_Si / Q_chem) * 100

print(f"ЧАСТКА ФІЗИЧНОГО ТЕПЛА: {p_phys:.1f} %")
print(f"ВНЕСОК C В ХІМІЧНЕ ТЕПЛО: {p_C:.1f} %")
print(f"ВНЕСОК Si В ХІМІЧНЕ ТЕПЛО: {p_Si:.1f} %")
print("-" * 60)
```

## 8. Висновки

### 8.1. Ключові результати та їх фізична інтерпретація

**1. Домінуюча роль фізичного тепла чавуну.** Фізичне тепло рідкого чавуну ($Q_{phys} = 860$–$871$ МДж) становить **50–61%** загального приходу тепла. Це підтверджує концепцію конвертера як автогенного агрегату: без безперервного постачання гарячого (T ≥ 1250 °C) рідкого чавуну жоден конвертер не може функціонувати. Охолодження чавуну лише на 50 °C під час транспортування знижує $Q_{phys}$ приблизно на 35–40 МДж, що еквівалентно здатності розплавити близько 25 кг холодного брухту. Звідси критична важливість *теплоізоляції торпедних ковшів* та мінімізації часу між доменним і конвертерним відділеннями.

**2. Вуглець — основне джерело хімічного тепла.** Внесок $Q_C$ у хімічний баланс становить **45–65%** від $Q_{chem}$. При збільшенні $PCR$ лише з 0.10 до 0.15 тепловий ефект окиснення одного кілограма вуглецю зростає з ~11 168 до ~12 173 кДж/кг (+9%). Маса окисненого вуглецю практично стала (≈31.5–31.9 кг) по всіх варіантах, що відображає стабільний початковий вміст C у доменному чавуні (≈4.2%).

**3. Кремній визначає «теплову потужність» конвертера.** $Q_{Si}$ зростає від 130 до 338 МДж (більш ніж вдвічі) при збільшенні $[Si]_{hm}$ з 0.45% до 1.40%. Різниця у 208 МДж між варіантом 1 і 20 теоретично відповідає можливості розплавити додатково ≈125–160 кг брухту на тонну шихти. Підприємства з висококременистими чавунами отримують суттєву конкурентну перевагу за рахунок збільшення частки дешевого металобрухту.

**4. PCR як стратегічний регулятор без зміни шихти.** Ступінь допалювання CO є найгнучкішим параметром управління тепловим балансом. Промислові значення PCR обмежені 0.20–0.22 через кінетичні та теплотехнічні обмеження. У цій моделі ми врахували ефективність теплопередачі ($\eta_{PCR} = 0.85$), що робить розрахунок більш реалістичним у порівнянні з ідеальними термодинамічними моделями.

**5. Стабільність $Q_{phys}$ як наслідок компенсаційного проектування.** Відносна незмінність фізичного тепла по варіантах відображає реальну практику виробництва: *теплова потужність плавки* підтримується постійною шляхом варіювання часток гарячого металу та холодного брухту залежно від теплового стану кожної плавки.

### 8.2. Інженерні застосування

- Результати прихідної частини балансу безпосередньо визначають **максимально можливу частку брухту** в шихті: для варіанта 20 (≈1735 МДж) потенціал — близько 30%, для варіанта 1 (≈1419 МДж) — близько 20%.
- При $[Si] > 1.0\%$ може знадобитися **подвійне скачування шлаку** або використання конвертерів більшої місткості.
- Математична модель є базою для **статичних комп'ютерних моделей** конвертерного процесу, що розраховують оптимальну масу кисню, вапна та охолоджувачів у реальному часі для кожної плавки.
- Підвищення температури чавуну на 10 °C (≈+6–7 МДж) та підвищення $PCR$ на 0.003–0.005 (≈+6–7 МДж) є **взаємозамінними** важелями в рамках управління тепловим балансом.

### 8.3. Обмеження застосованої моделі

- Емпірична залежність теплоємності $C_p$ враховує лише вміст кремнію; вплив інших елементів ($C$, $Mn$, $P$, $S$) та температурна нелінійність у цій наближеній моделі залишаються поза розглядом.
- Модель не враховує ряд другорядних джерел теплового приходу: **теплоту шлакоутворення** (екзотермічний ефект взаємодії оксидів, наприклад $CaO + SiO_2$), фізичне тепло дуття та теплоємність шлакоутворюючих матеріалів (вапна, флюсів), які вносяться холодними.
- Коефіцієнт теплопередачі від допалених газів до ванни ($\eta_{PCR}$) прийнятий рівним 0.85 (оптимістичний сценарій). У реальній практиці значення 0.50–0.60 є більш поширеними, що дещо знижує фактичний прихід тепла від вуглецю.
- Модель не розглядає **витратну частину балансу** (теплоємність сталі, шлаку, газів, теплові втрати), без якої неможливо оцінити кінцеву температуру сталі на випуску.
- Для більш точного розрахунку рекомендується використання **FactSage**, **SLAG-Opt** або **HSC Chemistry** — програмних пакетів для термодинамічного моделювання.

> **Загальний висновок.** Прихідна частина теплового балансу конвертера формується двома принципово різними за природою джерелами: фізичне тепло чавуну (≈50–61%) та хімічне тепло реакцій окиснення (≈39–50%). Вуглець та кремній разом забезпечують 80–90% хімічного тепла. Параметр $PCR$ є ключовим важелем управління без зміни складу шихти. Отримані значення $Q_{in} = 1374$–$1721$ МДж/т для всіх 20 варіантів перебувають у фізично обґрунтованому діапазоні та відповідають даним промислових конвертерів.
