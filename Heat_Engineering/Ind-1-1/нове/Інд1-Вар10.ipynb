### 1. Загальні положення та опис об'єкту дослідження

Дугова сталеплавильна піч (ДСП) — це сучасний металургійний агрегат, що використовується для виплавки сталі з металевого брухту та первинної сировини. В сучасних конструкціях ДСП з метою інтенсифікації технологічного процесу та продовження кампанії печі замість традиційної вогнетривкої цегляної футерівки стін і зводу широко застосовують трубчасті водоохолоджувані захисні панелі.

Використання водоохолоджуваних елементів має такі переваги:
- Істотне зниження питомих витрат вогнетривких матеріалів.
- Збільшення терміну служби огороджувальних конструкцій печі.
- Скорочення простоїв обладнання через проведення поточних та капітальних ремонтів.
- Можливість форсування теплових режимів плавки.

Незважаючи на збільшення втрат теплоти з охолоджувальною водою, це рішення є економічно та технологічно виправданим у сучасній металургійній практиці.

### 3. Тепловий та гідравлічний розрахунок

#### 3.1 Визначення середньої температури води в панелі
Середня температура води обчислюється як середнє арифметичне температур на вході та виході з контуру:
$$t_{av} = \frac{t_{out} + t_{in}}{2}$$

Підставимо чисельні значення:
$$t_{av} = \frac{55 + 20}{2} = 37,5\ ^\circ\text{C}$$

#### 3.2 Визначення коефіцієнта тепловіддачі
Згідно з рівнянням Ньютона-Ріхмана для конвективного теплообміну:
$$q = \alpha \cdot (t_1 - t_{av}) \implies \alpha = \frac{q}{t_1 - t_{av}}$$

Враховуючи, що $q = 230\ \text{кВт/м}^2 = 230\,000\ \text{Вт/м}^2$:
$$\alpha = \frac{230\,000}{75 - 37,5} = \frac{230\,000}{37,5} \approx 6133,33\ \text{Вт/(м}^2\cdot\text{К)}$$

#### 3.3 Обчислення числа Прандтля та швидкості руху води
Число Прандтля для охолоджувальної води:
$$Pr = \frac{\rho \cdot C \cdot \nu}{\lambda} = \frac{1000 \cdot 4200 \cdot 10^{-6}}{0,63} = 6,6667$$

Швидкість руху води $w$ в змійовику визначається з критеріального рівняння для турбулентного режиму течії:
$$w = \left(\frac{\alpha \cdot d_1^{0,2} \cdot \nu^{0,8}}{0,021 \cdot Pr^{0,43} \cdot \lambda}\right)^{1,25}$$

Внутрішній діаметр труби в метрах: $d_1 = 56\ \text{мм} = 0,056\ \text{м}$.
Підставимо значення:
$$w = \left(\frac{6133,33 \cdot 0,056^{0,2} \cdot (10^{-6})^{0,8}}{0,021 \cdot 6,6667^{0,43} \cdot 0,63}\right)^{1,25}$$

Обчислимо проміжні величини:
- $d_1^{0,2} = 0,056^{0,2} \approx 0,5623$
- $\nu^{0,8} = (10^{-6})^{0,8} = 10^{-4,8} \approx 1,5849 \cdot 10^{-5}$
- Чисельник: $6133,33 \cdot 0,5623 \cdot 1,5849 \cdot 10^{-5} \approx 0,05466$
- $Pr^{0,43} = 6,6667^{0,43} \approx 2,2544$
- Знаменник: $0,021 \cdot 2,2544 \cdot 0,63 \approx 0,02983$
- Дріб під степенем: $\frac{0,05466}{0,02983} \approx 1,8324$
- Швидкість: $w = 1,8324^{1,25} \approx 2,123\ \text{м/с}$

Розрахована швидкість води $w \approx 2,12\ \text{м/с}$ лежить в оптимальному рекомендованому експлуатаційному діапазоні $0,6 \text{--} 2,5\ \text{м/с}$.

#### 3.4 Визначення витрати води
Секундна витрата води в трубі:
$$Q = w \cdot \frac{\pi \cdot d_1^2}{4}$$
$$Q = 2,123 \cdot \frac{\pi \cdot 0,056^2}{4} \approx 2,123 \cdot 0,002463 \approx 0,005228\ \text{м}^3/\text{с}$$

Переведемо витрату води в об'ємну годинну витрату для зручності експлуатаційного контролю:
$$Q_{hour} = Q \cdot 3600 \approx 0,005228 \cdot 3600 \approx 18,82\ \text{м}^3/\text{год}$$

---

### 3.5 Розрахунок максимальної довжини змійовика
Довжина змійовика $L$ визначається з умови теплового балансу панелі. Кількість тепла, що передається променевим шляхом на зовнішню поверхню напівтруби змійовика довжиною $L$, має повністю відводитися потоком води:
$$q \cdot \pi \frac{d}{2} \cdot L = w \cdot \rho \cdot \frac{\pi \cdot d_1^2}{4} \cdot C \cdot (t_{out} - t_{in})$$

Звідси максимально допустима довжина змійовика $L$:
$$L = \frac{w \cdot \rho \cdot d_1^2 \cdot C \cdot (t_{out} - t_{in})}{2 \cdot q \cdot d}$$

Зовнішній діаметр труби: $d = 76\ \text{мм} = 0,076\ \text{м}$.
Підставимо чисельні значення:
$$L = \frac{2,123 \cdot 1000 \cdot 0,056^2 \cdot 4200 \cdot (55 - 20)}{2 \cdot 230\,000 \cdot 0,076}$$
$$L = \frac{2,123 \cdot 1000 \cdot 0,003136 \cdot 4200 \cdot 35}{34\,960}$$
$$L = \frac{978\,772}{34\,960} \approx 27,99\ \text{м}$$

Довжина змійовика $L \approx 27,99\ \text{м}$ відповідає рекомендованій інженерній межі $10 \text{--} 30\ \text{м}$. Панель з такою довжиною змійовика здатна забезпечити стаціонарний температурний режим без ризику закипання води на виході.

---

### 3.6 Температура робочої поверхні панелі
Температура зовнішньої поверхні труби $t_2$ визначається за рівнянням теплопровідності через циліндричну стінку змійовика:
$$t_2 = t_1 + \frac{q \cdot \ln(d/d_1)}{2 \cdot \pi \cdot \lambda_{pipe}}$$

Оскільки для Варіанту № 10 використовується **мідь** з коефіцієнтом теплопровідності $\lambda_{pipe} = 370\ \text{Вт/(м}\cdot\text{К)}$:
$$t_2 = 75 + \frac{230\,000 \cdot \ln(76/56)}{2 \cdot \pi \cdot 370}$$
$$t_2 = 75 + \frac{230\,000 \cdot 0,30538}{2324,78}$$
$$t_2 = 75 + \frac{70\,237,8}{2324,78} \approx 75 + 30,21 = 105,21\ ^\circ\text{C}$$

Гранична допустима температура експлуатації мідних панелей становить $260$ °С. Отримане значення $t_2 \approx 105,21$ °С є абсолютно безпечним і свідчить про наявність значного теплового запасу надійності (близько $154,79$ °С). Це досягається завдяки дуже високій теплопровідності міді порівняно зі сталлю Ст20.

---

### 3.7 Втрати тиску води в панелі

Загальні втрати тиску складаються з втрат на тертя по довжині труби та втрат у місцевих опорах.

#### 3.7.1 Втрати тиску на тертя по довжині змійовика:
$$\Delta P_{fr} = \mu_{fr} \cdot \frac{L}{d_1} \cdot \rho \cdot \frac{w^2}{2}$$
$$\Delta P_{fr} = 0,045 \cdot \frac{27,99}{0,056} \cdot 1000 \cdot \frac{2,123^2}{2}$$
$$\Delta P_{fr} = 0,045 \cdot 499,8 \cdot 1000 \cdot 2,2536 \approx 50\,670\ \text{Па} \approx 50,67\ \text{кПа}$$

#### 3.7.2 Місцеві втрати тиску:
У системі є $n_{90} = 2$ повороти на 90° та $n_{180} = 8$ поворотів на 180°.
$$\Delta P_{lr} = (n_{90} \cdot \xi_{lr}^1 + n_{180} \cdot \xi_{lr}^2) \cdot \rho \cdot \frac{w^2}{2}$$
$$\Delta P_{lr} = (2 \cdot 0,22 + 8 \cdot 0,31) \cdot 1000 \cdot \frac{2,123^2}{2}$$
$$\Delta P_{lr} = (0,44 + 2,48) \cdot 1000 \cdot 2,2536 = 2,92 \cdot 2253,6 \approx 6580\ \text{Па} \approx 6,58\ \text{кПа}$$

#### 3.7.3 Сумарні втрати тиску в панелі:
$$\Delta P_{loss} = \Delta P_{fr} + \Delta P_{lr} + \Delta P_g$$
Враховуючи, що статична складова $\Delta P_g = 0$:
$$\Delta P_{loss} = 50\,670 + 6580 = 57\,250\ \text{Па} \approx 57,24\ \text{кПа} \approx 0,0572\ \text{МПа}$$

#### 3.7.4 Перевірка відповідності експлуатаційним нормам:
Згідно з технічними вимогами, втрати тиску з урахуванням коефіцієнта запасу $1,25$ не повинні перевищувати наявний тиск води в цеховій магістралі $P_{wat} = 0,30$ МПа:
$$\Delta P_{loss} \cdot 1,25 \le P_{wat}$$
$$0,0572 \cdot 1,25 = 0,0716\ \text{МПа}$$

Оскільки $0,0716\ \text{МПа} \le 0,30\ \text{МПа}$, умова міцності та гідравлічної стабільності контуру повністю виконується. Наявний тиск води в цеху є достатнім для забезпечення необхідної витрати охолоджувального середовища.

### 5. Висновки

1. Проектована водоохолоджувана стінова панель ДСП для умов **Варіанту № 10** є повністю працездатною та теплогідравлічно стабільною. Всі ключові параметри знаходяться в межах встановлених інженерних та експлуатаційних допусків.
2. Розрахована швидкість течії води $2,12$ м/с гарантує розвинений турбулентний режим, що забезпечує високу інтенсивність тепловіддачі ($\alpha \approx 6133$ Вт/(м²·К)).
3. Використання мідних труб замість сталевих дозволило значно знизити температуру робочої поверхні панелі до безпечних $105,21$ °С (запас до критичної межі становить майже $155$ °С), що суттєво підвищує стійкість конструкції до термічних деформацій та втоми матеріалу.
4. Втрати тиску в контурі складають всього $0,057$ МПа, що з урахуванням коефіцієнта запасу становить $0,0716$ МПа. Це вчетверо менше за наявний цеховий тиск води ($0,30$ МПа), тому система охолодження охоплює високу гідравлічну стійкість до можливих коливань тиску в магістралі.

### 6. Додаток: Python-скрипт для верифікації розрахунків

```python
import math

# 1. Вхідні дані (Варіант 10)
q = 230  # тепловий потік, кВт/м²
d = 76 / 1000  # зовнішній діаметр труби, м
d1 = 56 / 1000  # внутрішній діаметр труби, м
n_90 = 2  # кількість поворотів на 90°
n_180 = 8  # кількість поворотів на 180°
t_in = 20  # температура води на вході, °С
P_wat = 0.30  # наявний тиск води в цеху, МПа
material = "Мідь"

# Константи
t1 = 75  # температура стінки водяного каналу, °С
t_out = 55  # температура води на виході, °С
dP_g = 0  # статичні втрати тиску, Па

lambda_cu = 370  # теплопровідність міді, Вт/(м·К)
lambda_water = 0.63  # теплопровідність води, Вт/(м·К)
rho = 1000  # щільність води, кг/м³
C = 4200  # теплоємність води, Дж/(кг·К)
nu = 1e-6  # коефіцієнт кінематичної в'язкості води, м²/с
mu_fr = 0.045  # коефіцієнт тертя води в трубі
xi_lr1 = 0.22  # коефіцієнт місцевого опору (90°)
xi_lr2 = 0.31  # коефіцієнт місцевого опору (180°)

# 2. Розрахунок швидкості та витрати води
q_w = q * 1000  # Вт/м²
t_av = (t_in + t_out) / 2
alpha = q_w / (t1 - t_av)
Pr = (rho * C * nu) / lambda_water
w = ((alpha * d1**0.2 * nu**0.8) / (0.021 * Pr**0.43 * lambda_water)) ** 1.25
Q = w * math.pi * d1**2 / 4
Q_hour = Q * 3600

# 3. Довжина змійовика
L = (w * rho * d1**2 * C * (t_out - t_in)) / (2 * q_w * d)

# 4. Температура робочої поверхні
t2 = t1 + (q_w * math.log(d/d1)) / (2 * math.pi * lambda_cu)

# 5. Втрати тиску
dP_fr = mu_fr * (L / d1) * rho * (w**2 / 2)
dP_lr = (n_90 * xi_lr1 + n_180 * xi_lr2) * rho * (w**2 / 2)
dP_loss = dP_fr + dP_lr + dP_g
dP_required = dP_loss * 1.25 / 1e6  # МПа

# 6. Виведення результатів
print(f"--- Результати розрахунку для Варіанту № 10 ({material}) ---")
print(f"Середня температура води: {t_av:.2f} °С")
print(f"Коефіцієнт тепловіддачі alpha: {alpha:.2f} Вт/(м²·К)")
print(f"Швидкість руху води w: {w:.3f} м/с (рекомендовано 0.6 - 2.5 м/с)")
print(f"Витрата води: {Q:.6f} м³/с ({Q_hour:.2f} м³/год)")
print(f"Максимальна довжина змійовика L: {L:.2f} м (рекомендовано 10 - 30 м)")
print(f"Температура зовнішньої поверхні t2: {t2:.2f} °С (допустима <= 260 °С)")
print(f"Втрати тиску на тертя: {dP_fr/1000:.2f} кПа")
print(f"Місцеві втрати тиску: {dP_lr/1000:.2f} кПа")
print(f"Сумарні втрати тиску dP_loss: {dP_loss/1000:.2f} кПа")
print(f"Необхідний тиск з коефіцієнтом запасу 1.25: {dP_required:.4f} МПа (наявний: {P_wat:.2f} МПа)")
```
