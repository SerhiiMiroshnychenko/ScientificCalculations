# Розв'язання задачі з питомої потужності перемішування сталі у пристрої ківш-піч

## Умова задачі

Технологічний регламент позапічної обробки сталі в пристрої ківш-піч потребує забезпечити питому потужність перемішування 12 Вт/т. Визначити необхідну витрату аргону через пористу донну пробку (м³/годину за нормальних умов). Вихідні дані: глибина ванни 3,2 м; маса металу у ківші 120 т; температура металу 1595 °С, молекулярна маса газу 40 кг/кмоль, густина газу за нормальних умов 1,78 кг/м³, універсальна газова стала 8314 Дж/(кмоль К).

## 1. Теоретичне обґрунтування процесу

### 1.1. Роль перемішування в металургійних процесах

Перемішування розплаву є критично важливим етапом позапічної обробки сталі. Ефективне перемішування забезпечує:

1. **Гомогенізацію хімічного складу** - рівномірний розподіл легуючих елементів по всьому об'єму розплаву.
2. **Вирівнювання температури** - усунення температурної стратифікації в об'ємі металу.
3. **Прискорення масопереносу** - інтенсифікація хімічних реакцій між металом і шлаком.
4. **Видалення неметалевих включень** - флотація включень до шлакової фази.
5. **Десульфурацію та дефосфорацію** - покращення взаємодії металу зі шлаком.

### 1.2. Фізична природа та термодинаміка пневматичного перемішування

Пневматичне перемішування (барботаж) ґрунтується на введенні інертного газу (найчастіше аргону) в нижню частину ковша. При продувці глибоких ванн енергія, що передається розплаву, складається переважно з енергії розширення газу.

Під час спливання бульбашок від дна ковша до поверхні гідростатичний тиск стовпа рідкого металу постійно зменшується. Оскільки температура металу висока і в межах ванни практично постійна, розширення бульбашок аргону вважається **ізотермічним**. 

Робота розширення 1 моля газу визначається за рівнянням:
$$W = R \cdot T \cdot \ln\left(\frac{P_{bottom}}{P_{surface}}\right)$$

Відношення тисків можна представити через «еквівалентну висоту стовпа металу». Для рідкої сталі ($\rho \approx 7000$ кг/м³) тиск в 1 атм (101325 Па) відповідає стовпу металу висотою $\approx 1{,}48$ м. Таким чином, аргумент логарифма набуває вигляду $\left(1 + \frac{H}{1{,}48}\right)$.

### 1.3. Теорія розрахунку питомої потужності

Питома потужність перемішування ($P_п$) є ключовим параметром інтенсивності масообмінних процесів і визначається як відношення загальної потужності ($P$) до маси металу ($m$).

Після переходу від молярної витрати до об'ємної витрати за нормальних умов ($Q_н$) отримують зручну розрахункову залежність:

$$P_п = \frac{\rho_н \cdot Q_н \cdot R \cdot T \cdot \ln\left(1 + \frac{H}{1{,}48}\right)}{m \cdot M}$$

Саме ця форма враховує фундаментальну роботу ізотермічного розширення і використовується для проектування технологічних режимів продувки.

## 2. Вихідні дані

- Глибина ванни: $H = 3{,}2\;\text{м}$
- Маса металу у ківші: $m = 120\;\text{т}$
- Температура металу: $T = 1595\;^\circ\text{C} = 1868{,}15\;\text{К}$
- Молекулярна маса газу: $M = 40\;\text{кг}/\text{кмоль}$
- Густина газу за нормальних умов: $\rho_н = 1{,}78\;\text{кг}/\text{м}^3$
- Універсальна газова стала: $R = 8314\;\text{Дж}/(\text{кмоль} \cdot \text{К})$
- Задана питома потужність перемішування: $P_п = 12\;\text{Вт}/\text{т}$

## 3. Розв'язання

### 3.1. Визначення логарифмічного комплексу

Для врахування впливу гідростатичного тиску стовпа рідкого металу обчислюємо логарифмічний комплекс, що відображає ступінь розширення газу:

$$\ln\left(1 + \frac{H}{1{,}48}\right) = \ln\left(1 + \frac{3{,}2}{1{,}48}\right)$$

Розрахунок:

$$\ln\left(1 + \frac{3{,}2}{1{,}48}\right) = \ln\left(1 + 2{,}1622\right) = \ln\left(3{,}1622\right) = 1{,}1513$$

Цей коефіцієнт показує, що газ під час спливання від дна до поверхні ванни розширюється більш ніж у 3 рази.

### 3.2. Розрахунок необхідної загальної потужності перемішування

Загальна потужність перемішування визначається як добуток заданої питомої потужності на масу металу:

$$P = P_п \cdot m = 12 \cdot 120 = 1440\;\text{Вт}$$

### 3.3. Визначення витрати газу за нормальною формулою

Для розрахунку витрати аргону за нормальних умов ($Q_н$) виражаємо її з базового рівняння потужності:

$$Q_н = \frac{P_п \cdot m \cdot M}{\rho_н \cdot R \cdot T \cdot \ln\left(1 + \frac{H}{1{,}48}\right)}$$

Розрахунок (використовуючи температуру в Кельвінах $1868{,}15$ К):

$$Q_н = \frac{12 \cdot 120 \cdot 40}{1{,}78 \cdot 8314 \cdot 1868{,}15 \cdot 1{,}1513} = \frac{57600}{31820261} \approx 0{,}001810\;\text{м}^3/\text{с}$$

### 3.4. Переведення у погодинну витрату

Оскільки за умовою задачі витрату аргону потрібно подати у $\text{м}^3/\text{год}$ (стандартна технологічна одиниця вимірювання витрати газів), виконуємо переведення:

$$Q_н = 0{,}001810 \cdot 3600 = 6{,}515\;\text{м}^3/\text{год} \approx 6{,}52\;\text{м}^3/\text{год}$$

## 4. Аналіз та висновки

Для забезпечення питомої потужності перемішування $12\;\text{Вт}/\text{т}$ у ківш-печі з масою металу $120\;\text{т}$ необхідна витрата аргону через пористу донну пробку становить **$6{,}52\;\text{м}^3/\text{год}$** за нормальних умов.

**Технологічна оцінка результату:**
1. **Режим продувки:** Витрата $\approx 6{,}5-7{,}0\;\text{м}^3/\text{год}$ (або $\approx 108-116\;\text{л}/\text{хв}$) забезпечує режим «м'якої продувки». Такий режим є оптимальним для гомогенізації температури, десульфурації та флотації неметалевих включень без надмірного розриву шлакового покриву («оголення ока металу»), що мінімізує вторинне окиснення сталі атмосферою.
2. **Вплив глибини ванни:** Аналіз розрахункової формули показує, що зі збільшенням глибини ванни ($H$) ефективність використання кожної одиниці об'єму газу зростає завдяки виконанню більшої роботи ізотермічного розширення.
3. **Порівняння з ЕМП:** Пневматичне перемішування з потужністю $12\;\text{Вт}/\text{т}$ дозволяє досягти високої однорідності розплаву. Для порівняння, системи електромагнітного перемішування (ЕМП) можуть генерувати аналогічну або більшу питому потужність (11–105 Вт/т), проте пневматичне перемішування залишається більш поширеним і економічно виправданим інструментом на установках ківш-піч.

## 5. Додаток: Програмна реалізація розрахунку (Python)

Нижче наведено скрипт для автоматизації розрахунку та моделювання різних технологічних режимів.

```python
print("="*65)
print("ВИХІДНІ ДАНІ:")
print(f"Глибина ванни: {H} м")
print(f"Маса металу у ківші: {m} т")
print(f"Температура металу: {T-273.15} °C ({T:.2f} К)")
print(f"Молекулярна маса аргону: {M} кг/кмоль")
print(f"Густина аргону за нормальних умов: {rho_n} кг/м³")
print(f"Задана питома потужність перемішування: {P_p} Вт/т")
print("="*65)

# 1. Визначення логарифмічного комплексу (ступінь розширення газу)
log_argument = 1 + H / 1.48
log_term = math.log(log_argument)
print(f"\n1. ВИЗНАЧЕННЯ ЛОГАРИФМІЧНОГО КОМПЛЕКСУ (Енергія розширення)")
print(f"1 + H/1.48 = 1 + {H}/1.48 = {log_argument:.4f}")
print(f"ln(1 + H/1.48) = ln({log_argument:.4f}) = {log_term:.6f}")

# 2. Розрахунок необхідної загальної потужності перемішування
P = P_p * m
print(f"\n2. РОЗРАХУНОК НЕОБХІДНОЇ ЗАГАЛЬНОЇ ПОТУЖНОСТІ ПЕРЕМІШУВАННЯ")
print(f"P = P_п·m = {P_p}·{m} = {P} Вт")

# 3. Визначення витрати газу за термодинамічною формулою
Q_n = (P_p * m * M) / (rho_n * R * T * log_term)
print(f"\n3. ВИЗНАЧЕННЯ ВИТРАТИ ГАЗУ ЗА НОРМАЛЬНИХ УМОВ")
print(f"Q_н = (P_п·m·M)/(ρ_н·R·T·ln(1 + H/1.48)) = ({P_p}·{m}·{M})/({rho_n}·{R}·{T:.2f}·{log_term:.6f}) = {Q_n:.6f} м³/с")

# 4. Переведення у погодинну витрату
Q_n_per_hour = Q_n * 3600
print(f"\n4. ПЕРЕВЕДЕННЯ У ТЕХНОЛОГІЧНУ ПОГОДИННУ ВИТРАТУ")
print(f"Q_н(год) = {Q_n:.6f} · 3600 = {Q_n_per_hour:.3f} м³/год")

print("\n"+"="*65)
print("ВИСНОВОК:")
print(f"Для забезпечення питомої потужності перемішування {P_p} Вт/т")
print(f"у ківш-печі з масою металу {m} т необхідна витрата аргону")
print(f"через пористу донну пробку становить {Q_n_per_hour:.3f} м³/год за н.у.")
print("="*65)
```
