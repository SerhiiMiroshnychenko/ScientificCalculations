Рівняння $\frac{PV}{T} = const$ є формою об'єднаного газового закону (або рівняння Клапейрона), який описує поведінку ідеального газу. Це рівняння справедливе за умови, що кількість речовини (або маса) газу залишається незмінною.

Розглянемо варіанти відповідей:

а) нормального атмосферного тиску - це конкретне значення тиску, а не загальна умова.

б) постійної концентрації молекул - концентрація залежить від об'єму та кількості молекул. Постійна концентрація не є умовою для цього рівняння.

в) постійної маси газу - якщо маса газу постійна, то і кількість молів (n) є постійною. З рівняння стану ідеального газу $PV = nRT$, де R - універсальна газова стала, ми можемо переписати його як $\frac{PV}{T} = nR$. Оскільки n і R постійні, то $\frac{PV}{T} = const$.

г) постійної швидкості руху молекул - швидкість руху молекул пов'язана з температурою. Постійна швидкість означала б постійну температуру, що є умовою для закону Бойля-Маріотта (PV=const при T=const), але не для об'єднаного газового закону.

Таким чином, вірною умовою є **постійна маса газу**.

**Відповідь: в) постійної маси газу**

Визначте процес, зображений на графіку

На графіку зображена залежність тиску (P) від температури (T). Графік є прямою лінією, що проходить через початок координат. Це означає, що тиск прямо пропорційний температурі: $P \propto T$, або $\frac{P}{T} = const$.

З рівняння стану ідеального газу для постійної маси ($PV = nRT$) ми можемо вивести співвідношення між P і T: $\frac{P}{T} = \frac{nR}{V}$.
Оскільки $n$ (кількість речовини) і $R$ (універсальна газова стала) є постійними для постійної маси газу, то для того, щоб відношення $\frac{P}{T}$ було постійним, об'єм $V$ також має бути постійним.

Процес, що відбувається за постійного об'єму, називається ізохорним.

Розглянемо інші варіанти:

а) Адіабатний - процес без теплообміну. На P-T діаграмі виглядає інакше.

б) Ізотермічний - процес за постійної температури. На P-T діаграмі це вертикальна лінія.

в) Ізобарний - процес за постійного тиску. На P-T діаграмі це горизонтальна лінія.

Таким чином, графік, де $P \propto T$ і лінія проходить через початок координат, відповідає ізохорному процесу.

**Відповідь: г) Ізохорний**

Визначить роботу ідеального газу згідно графіку зміни його стану

Робота, виконана газом під час зміни його стану, на P-V діаграмі чисельно дорівнює площі під кривою процесу. В даному випадку графік складається з двох відрізків.

Перший відрізок: від об'єму 10 м³ до 20 м³ при постійному тиску 50 Па. Це ізобарний процес.
Робота в ізобарному процесі розраховується за формулою $W_1 = P \Delta V$.
$$W_1 = 50 \, \text{Па} \times (20 \, \text{м}^3 - 10 \, \text{м}^3) = 50 \, \text{Па} \times 10 \, \text{м}^3 = 500 \, \text{Дж}$$

Другий відрізок: від об'єму 20 м³ до 30 м³, при цьому тиск лінійно зростає від 50 Па до 100 Па. Площа під цим відрізком є площею трапеції.
Площа трапеції розраховується як $\frac{1}{2} \times (\text{сума основ}) \times \text{висота}$. У нашому випадку "основи" - це тиски, а "висота" - зміна об'єму.
$$W_2 = \frac{1}{2} (P_2 + P_3) (V_3 - V_2)$$
$$W_2 = \frac{1}{2} (50 \, \text{Па} + 100 \, \text{Па}) (30 \, \text{м}^3 - 20 \, \text{м}^3)$$
$$W_2 = \frac{1}{2} (150 \, \text{Па}) (10 \, \text{м}^3)$$
$$W_2 = \frac{1}{2} \times 1500 \, \text{Дж} = 750 \, \text{Дж}$$

Повна робота, виконана газом, дорівнює сумі робіт на кожному відрізку:
$$W_{загальна} = W_1 + W_2 = 500 \, \text{Дж} + 750 \, \text{Дж} = 1250 \, \text{Дж}$$

Переведемо джоулі в кілоджоулі:
$$1250 \, \text{Дж} = 1.25 \, \text{кДж}$$.

Порівнюємо з варіантами відповідей:

а) 3,0 кДж

б) 1,5 кДж

в) 1,25 кДж

г) 2,65 кДж

Розраховане значення збігається з варіантом в).

**Відповідь: в) 1,25 кДж**

На якому графіку зображено ізотермічний процес

Ізотермічний процес - це процес зміни стану газу, що відбувається за постійної температури ($T = const$). Для ідеального газу при постійній масі, ізотермічний процес описується законом Бойля-Маріотта: $PV = const$.

Розглянемо кожен графік:

1. Графік V(T): Пряма лінія, що проходить через початок координат. Це означає $V \propto T$. Це графік ізобарного процесу (при постійному тиску).

2. Графік P(V): Крива, яка показує обернену залежність тиску від об'єму ($P \propto \frac{1}{V}$ або $PV = const$). Це графік ізотермічного процесу.

3. Графік P(T): Пряма лінія, що проходить через початок координат. Це означає $P \propto T$. Це графік ізохорного процесу (при постійному об'ємі).

4. Графік P(T): Крива, яка не відповідає жодному з основних ізо-процесів на цій діаграмі. Ізотермічний процес на P-T діаграмі виглядав би як вертикальна лінія (при постійній T).

Таким чином, графік, що зображує ізотермічний процес ($PV = const$), це графік під номером 2.

**Відповідь: б) 2**

Під час ізотермічного процесу тиск газу збільшився в 3 рази. Визначте, як змінився об'єм газу.

Ізотермічний процес означає, що температура газу залишається постійною ($T = const$). Для ідеального газу постійної маси ізотермічний процес описується законом Бойля-Маріотта:
$P_1V_1 = P_2V_2$

За умовою, тиск газу збільшився в 3 рази, тобто $P_2 = 3P_1$. Підставимо це співвідношення в рівняння Бойля-Маріотта:
$P_1V_1 = (3P_1)V_2$

Оскільки $P_1 \neq 0$, ми можемо розділити обидві частини рівняння на $P_1$:
$V_1 = 3V_2$

Звідси випливає:
$V_2 = \frac{V_1}{3}$

Таким чином, кінцевий об'єм $V_2$ став у 3 рази меншим за початковий об'єм $V_1$. Це означає, що об'єм газу зменшився у 3 рази.

Розглянемо варіанти відповідей:

а) Збільшився у 4 рази - Неправильно

б) Зменшився у 3 рази - Правильно

в) Збільшився у 3 рази - Неправильно

г) Не змінився - Неправильно (це ізохорний процес)

**Відповідь: б) Зменшився у 3 рази**

Під час ізохорного процесу тиск газу збільшився в 3 рази. Визначте, як змінився об'єм газу.

Ізохорний процес - це процес зміни стану газу, що відбувається за постійного об'єму ($V = const$). За визначенням ізохорного процесу, об'єм газу не змінюється незалежно від того, як змінюються його тиск або температура.

Згідно з умовою, тиск газу збільшився в 3 рази. Для ізохорного процесу (при постійній масі газу) це означає, що температура також збільшилася в 3 рази згідно із законом Шарля (або другим законом Гей-Люссака): $\frac{P_1}{T_1} = \frac{P_2}{T_2}$ при $V=const$, звідки $\frac{P_2}{P_1} = \frac{T_2}{T_1}$. Якщо $P_2 = 3P_1$, то $T_2 = 3T_1$.

Однак, запитання стосується зміни об'єму. Оскільки процес є ізохорним, об'єм залишається незмінним.

Розглянемо варіанти відповідей:

а) Збільшився в 3 рази - Неправильно.

б) Не змінився - Правильно, за визначенням ізохорного процесу.

в) Збільшився в 6 раз - Неправильно.

г) Зменшився в 3 рази - Неправильно.

**Відповідь: б) Не змінився**

Ідеальний газ не виконує механічної роботи при

Механічна робота, що виконується газом, визначається за формулою $W = \int P \, dV$. Для виконання механічної роботи газом необхідна зміна його об'єму ($\Delta V \neq 0$).

Розглянемо кожен із зазначених процесів:

а) Адіабатний процес: відбувається без теплообміну з навколишнім середовищем ($Q=0$). Робота в адіабатному процесі, як правило, не дорівнює нулю, якщо змінюється об'єм.

б) Ізотермічний процес: відбувається за постійної температури ($T=const$). Робота в ізотермічному процесі не дорівнює нулю, якщо змінюється об'єм ($W = nRT \ln(V_2/V_1)$).

в) Ізобарний процес: відбувається за постійного тиску ($P=const$). Робота в ізобарному процесі дорівнює $W = P \Delta V$, і вона не дорівнює нулю, якщо змінюється об'єм.

г) Ізохорний процес: відбувається за постійного об'єму ($V=const$). В цьому випадку зміна об'єму $\Delta V = 0$. Відповідно, робота, що виконується газом, дорівнює $W = P \Delta V = P \times 0 = 0$.

Таким чином, ідеальний газ не виконує механічної роботи під час ізохорного процесу, оскільки його об'єм залишається незмінним.

**Відповідь: г) Ізохорному процесі**

Ідеальна теплова машина отримала від нагрівача 600 Дж теплоти та передала охолоджувачу 480 Дж. Знайти ККД машини.

Коефіцієнт корисної дії (ККД, $\eta$) теплової машини визначається як відношення корисної роботи ($W$), виконаної машиною, до кількості теплоти ($Q_H$), отриманої від нагрівача:
$$\eta = \frac{W}{Q_H}$$

Згідно з першим законом термодинаміки, робота, виконана тепловою машиною, дорівнює різниці між теплотою, отриманою від нагрівача ($Q_H$), та теплотою, переданою охолоджувачу ($Q_C$):
$$W = Q_H - Q_C$$

За умовою задачі:
Теплота від нагрівача $Q_H = 600 \, \text{Дж}$
Теплота до охолоджувача $Q_C = 480 \, \text{Дж}$

Спочатку розрахуємо виконану роботу:
$$W = 600 \, \text{Дж} - 480 \, \text{Дж} = 120 \, \text{Дж}$$

Тепер розрахуємо ККД:
$$\eta = \frac{W}{Q_H} = \frac{120 \, \text{Дж}}{600 \, \text{Дж}} = \frac{120}{600} = \frac{1}{5} = 0.2$$

Щоб виразити ККД у відсотках, помножимо на 100%:
$$\eta = 0.2 \times 100\% = 20\%$$

ККД теплової машини становить 20%.

Порівняємо з варіантами відповідей:

а) 15% - Неправильно.

б) 20% - Правильно.

в) 10% - Неправильно.

г) 25% - Неправильно.

**Відповідь: б) 20%**

Вказати розмірність коефіцієнту теплопровідності.

Коефіцієнт теплопровідності ($\lambda$) - це фізична величина, що характеризує здатність матеріалу проводити тепло. Він входить до закону Фур'є для теплопровідності, який у одновимірному випадку можна записати як:
$$Q = -\lambda A \frac{dT}{dx}$$
де:

$Q$ - швидкість передачі теплоти (потужність), вимірюється у Ватах (Вт).

$A$ - площа поверхні, через яку проходить тепловий потік, вимірюється у квадратних метрах (м²).

$\frac{dT}{dx}$ - градієнт температури (зміна температури на одиницю довжини), вимірюється у Кельвінах на метр (К/м).

$\lambda$ - коефіцієнт теплопровідності.

З цього рівняння можна виразити розмірність коефіцієнта теплопровідності:
$$[\lambda] = \frac{[Q]}{[A] \cdot [\frac{dT}{dx}]} = \frac{\text{Вт}}{\text{м}^2 \cdot \frac{\text{К}}{\text{м}}} = \frac{\text{Вт} \cdot \text{м}}{\text{м}^2 \cdot \text{К}} = \frac{\text{Вт}}{\text{м} \cdot \text{К}}$$

Отже, розмірність коефіцієнта теплопровідності - Вт/(м·К).

Порівняємо з запропонованими варіантами:

а) Вт/м² - Розмірність густини теплового потоку.

б) Вт/(м²·К) - Неправильна розмірність.

в) Вт/(м·К) - Правильна розмірність коефіцієнта теплопровідності.

г) Дж/(кг·К) - Розмірність питомої теплоємності.

**Відповідь: в) Вт/(м·К)**

# 12
![12](12.png "12")

Який з перелічених заходів не сприятиме інтенсифікації теплопередачі між двома середовищами, розділеними плоскою одношаровою стінкою:

Інтенсифікація теплопередачі означає збільшення швидкості передачі теплоти. Швидкість теплопередачі через плоску одношарову стінку описується формулою:
$$Q = K A \Delta T$$
де
$Q$ - швидкість теплопередачі,

$K$ - коефіцієнт загальної теплопередачі,

$A$ - площа поверхні теплообміну,

$\Delta T$ - різниця температур між середовищами.

Для збільшення $Q$ потрібно збільшити $K$, $A$ або $\Delta T$. Заходи в варіантах стосуються коефіцієнта загальної теплопередачі $K$.

Коефіцієнт загальної теплопередачі для плоскої одношарової стінки розраховується за формулою:
$$\frac{1}{K} = \frac{1}{\alpha_1} + \frac{\delta}{\lambda} + \frac{1}{\alpha_2}$$
де

$\alpha_1$ - коефіцієнт тепловіддачі від першого середовища до стінки,

$\delta$ - товщина стінки,

$\lambda$ - коефіцієнт теплопровідності матеріалу стінки,

$\alpha_2$ - коефіцієнт тепловіддачі від стінки до другого середовища.

Щоб збільшити $K$, потрібно зменшити $\frac{1}{K}$. Розглянемо вплив кожного заходу на $\frac{1}{K}$:

а) підвищення коефіцієнту тепловіддачі від теплоносія до стінки ($\alpha_1$ збільшується): При збільшенні $\alpha_1$ дріб $\frac{1}{\alpha_1}$ зменшується. Отже, $\frac{1}{K}$ зменшується, а $K$ збільшується. Це сприяє інтенсифікації.

б) підвищення коефіцієнту тепловіддачі від стінки до середовища, що нагрівається ($\alpha_2$ збільшується): При збільшенні $\alpha_2$ дріб $\frac{1}{\alpha_2}$ зменшується. Отже, $\frac{1}{K}$ зменшується, а $K$ збільшується. Це сприяє інтенсифікації.

в) зменшення товщини стінки ($\delta$ зменшується): При зменшенні $\delta$ дріб $\frac{\delta}{\lambda}$ зменшується. Отже, $\frac{1}{K}$ зменшується, а $K$ збільшується. Це сприяє інтенсифікації.

г) заміна стінки на матеріал з меншим коефіцієнтом теплопровідності ($\lambda$ зменшується): При зменшенні $\lambda$ дріб $\frac{\delta}{\lambda}$ збільшується (знаменник меншає). Отже, $\frac{1}{K}$ збільшується, а $K$ зменшується. Це не сприяє інтенсифікації, а навпаки, зменшує теплопередачу.

Таким чином, захід, який не сприятиме інтенсифікації теплопередачі, - це заміна стінки на матеріал з меншим коефіцієнтом теплопровідності.

**Відповідь: г) заміна стінки на матеріал з меншим коефіцієнтом теплопровідності**

# 13
![13](13.png "13")

Який з перелічених заходів найбільш ефективно сприятиме інтенсифікації теплопередачі між двома середовищами з коефіцієнтами тепловіддачі $\alpha_1$ і $\alpha_2$ (за умов $\alpha_1 > \alpha_2$), розділеними плоскою одношаровою стінкою:

Загальний термічний опір теплопередачі через плоску стінку виражається як сума трьох опорів: опору тепловіддачі від першого середовища до стінки ($\frac{1}{\alpha_1 A}$), опору теплопровідності стінки ($\frac{\delta}{\lambda A}$) та опору тепловіддачі від стінки до другого середовища ($\frac{1}{\alpha_2 A}$). Загальний коефіцієнт теплопередачі $K$ пов'язаний із загальним термічним опором $R_{total}$ співвідношенням $R_{total} = \frac{1}{KA}$.

$$R_{total} = \frac{1}{\alpha_1 A} + \frac{\delta}{\lambda A} + \frac{1}{\alpha_2 A}$$
$$\frac{1}{KA} = \frac{1}{A} \left( \frac{1}{\alpha_1} + \frac{\delta}{\lambda} + \frac{1}{\alpha_2} \right)$$
$$\frac{1}{K} = \frac{1}{\alpha_1} + \frac{\delta}{\lambda} + \frac{1}{\alpha_2}$$

Щоб інтенсифікувати теплопередачу (збільшити $K$), потрібно зменшити загальний термічний опір $\frac{1}{K}$. Найбільший внесок у загальний опір вносить та складова, яка є найбільшою.

За умовою $\alpha_1 > \alpha_2$, з чого випливає, що $\frac{1}{\alpha_1} < \frac{1}{\alpha_2}$. Тобто опір тепловіддачі з боку середовища з коефіцієнтом $\alpha_2$ більший, ніж з боку середовища з коефіцієнтом $\alpha_1$. Цей більший опір є "вузьким місцем" у процесі теплопередачі.

Розглянемо варіанти:

а) оребрення стінки з боку середовища з коефіцієнтом тепловіддачі $\alpha_1$: Оребрення збільшує площу поверхні тепловіддачі. Це ефективно там, де коефіцієнт тепловіддачі низький. Оскільки $\alpha_1$ більший, опір на цій стороні менший, і ефект від оребрення буде меншим, ніж з боку з нижчим коефіцієнтом.

б) зменшення товщини стінки: Зменшення $\delta$ зменшує опір теплопровідності стінки ($\frac{\delta}{\lambda}$), що сприяє збільшенню $K$. Це ефективно, особливо для товстих стінок або матеріалів з низькою теплопровідністю.

в) збільшення товщини стінки: Збільшення $\delta$ збільшує опір теплопровідності стінки, що зменшує $K$. Це не сприяє інтенсифікації.

г) оребрення стінки з боку середовища з коефіцієнтом тепловіддачі $\alpha_2$: Оскільки $\alpha_2$ є меншим коефіцієнтом тепловіддачі, опір на цій стороні ($\frac{1}{\alpha_2}$) є більшим. Оребрення цієї сторони значно збільшує площу тепловіддачі на стороні найбільшого опору, що призводить до найбільш суттєвого зменшення загального термічного опору і, відповідно, найбільшого збільшення коефіцієнта загальної теплопередачі $K$.

Найбільш ефективним заходом для інтенсифікації теплопередачі, коли один з коефіцієнтів тепловіддачі значно менший за інший, є оребрення поверхні з боку меншого коефіцієнта тепловіддачі.

**Відповідь: г) оребрення стінки з боку середовища з коефіцієнтом тепловіддачі $\alpha_2$**

Який з процесів теплообміну можливо реалізувати в умовах вакууму

Розглянемо основні види теплообміну:

а) **Теплопровідність**: Передача теплоти шляхом безпосереднього контакту частинок середовища. Цей процес потребує наявності речовини (твердої, рідкої або газоподібної), оскільки передача енергії відбувається через взаємодію молекул або інших частинок. г) Усі процеси: Неправильно, оскільки теплопровідність та конвекція потребують середовища.

Таким чином, єдиним видом теплообміну, що може відбуватися в умовах вакууму, є випромінювання.

Рівняння Ньютона-Ріхмана, що описує процес теплообміну конвекцією має вигляд:

Рівняння Ньютона-Ріхмана (також відоме як закон Ньютона для охолодження у контексті конвективного теплообміну) описує швидкість теплообміну між поверхнею та рухомим середовищем (рідиною або газом) через конвекцію. Це рівняння формулюється таким чином: швидкість теплообміну пропорційна площі поверхні контакту та різниці температур між поверхнею та середовищем.

У перерахунку на густину теплового потоку ($q$, теплота за одиницю площі за одиницю часу), рівняння має вигляд:
$$q = \alpha \Delta t$$
або
$$q = \alpha (t_{поверхні} - t_{середовища})$$

де:

$q$ - густина теплового потоку (Вт/м²)

$\alpha$ - коефіцієнт тепловіддачі (Вт/(м²·К) або Вт/(м²·°C))

$\Delta t$ - різниця температур між поверхнею та середовищем (°C або К).

Враховуючи позначення температур як $t_1$ та $t_2$, рівняння набуває вигляду $$q = \alpha (t_1 - t_2)$$

Розглянемо запропоновані варіанти:

а) $q = \alpha (t_1 - t_2)$: Це рівняння відповідає закону Ньютона-Ріхмана для конвективного теплообміну. $\alpha$ - коефіцієнт тепловіддачі.

б) $q = \sigma (T_1^4 - T_2^4)$: Це закон Стефана-Больцмана для теплообміну випромінюванням між двома абсолютно чорними тілами при температурах $T_1$ та $T_2$.

в) $q = \frac{\lambda}{\delta} (t_1 - t_2)$: Це рівняння описує теплопровідність через плоску стінку товщиною $\delta$ з коефіцієнтом теплопровідності $\lambda$ при різниці температур $t_1 - t_2$ між поверхнями.

г) $q = (\Sigma \frac{\lambda_i}{\delta_i}) (t_1 - t_2)$: Це рівняння описує теплопровідність через багатошарову стінку.

Таким чином, рівняння, що описує процес теплообміну конвекцією згідно з законом Ньютона-Ріхмана, - це варіант а).

**Відповідь: а) $q = \alpha (t_1 - t_2)$**

Закон Фур'є, що визначає процес теплообміну теплопровідністю через плоску одношарову стінку, записується так:

Закон Фур'є для теплопровідності в одновимірному випадку (через плоску стінку) стверджує, що кількість теплоти, що передається за одиницю часу через одиницю площі (густина теплового потоку $q$), пропорційна градієнту температури. Математично це виглядає так:
$$q = -\lambda \frac{dt}{dx}$$

Для плоскої одношарової стінки товщиною $\delta$ з різницею температур $t_1$ та $t_2$ на її поверхнях, градієнт температури можна вважати постійним і рівним $\frac{t_2 - t_1}{\delta}$. Якщо прийняти, що теплота тече від поверхні з температурою $t_1$ до поверхні з температурою $t_2$ (тобто $t_1 > t_2$), то $dt$ буде від'ємним при зростанні $x$ (напрямку передачі теплоти). Щоб отримати додатну густину теплового потоку в напрямку від $t_1$ до $t_2$, використовують різницю температур $(t_1 - t_2)$ і записують:
$$q = \lambda \frac{t_1 - t_2}{\delta}$$
або
$$q = \frac{\lambda}{\delta} (t_1 - t_2)$$

Порівняємо з запропонованими варіантами:

а) $q = \alpha (t_1 - t_2)$: Це закон Ньютона-Ріхмана для конвективного теплообміну.

б) $q = \sigma (T_1^4 - T_2^4)$: Це закон Стефана-Больцмана для теплообміну випромінюванням.

в) $q = \frac{\lambda}{\delta} (t_1 - t_2)$: Це рівняння точно відповідає закону Фур'є для теплопровідності через плоску одношарову стінку.

г) $q = (\Sigma \frac{\lambda_i}{\delta_i}) (t_1 - t_2)$: Це рівняння не є стандартним записом закону Фур'є для багатошарової стінки у такому вигляді.

**Відповідь: в) $q = \frac{\lambda}{\delta} (t_1 - t_2)$**

Рівняння Стефана-Больцмана стосовно процесу теплообміну випромінюванням має наступний вигляд:

Закон Стефана-Больцмана описує енергію, випромінювану абсолютно чорним тілом. У випадку теплообміну випромінюванням між двома тілами, густина чистого теплового потоку ($q$) між ними (для абсолютно чорних тіл) пропорційна різниці четвертих ступенів їх абсолютних температур. Рівняння має вигляд:
$$q = \sigma (T_1^4 - T_2^4)$$
де:

$q$ - густина чистого теплового потоку випромінюванням (Вт/м²)

$\sigma$ - стала Стефана-Больцмана ($\approx 5.67 \times 10^{-8} \, \text{Вт}/(\text{м}^2 \cdot \text{К}^4)$)

$T_1$ та $T_2$ - абсолютні температури поверхонь (в Кельвінах).

Порівняємо з запропонованими варіантами:

а) $q = \alpha (t_1 - t_2)$: Це закон Ньютона-Ріхмана для конвективного теплообміну.

б) $q = \sigma (T_1^4 - T_2^4)$: Це рівняння відповідає закону Стефана-Больцмана для теплообміну випромінюванням між абсолютно чорними тілами.

в) $q = \frac{\lambda}{\delta} (t_1 - t_2)$: Це закон Фур'є для теплопровідності.

г) $q = (\Sigma \frac{\lambda_i}{\delta_i}) (t_1 - t_2)$: Це рівняння пов'язане з теплопровідністю, а не випромінюванням.

**Відповідь: б) $q = \sigma (T_1^4 - T_2^4)$**

Визначити температуру плоскої одношарової стінки товщиною 10 мм за умов: стаціонарний тепловий стан, тепловий потік через стінку 105 кВт/м², коефіцієнт теплопровідності стінки 350 Вт/(м·К), температура стінки з боку води 45 °С.

Для розрахунку температури іншої сторони стінки скористаємося законом Фур'є для стаціонарної теплопровідності через плоску стінку. Густина теплового потоку $q$ визначається за формулою:
$$q = \frac{\lambda}{\delta} (t_1 - t_2)$$
де:

$q$ - густина теплового потоку (Вт/м²)

$\lambda$ - коефіцієнт теплопровідності матеріалу стінки (Вт/(м·К))

$\delta$ - товщина стінки (м)

$t_1$ та $t_2$ - температури на поверхнях стінки (°С або К).

За умовою задачі маємо:

Товщина стінки $\delta = 10 \, \text{мм} = 0.010 \, \text{м}$

Густина теплового потоку $q = 105 \, \text{кВт/м}^2 = 105 \times 10^3 \, \text{Вт/м}^2 = 105000 \, \text{Вт/м}^2$

Коефіцієнт теплопровідності стінки $\lambda = 350 \, \text{Вт/(м·К)}$

Температура стінки з боку води $t_{\text{води}} = 45 \, \text{°С}$

Нехай $t_1$ буде температура з одного боку стінки, а $t_2$ - температура з іншого боку. Різниця температур $(t_1 - t_2)$ викликає тепловий потік $q$. Модуль різниці температур через стінку можна знайти з формули:
$$|t_1 - t_2| = \frac{q \delta}{\lambda}$$

Підставимо відомі значення:
$$|t_1 - t_2| = \frac{105000 \, \text{Вт/м}^2 \times 0.010 \, \text{м}}{350 \, \text{Вт/(м·К)}} = \frac{1050}{350} \, \text{К} = 3 \, \text{К}$$

Оскільки різниця в Кельвінах дорівнює різниці в градусах Цельсія, то різниця температур через стінку становить 3 °С.

Температура стінки з боку води дорівнює 45 °С. Інша сторона стінки може бути або на 3 °С гарячішою, або на 3 °С холоднішою за 45 °С. Тобто температура іншої сторони може бути $45 + 3 = 48$ °С або $45 - 3 = 42$ °С.

Розглянемо варіанти відповідей:

а) 48 °С

б) 45,3 °С

в) 115,5 °С

г) 92 °С

Серед запропонованих варіантів є 48 °С. Якщо температура іншої сторони становить 48 °С, а з боку води 45 °С, то тепловий потік відбувається від 48 °С до 45 °С, що відповідає додатному тепловому потоку в цьому напрямку.

Розрахунок для перевірки: якщо $t_1 = 48$ °С та $t_2 = 45$ °С, то $q = \frac{350}{0.010} (48 - 45) = 35000 \times 3 = 105000 \, \text{Вт/м}^2 = 105 \, \text{кВт/м}^2$. Це збігається із заданим тепловим потоком.

Якщо б інша сторона була 42 °С, а з боку води 45 °С, то тепловий потік також був би 105 кВт/м², але в протилежному напрямку. Однак 42 °С немає серед варіантів.

Отже, найбільш ймовірний варіант, що відповідає умові та варіантам відповіді, - це температура іншої сторони стінки 48 °С.

**Відповідь: а) 48 °С**

визначити ланку, що лімітує процес в цілому, за умов:

$\delta_2 = 0,04 \, \text{м}$;

$\lambda_2 = 2 \, \text{Вт/(мК)}$;

$\alpha_1 = 3500 \, \text{Вт/(м}^2\text{К)}$;

$\alpha_2 = 6200 \, \text{Вт/(м}^2\text{К)}$.

Ланкою, що лімітує процес теплопередачі (або визначає "вузьке місце"), є складова загального термічного опору, яка має найбільше значення. Загальний термічний опір (на одиницю площі) у цьому випадку дорівнює знаменнику дробу у формулі для $q$:
$$R_{total} = \frac{1}{\alpha_1} + \frac{\delta_1}{\lambda_1} + \frac{\delta_2}{\lambda_2} + \frac{1}{\alpha_2}$$

Окремі терми представляють собою термічні опори:

Опір тепловіддачі з боку першого середовища: $R_1 = \frac{1}{\alpha_1}$

Опір теплопровідності першого шару: $R_2 = \frac{\delta_1}{\lambda_1}$

Опір теплопровідності другого шару: $R_3 = \frac{\delta_2}{\lambda_2}$

Опір тепловіддачі з боку другого середовища: $R_4 = \frac{1}{\alpha_2}$

Розрахуємо значення кожного опору за заданими умовами:

Фізичне тіло у процесах нагріву-охолодження вважається теплотехнічно масивним за умов:

Теплотехнічна масивність тіла в контексті процесів нагріву або охолодження пов'язана з тим, наскільки швидко вирівнюється температура всередині тіла порівняно зі швидкістю теплообміну з навколишнім середовищем. Це поняття описується за допомогою критерію Біо (Bi).

Критерій Біо - це відношення термічного опору теплопровідності всередині тіла до термічного опору тепловіддачі з поверхні тіла:
$$Bi = \frac{R_{внутрішній}}{R_{зовнішній}} = \frac{L_c / \lambda}{1 / \alpha} = \frac{\alpha L_c}{\lambda}$$
де:

$L_c$ - характерний лінійний розмір тіла

$\lambda$ - коефіцієнт теплопровідності матеріалу тіла

$\alpha$ - коефіцієнт тепловіддачі від поверхні тіла до середовища.

- Якщо $Bi \ll 1$ (зазвичай $Bi < 0.1$), внутрішній термічний опір значно менший за зовнішній. Температура всередині тіла майже однакова в будь-який момент часу (модель тіла з зосередженими параметрами). Таке тіло вважається теплотехнічно тонким.
- Якщо $Bi > 0.1$, внутрішній термічний опір є порівнянним або більшим за зовнішній. Існують значні градієнти температури всередині тіла. Таке тіло вважається теплотехнічно масивним або товстим. Різні джерела використовують дещо різні порогові значення (наприклад, $Bi > 0.5$ або $Bi > 1$) для чіткого розмежування, але ідея полягає в тому, що при достатньо великому Bi внутрішні градієнти температури є суттєвими.

Розглянемо варіанти відповідей:

а) Критерій Біо > 0,5: Це умова, яка вказує на значний внутрішній термічний опір порівняно із зовнішнім, що характерно для теплотехнічно масивних тіл.

б) Критерій Фур'є ≤ 12: Критерій Фур'є (Fo) характеризує швидкість поширення зміни температури всередині тіла з часом. Він важливий для нестаціонарних процесів, але не визначає теплотехнічну масивність в термінах співвідношення внутрішнього та зовнішнього опорів.

в) Критерій Рейнольдса > 20000: Критерій Рейнольдса визначає режим руху рідини (ламінарний чи турбулентний) і не стосується теплотехнічної масивності твердого тіла.

г) Лінійні розміри тіла зіставні з розмірами печі: Розміри тіла є фактором, що впливає на теплообмін, але само по собі порівняння розмірів тіла та печі не є точним критерієм теплотехнічної масивності. Критерій Біо враховує теплофізичні властивості тіла та умови теплообміну.

Таким чином, тіло вважається теплотехнічно масивним за умов, коли критерій Біо є достатньо великим, наприклад, більшим за 0,5.

**Відповідь: а) Критерій Біо > 0,5**

Вкажіть розмірність коефіцієнту конвективної тепловіддачі.

Коефіцієнт конвективної тепловіддачі ($\alpha$) є мірою інтенсивності теплообміну між поверхнею та рухомим середовищем (рідиною або газом). Згідно з законом Ньютона-Ріхмана, густина теплового потоку при конвекції ($q$) визначається за формулою:
$$q = \alpha \Delta t$$
де:

$q$ - густина теплового потоку (теплова потужність на одиницю площі), вимірюється у Ватах на квадратний метр (Вт/м²).

$\Delta t$ - різниця температур між поверхнею та середовищем, вимірюється у Кельвінах (К) або градусах Цельсія (°С). Оскільки нас цікавить розмірність, різниця температур має розмірність Кельвіна.

$\alpha$ - коефіцієнт конвективної тепловіддачі.

З цієї формули можемо виразити розмірність коефіцієнта тепловіддачі:
$$[\alpha] = \frac{[q]}{[\Delta t]} = \frac{\text{Вт/м}^2}{\text{К}} = \frac{\text{Вт}}{\text{м}^2 \cdot \text{К}}$$

Отже, розмірність коефіцієнта конвективної тепловіддачі - Вт/(м²·К).

Порівняємо з запропонованими варіантами:

а) Вт/м² - Розмірність густини теплового потоку.

б) Вт/(м²·К) - Правильна розмірність коефіцієнта конвективної тепловіддачі.

в) Вт/(м·К) - Розмірність коефіцієнта теплопровідності.

г) Дж/(кг·К) - Розмірність питомої теплоємності.

**Відповідь: б) Вт/(м²·К)**

Визначить коефіцієнти взаємного опромінення поверхонь 1 та 2 ($\Phi_{11}, \Phi_{12}, \Phi_{21}, \Phi_{22}$), площі яких становлять F₁ та F₂, відповідно.

Коефіцієнт взаємного опромінення $\Phi_{ij}$ - це частка випромінювання, що залишає поверхню $i$ і досягає поверхні $j$.

Проаналізуємо коефіцієнти для даної конфігурації:
- $\Phi_{11}$: Коефіцієнт опромінення поверхні 1 на себе. Оскільки поверхня 1 повністю оточена поверхнею 2 і не має увігнутостей, жодне випромінювання, що залишає поверхню 1, не потрапляє назад на неї. Отже, $\Phi_{11} = 0$.

- $\Phi_{12}$: Коефіцієнт опромінення поверхні 1 на поверхню 2. Оскільки поверхня 1 повністю знаходиться всередині поверхні 2, все випромінювання, що залишає поверхню 1, потрапляє на поверхню 2. Отже, $\Phi_{12} = 1$.

- $\Phi_{21}$: Коефіцієнт опромінення поверхні 2 на поверхню 1. Використаємо закон взаємності для коефіцієнтів опромінення: $F_i \Phi_{ij} = F_j \Phi_{ji}$.

Для поверхонь 1 і 2 маємо: $F_1 \Phi_{12} = F_2 \Phi_{21}$.

Оскільки $\Phi_{12} = 1$, то $F_1 \times 1 = F_2 \Phi_{21}$, звідки $\Phi_{21} = \frac{F_1}{F_2}$.

- $\Phi_{22}$: Коефіцієнт опромінення поверхні 2 на себе. Сума коефіцієнтів опромінення від поверхні до всіх поверхонь, що утворюють замкнену систему (включаючи саму себе), дорівнює 1. У даній системі випромінювання з поверхні 2 потрапляє або на поверхню 1, або назад на поверхню 2.

Отже, $\Phi_{21} + \Phi_{22} = 1$.

Знаючи $\Phi_{21} = \frac{F_1}{F_2}$, отримаємо $\Phi_{22} = 1 - \Phi_{21} = 1 - \frac{F_1}{F_2}$.

Таким чином, коефіцієнти взаємного опромінення:
$$\Phi_{11} = 0$$
$$\Phi_{12} = 1$$
$$\Phi_{21} = \frac{F_1}{F_2}$$
$$\Phi_{22} = 1 - \frac{F_1}{F_2}$$

Порівняємо з запропонованими варіантами:

а) $\Phi_{11}=0, \Phi_{12}=1, \Phi_{21}=0, \Phi_{22}=1$ - Неправильно.

б) $\Phi_{11}=1, \Phi_{12}=1- F_1 / F_2, \Phi_{21}= F_2 / F_1, \Phi_{22}=1$ - Неправильно.

в) $\Phi_{11}=0, \Phi_{12}=1, \Phi_{21}= F_1 / F_2, \Phi_{22}= 1- F_1 / F_2$ - Правильно.

г) $\Phi_{11}=0, \Phi_{12}=1- F_1 / F_2, \Phi_{21}= F_2 / F_1, \Phi_{22}= 1- F_1 / F_2$ - Неправильно.

**Відповідь: в) $\Phi_{11}=0, \Phi_{12}=1, \Phi_{21}= F_1 / F_2, \Phi_{22}= 1- F_1 / F_2$**

За якою формулою визначається термічний опір теплопровідності багатошарової плоскої стінки? Позначення: $\alpha$ – коефіцієнт тепловіддачі, $\lambda$ – коефіцієнт теплопровідності, $\delta_i$ – товщина $i$-го шару стінки.

Термічний опір теплопровідності окремого шару плоскої стінки товщиною $\delta$ і коефіцієнтом теплопровідності $\lambda$ (на одиницю площі) визначається за формулою $R = \frac{\delta}{\lambda}$.

Для багатошарової плоскої стінки, що складається з $n$ шарів, які з'єднані послідовно (тепло проходить через кожен шар по черзі), загальний термічний опір теплопровідності (на одиницю площі) дорівнює сумі термічних опорів окремих шарів:
$$R_{теплопровідності, загальний} = R_1 + R_2 + ... + R_n = \frac{\delta_1}{\lambda_1} + \frac{\delta_2}{\lambda_2} + ... + \frac{\delta_n}{\lambda_n} = \sum_{i=1}^{n} \frac{\delta_i}{\lambda_i}$$

Розглянемо запропоновані формули:

а) $R = \frac{1}{\alpha}$: Це термічний опір тепловіддачі (конвекції) на одиницю площі.

б) $R = \sum_{i=1}^{n} \frac{\delta_i}{\lambda_i}$: Ця формула є сумою термічних опорів теплопровідності кожного шару, що відповідає загальному термічному опору теплопровідності багатошарової стінки.

в) $R = \frac{\lambda}{\delta}$: Це термічна провідність одного шару на одиницю площі (обернена до термічного опору теплопровідності).

г) $R = \frac{\lambda}{\delta} + \frac{1}{\alpha}$: Ця формула змішує термічну провідність і термічний опір і не є коректною для опису загального термічного опору теплопровідності багатошарової стінки.

Отже, формула, що визначає термічний опір теплопровідності багатошарової плоскої стінки, - це сума термічних опорів окремих шарів, яка представлена у варіанті б).

**Відповідь: б) $R = \sum_{i=1}^{n} \frac{\delta_i}{\lambda_i}$**

При теплообміні випромінюванням через непрозорий екран того ж ступеня чорноти, як і поверхня що екранується, тепловий потік, що сприймає дана поверхня

Розглянемо випадок теплообміну випромінюванням між двома паралельними плоскими поверхнями з однаковою площею та ступенем чорноти $\epsilon$. Тепловий потік між ними (без екрана) пропорційний $\sigma (T_1^4 - T_2^4)$ та обернено пропорційний термічному опору, який залежить від ступеня чорноти.

При розміщенні між цими поверхнями непрозорого екрана того ж ступеня чорноти $\epsilon$, система перетворюється на послідовне з'єднання радіаційних теплообмінів: від першої поверхні до екрана, і від екрана до другої поверхні. Кожен з цих процесів має свій термічний опір.

Для випадку двох паралельних плоских поверхонь з однаковою площею $A$ та однаковим ступенем чорноти $\epsilon$, тепловий потік без екрана (для випадку, коли $T_1 > T_2$) можна виразити як:
$$Q_{12} = \frac{\sigma (T_1^4 - T_2^4)}{\frac{1}{\epsilon} + \frac{1}{\epsilon} - 1} A = \frac{\sigma (T_1^4 - T_2^4)}{\frac{2}{\epsilon} - 1} A$$

Якщо між поверхнями 1 і 2 розмістити екран 3 з тим же ступенем чорноти $\epsilon$, то тепловий потік буде передаватися послідовно від 1 до 3, а потім від 3 до 2. При стаціонарному режимі тепловий потік від 1 до 3 дорівнює тепловому потоку від 3 до 2, і дорівнює загальному тепловому потоку від 1 до 2 через екран. Припускаючи, що екран має проміжну температуру $T_3$.

Тепловий потік від 1 до 3: $Q_{13} = \frac{\sigma (T_1^4 - T_3^4)}{\frac{1}{\epsilon} + \frac{1}{\epsilon} - 1} A = \frac{\sigma (T_1^4 - T_3^4)}{\frac{2}{\epsilon} - 1} A$

Тепловий потік від 3 до 2: $Q_{32} = \frac{\sigma (T_3^4 - T_2^4)}{\frac{1}{\epsilon} + \frac{1}{\epsilon} - 1} A = \frac{\sigma (T_3^4 - T_2^4)}{\frac{2}{\epsilon} - 1} A$

Оскільки $Q_{13} = Q_{32}$, то $T_1^4 - T_3^4 = T_3^4 - T_2^4$, звідки $2 T_3^4 = T_1^4 + T_2^4$, або $T_3^4 = \frac{T_1^4 + T_2^4}{2}$.

Підставимо $T_3^4$ у формулу для $Q_{13}$ (або $Q_{32}$):
$$Q_{12, з \, екраном} = \frac{\sigma (T_1^4 - \frac{T_1^4 + T_2^4}{2})}{\frac{2}{\epsilon} - 1} A = \frac{\sigma \frac{2T_1^4 - T_1^4 - T_2^4}{2}}{\frac{2-\epsilon}{\epsilon}} A = \frac{\sigma \frac{T_1^4 - T_2^4}{2}}{\frac{2-\epsilon}{\epsilon}} A = \frac{\sigma \epsilon (T_1^4 - T_2^4)}{2(2-\epsilon)} A$$

Порівняємо теплові потоки з екраном і без екрана (припускаючи, що формула без екрана була $Q_{12} = \frac{\sigma \epsilon (T_1^4 - T_2^4)}{2 - \epsilon} A$):
$$\frac{Q_{12, з \, екраном}}{Q_{12, без \, екрана}} = \frac{\frac{\sigma \epsilon (T_1^4 - T_2^4)}{2(2-\epsilon)} A}{\frac{\sigma \epsilon (T_1^4 - T_2^4)}{2 - \epsilon} A} = \frac{1}{2}$$

Таким чином, тепловий потік зменшується в 2 рази.

Дати оцінку швидкості нагріву тіл за умов рівномірного теплового навантаження у порядку збільшення: для пластини, циліндру та сфероїду (однакові значення маси, діючого теплового потоку q, вихідної температури)

Швидкість нагріву тіла пропорційна кількості теплоти, що надходить до тіла за одиницю часу, та обернено пропорційна його теплоємності. Кількість теплоти, що надходить, дорівнює тепловому потоку $q$, помноженому на площу поверхні тіла $A$: $Q = q \cdot A$. Теплоємність тіла $C = m \cdot c_p$, де $m$ - маса, $c_p$ - питома теплоємність.

Швидкість зміни температури (швидкість нагріву) приблизно дорівнює $\frac{\Delta T}{\Delta \tau} \propto \frac{Q}{C} = \frac{q \cdot A}{m \cdot c_p}$.

За умовою задачі, маса ($m$), діючий тепловий потік ($q$) та вихідна температура однакові для всіх тіл. Припускаючи, що тіла виготовлені з одного матеріалу, їх питома теплоємність ($c_p$) також однакова. Таким чином, швидкість нагріву прямо пропорційна площі поверхні тіла $A$:
$$\frac{\Delta T}{\Delta \tau} \propto A$$

Для однакової маси (а отже, однакового об'єму, якщо густина стала), тіло з більшою площею поверхні буде нагріватися швидше, а тіло з меншою площею поверхні - повільніше. Серед тіл однакового об'єму сфера має найменшу площу поверхні.

Порівняємо площі поверхонь пластини, циліндра та сфери однакової маси (об'єму). Сфера (3) має найменшу площу поверхні серед тіл однакового об'єму, тому вона буде нагріватися найповільніше. Пластина (1), особливо якщо вона тонка, матиме відносно велику площу поверхні. Циліндр (2) матиме проміжну площу поверхні (залежно від співвідношення висоти до радіуса).

Зображення візуально підтверджує цю ідею: сфера є найбільш "компактною", тоді як пластина виглядає тонкою і розширеною, що вказує на велику відносну площу поверхні.

Нагрів масивного тіла при граничних умовах третього роду включає інерційний період, впродовж якого перепад температур поверхня - вісь

При нагріві масивного тіла (тіла зі значним термічним опором теплопровідності всередині) відносно холодним середовищем за граничних умов третього роду (конвективний теплообмін з навколишнім середовищем) спостерігається так званий "інерційний період" нестаціонарного теплообміну.

На початку процесу нагріву теплота надходить на поверхню тіла з навколишнього середовища. Температура поверхні тіла починає швидко зростати. Однак, через значний термічний опір теплопровідності матеріалу "масивного" тіла, теплота повільно поширюється вглиб тіла, до його осі або центру.

Протягом цього "інерційного періоду" температура поверхні значно випереджає температуру внутрішніх шарів тіла. Перепад температур між поверхнею та віссю (центром) тіла зростає від початкового значення (яке дорівнювало нулю, якщо початковий розподіл температури був рівномірним). Цей перепад температур є рушійною силою для подальшого поширення теплоти всередину тіла.

Максимальний перепад температури між поверхнею та віссю тіла зазвичай досягається протягом нестаціонарного процесу нагріву. Інерційний період є початковою фазою цього процесу, коли перепад температур інтенсивно зростає. Хоча абсолютний максимум перепаду може бути досягнутий дещо пізніше завершення власне інерційного періоду (залежно від точного визначення цього періоду), під час інерційного періоду перепад температур є значним і збільшується.

Зважаючи на варіанти відповідей, найбільш точним описом того, що відбувається з перепадом температур поверхня - вісь протягом інерційного періоду нагріву масивного тіла, є його зростання і досягнення значних значень, що відповідає концепції "сягає максимуму" в контексті нестаціонарного процесу.

# Питання 2-1
В механікі рідини і газу рівняння балансу енергії пов'язує зміну питомої внутрішньої енергії, що припадає на одиницю маси в одиницю часу, та...
* A) Роботу переміщення елементарного об'єму середовища під дією неврівноваженого тиску,
* B) Роботу сил в'язкості з деформації елементарного об'єму,
* C) Теплообмін та робота поля зовнішніх сил,
* D) Всі перелічені впливи.

Рівняння балансу енергії для елементарного об'єму рідини або газу є диференціальною формою першого закону термодинаміки. Воно стверджує, що швидкість зміни повної енергії елементарного об'єму дорівнює сумі швидкостей надходження теплоти та роботи, виконаної над цим об'ємом.

Повна питома енергія ($e$) рідини/газу складається з питомої внутрішньої енергії ($e_u$), питомої кінетичної енергії ($\frac{w^2}{2}$) та питомої потенціальної енергії ($\Pi$):
$$e = e_u + \frac{w^2}{2} + \Pi$$

Рівняння балансу енергії в найзагальнішому вигляді (для одиниці маси) можна записати як:
$$\frac{De}{D\tau} = \frac{1}{\rho} \nabla \cdot (\lambda \nabla T) + \frac{1}{\rho} \nabla \cdot (\textbf{w} \cdot \underline{\underline{\sigma}}) + \textbf{w} \cdot \textbf{f}$$

Де:
* $\frac{De}{D\tau}$ — субстанціональна похідна питомої повної енергії ($e$) за часом, що представляє швидкість зміни повної енергії, що припадає на одиницю маси елементарного об'єму.
* $\frac{1}{\rho} \nabla \cdot (\lambda \nabla T)$ — швидкість надходження теплоти за рахунок теплопровідності (теплообмін). $\lambda$ — коефіцієнт теплопровідності, $T$ — температура, $\rho$ — густина.
* $\frac{1}{\rho} \nabla \cdot (\textbf{w} \cdot \underline{\underline{\sigma}})$ — робота, виконана силами, що діють на поверхню елементарного об'єму (робота тиску та в'язких сил). Тут $\textbf{w}$ — вектор швидкості, $\underline{\underline{\sigma}}$ — тензор напружень.
    * Тензор напружень $\underline{\underline{\sigma}}$ може бути розкладений на ізотропну частину (тиск) та девіаторну частину (в'язкі напруження): $\underline{\underline{\sigma}} = -p\underline{\underline{I}} + \underline{\underline{\tau}}$, де $p$ — тиск, $\underline{\underline{I}}$ — одиничний тензор, $\underline{\underline{\tau}}$ — тензор в'язких напружень.
    * Відповідно, член $\frac{1}{\rho} \nabla \cdot (\textbf{w} \cdot \underline{\underline{\sigma}})$ включає в себе **роботу переміщення елементарного об'єму під дією тиску** та **роботу сил в'язкості з деформації**.
* $\textbf{w} \cdot \textbf{f}$ — робота поля зовнішніх масових сил. $\textbf{f}$ — вектор зовнішніх масових сил (наприклад, гравітації) на одиницю маси.

Якщо ми розкладаємо це рівняння на складові і розглядаємо окремо рівняння для питомої внутрішньої енергії ($e_u$), воно набуде вигляду:
$$\frac{De_u}{D\tau} = \frac{1}{\rho} \nabla \cdot (\lambda \nabla T) - p \frac{D}{D\tau}\left(\frac{1}{\rho}\right) + \frac{1}{\rho} (\underline{\underline{\tau}} : \nabla \textbf{w})$$

Де:
* $\frac{De_u}{D\tau}$ — швидкість зміни питомої внутрішньої енергії.
* $\frac{1}{\rho} \nabla \cdot (\lambda \nabla T)$ — член, що описує **теплообмін** (теплопровідність).
* $- p \frac{D}{D\tau}\left(\frac{1}{\rho}\right)$ — робота, виконана силами тиску при зміні об'єму (робота розширення/стиснення). Це **робота переміщення елементарного об'єму середовища під дією неврівноваженого тиску**.
* $\frac{1}{\rho} (\underline{\underline{\tau}} : \nabla \textbf{w})$ — дисипативна функція, що представляє роботу сил в'язкості, яка перетворюється на внутрішню енергію. Це **робота сил в'язкості з деформації елементарного об'єму**.

Крім того, вплив зовнішніх сил (наприклад, гравітації) на повну енергію може опосередковано впливати на внутрішню енергію через перетворення форм енергії. Тому **робота поля зовнішніх сил** також є важливою складовою загального енергетичного балансу.

Таким чином, рівняння балансу енергії (зокрема, його частина, що стосується внутрішньої енергії) дійсно пов'язує зміну питомої внутрішньої енергії з усіма переліченими впливами:

* **Робота переміщення елементарного об'єму середовища під дією неврівноваженого тиску** (пов'язана з членом тиску в тензорі напружень).
* **Робота сил в'язкості з деформації елементарного об'єму** (пов'язана з девіаторною частиною тензора напружень).
* **Теплообмін** (член з теплопровідністю).
* **Робота поля зовнішніх сил** (яка входить до загального рівняння повної енергії та може впливати на внутрішню енергію).

**Висновок:** Всі перелічені у варіантах A, B, C фактори є невід'ємними частинами рівняння балансу енергії, що впливають на зміну внутрішньої енергії рідини або газу.

Правильна відповідь: $\boxed{D}$

# Питання 2-2
В структурі вільного затопленого струменю виділяють "початкову дільницю". Вона характеризується тим, що
* A) Осьова швидкість у поперечному перерізі струменю =0,
* B) Осьова швидкість у поперечному перерізі струменю дорівнює осьової швидкості у вихідному перерізі сопла,
* C) Осьова швидкість у поперечному перерізі струменю дорівнює швидкості на периферії струменю,
* D) Осьова швидкість у поперечному перерізі струменю більше периферійної швидкості.

Вільний затоплений струмінь - це потік рідини або газу, що витікає з сопла (форсунки) в нескінченний (або дуже великий) простір, заповнений тим самим середовищем, що і струмінь, але в стані спокою. Його структура ділиться на дві основні ділянки: початкову та основну.

1.  **Початкова ділянка (або ділянка формування):** Ця ділянка розташована безпосередньо біля вихідного перерізу сопла.
    * **Характеристика:** В межах цієї ділянки струмінь ще не повністю взаємодіяв з навколишнім середовищем. Основною особливістю початкової ділянки є наявність **ядра потенційного потоку (або ядра незміненої швидкості)**. У цьому ядрі швидкість потоку залишається практично незмінною і дорівнює швидкості витоку з сопла. Тобто, осьова швидкість в будь-якому поперечному перерізі в межах цього ядра дорівнює осьовій швидкості у вихідному перерізі сопла ($w_x = w_0$).
    * По краях струменя (на периферії) відбувається змішування з навколишнім середовищем, формуються зони турбулентного змішування (зони зсуву), де швидкість поступово зменшується від швидкості ядра до нуля (швидкості спокійного середовища). Однак, в центральній частині струменя швидкість залишається постійною і високою.
    * Довжина початкової ділянки залежить від режиму течії (ламінарний чи турбулентний) та форми сопла. Для турбулентних струменів вона є відносно короткою.

2.  **Основна ділянка (або ділянка стабілізованого потоку):** Після початкової ділянки ядро потенційного потоку зникає, і струмінь повністю змішується з навколишнім середовищем.
    * **Характеристика:** Осьова швидкість потоку починає зменшуватися вздовж осі струменя. Профіль швидкостей по всьому перерізу стає параболічним (або гаусовим), з максимальним значенням на осі, яке зменшується в міру віддалення від сопла.

**Аналіз варіантів:**

* **A) Осьова швидкість у поперечному перерізі струменю =0:** Це невірно. Осьова швидкість дорівнює нулю лише далеко від струменя в спокійному середовищі.
* **B) Осьова швидкість у поперечному перерізі струменю дорівнює осьової швидкості у вихідному перерізі сопла:** Це вірно для ядра потенційного потоку, яке існує на початковій ділянці. Саме ця характеристика визначає початкову ділянку.
* **C) Осьова швидкість у поперечному перерізі струменю дорівнює швидкості на периферії струменю:** Це невірно. На периферії швидкість значно менша (аж до нуля) через змішування з навколишнім середовищем.
* **D) Осьова швидкість у поперечному перерізі струменю більше периферійної швидкості:** Хоча це технічно вірно (осьова швидкість завжди вища за периферійну в струмені, що розтікається), це не є **характерною особливістю** саме початкової ділянки, яка відрізняє її від основної ділянки. На основній ділянці осьова швидкість також більша за периферійну. Характерною ознакою початкової ділянки є саме збереження постійної швидкості в ядрі, рівній швидкості на виході з сопла.

Таким чином, ключовою характеристикою початкової ділянки вільного затопленого струменя є наявність ядра, де швидкість потоку залишається незмінною і дорівнює швидкості на виході з сопла.

Правильна відповідь: $\boxed{B}$

# Питання 2-3
В якому із варіантів рекуператора: прямоточному чи протиточному можна досягнути більш ефективного, при інших рівних умовах, нагріву середовища, що нагрівають.
* A) В прямоточному,
* B) В протиточному,
* C) Результат буде однаковим.

Рекуператори - це теплообмінні апарати, в яких теплота від гарячого теплоносія передається до холодного теплоносія через стінку, що їх розділяє, без змішування. Існують дві основні схеми руху теплоносіїв: прямоточна та протиточна.

1.  **Прямоточний (паралельний) теплообмінник:**
    * В обох теплоносіїв напрямок руху **однаковий**.
    * Гарячий теплоносій і холодний теплоносій входять в теплообмінник з одного кінця і виходять з іншого.
    * На вході перепад температур є максимальним, що забезпечує інтенсивний теплообмін. Однак, в міру руху вздовж теплообмінника, температура гарячого теплоносія знижується, а холодного - підвищується. В результаті, перепад температур між ними зменшується, і теплообмін стає менш інтенсивним.
    * **Ключовий недолік:** Температура холодного теплоносія на виході з прямоточного теплообмінника **ніколи не може перевищити (або навіть досягти) температури гарячого теплоносія на виході** з цього ж теплообмінника. Це обмежує ступінь нагріву холодного середовища.

2.  **Протиточний теплообмінник:**
    * Теплоносії рухаються в **протилежних напрямках**.
    * Гарячий теплоносій входить з одного кінця, а холодний теплоносій - з протилежного.
    * Завдяки протилежному руху, відносно великий перепад температур підтримується практично по всій довжині теплообмінника. Гарячий теплоносій, який вже віддав більшу частину своєї теплоти, контактує з холодним теплоносієм, який тільки починає нагріватися. І навпаки, найгарячіший холодний теплоносій контактує з найгарячішим гарячим теплоносієм.
    * **Ключова перевага:** У протиточному теплообміннику температура холодного теплоносія на виході **може бути вищою**, ніж температура гарячого теплоносія на виході. Теоретично, вона може наблизитися до температури гарячого теплоносія на вході, що дозволяє досягти максимально можливого нагріву.

**Висновок:**

Через можливість підтримання значного перепаду температур по всій довжині апарату та досягнення більш високих температур нагрітого середовища на виході, **протиточний рекуператор є значно ефективнішим** для нагріву середовища за інших рівних умов (однакова площа теплообміну, однакові теплоносії та їх витрати).

Правильна відповідь: $\boxed{B}$

# Питання 2-4
Вкажіть інтервал теплотворності природного газу, як основного виду палива (МДж/м3).
* A) 12-16,
* B) 33-36,
* C) 25-28,
* D) 3-6

Теплотворність (або питома теплота згоряння) палива – це кількість теплоти, яка виділяється при повному згорянні одиниці його маси або об'єму. Природний газ є одним з найпоширеніших видів палива у світі і широко використовується в промисловості, енергетиці та побуті.

Основною складовою природного газу є метан (CH₄), вміст якого може коливатися від 70% до 99%. Крім метану, природний газ може містити невеликі домішки інших вуглеводнів (етан, пропан, бутан) та негорючих газів (азот, вуглекислий газ).

Питома теплота згоряння природного газу може дещо варіюватися залежно від його конкретного складу та умов вимірювання (температура, тиск). Однак, для стандартних умов (0°C і 101,325 кПа) та типового складу, його теплотворність становить:

* **Нижча теплота згоряння (без урахування теплоти конденсації водяної пари):** зазвичай знаходиться в діапазоні $33-36 \, \text{МДж/м}^3$.
* **Вища теплота згоряння (з урахуванням теплоти конденсації водяної пари):** дещо вища, близько $37-40 \, \text{МДж/м}^3$.

Зазвичай, у теплотехнічних розрахунках, якщо не вказано інше, мають на увазі нижчу теплотворність. Розглянемо надані варіанти:

* A) $12-16 \, \text{МДж/м}^3$: Це значно нижче типових значень для природного газу. Така теплотворність може бути у деяких бідних газів або генераторних газів.
* B) $33-36 \, \text{МДж/м}^3$: Цей діапазон точно відповідає типовим значенням нижчої теплотворності природного газу.
* C) $25-28 \, \text{МДж/м}^3$: Це також значно нижче, ніж у більшості видів природного газу.
* D) $3-6 \, \text{МДж/м}^3$: Це дуже низька теплотворність, характерна для дуже бідних газів.

Отже, інтервал $33-36 \, \text{МДж/м}^3$ є найбільш точним для теплотворності природного газу як основного виду палива.

Правильна відповідь: $\boxed{B}$

# Питання 2-5
Вкажіть основну відмінність вуглекислотної (ВК) та пароводяної (ПВ) конверсії природного газу.

Основна відмінність між **вуглекислотною (ВК)** та **пароводяною (ПВ)** конверсією природного газу полягає у **використовувальному окиснювачі** для перетворення метану (\$CH\_4\$) на синтез-газ (суміш \$H\_2\$ і \$CO\$):

* **Пароводяна конверсія (ПВ)**:
  Окиснювачем є **водяна пара (\$H\_2O\$)**.
  Основна реакція:

  $$
  CH_4 + H_2O \rightarrow CO + 3H_2 \quad (\Delta H > 0)
  $$

  Це **ендотермічний процес**, тобто вимагає підведення тепла.

* **Вуглекислотна конверсія (ВК)**:
  Окиснювачем виступає **вуглекислий газ (\$CO\_2\$)**.
  Основна реакція (реакція ріформінгу з CO₂):

  $$
  CH_4 + CO_2 \rightarrow 2CO + 2H_2 \quad (\Delta H > 0)
  $$

  Також **ендотермічна**.

### Порівняння:

| Ознака              | Пароводяна конверсія (ПВ)             | Вуглекислотна конверсія (ВК)    |
| ------------------- | ------------------------------------- | ------------------------------- |
| Окиснювач           | Водяна пара (\$H\_2O\$)               | Вуглекислий газ (\$CO\_2\$)     |
| Основні продукти    | \$H\_2\$, \$CO\$                      | \$H\_2\$, \$CO\$                |
| Вихід водню         | Вищий (H₂/CO ≈ 3)                     | Нижчий (H₂/CO ≈ 1)              |
| Джерело CO₂         | Не використовується                   | Утилізується CO₂                |
| Енергетичний баланс | Високі енерговитрати                  | Менші при спільному ріформінгу  |
| Застосування        | Водень для аміаку, паливних елементів | Синтез метанолу, утилізація CO₂ |

**Висновок:**
Основна відмінність полягає у типі окиснювача: ПВ-конверсія використовує **пару**, а ВК-конверсія — **вуглекислий газ**. Це впливає на співвідношення продуктів, енергетичні витрати та екологічну доцільність процесу.

# Питання 2-6
Вкажіть фактори, що при незмінній електричній потужності сприяють підвищенню енергоефективності дугової сталеплавильної печі в технологічний період плавки.
* A) Зменшення струму дуги разом із застосуванням технології спіненого шлаку,
* B) Зменшення струму дуги без застосування технології спіненого шлаку,
* C) Збільшення струму дуги без застосування технології спіненого шлаку,
* D) Збільшення струму дуги разом із застосуванням технології спіненого шлаку.

Для підвищення енергоефективності дугової сталеплавильної печі (ДСП) в технологічний період плавки необхідно мінімізувати втрати енергії та максимально ефективно використовувати електричну потужність. Розглянемо фактори, зазначені у варіантах:

1.  **Струм дуги:**
    * **Збільшення струму дуги:** При незмінній електричній потужності (яка є добутком напруги та струму, $P = U \cdot I \cdot \text{cos}\phi$), збільшення струму означає зменшення напруги на дузі. Це призводить до скорочення довжини дуги. Короткі дуги є більш "жорсткими" і стабільними, що зменшує флуктуації потужності та покращує перенос енергії в розплав. Крім того, короткі дуги мають менші променисті втрати тепла в атмосферу печі та футерівку, оскільки основна частина тепла зосереджується безпосередньо в розплаві. Довгі дуги, навпаки, розсіюють більше тепла на стінки та склепіння печі, підвищуючи теплові втрати і знижуючи ефективність.

2.  **Технологія спіненого шлаку:**
    * **Застосування технології спіненого шлаку:** Спінений шлак є одним з найефективніших засобів підвищення енергоефективності ДСП. Шлак спінюється шляхом продувки вуглецевмісних матеріалів (наприклад, графіту, коксу) та кисню в шлак.
        * **Тепловий захист:** Спінений шлак утворює товстий шар, що плаває на поверхні металу. Цей шар діє як чудовий теплоізолятор, зменшуючи втрати тепла випромінюванням від металу до склепіння, стінок печі та в атмосферу. Це значно знижує питоме споживання електроенергії.
        * **Подовження дуги:** Спінений шлак "занурює" дуги в розплав, дозволяючи використовувати довші дуги (за рахунок підвищеної напруги при тій же потужності), які були б неприйнятними без спіненого шлаку через руйнування футерівки. При цьому, дуга знаходиться всередині шлаку, що також сприяє передачі тепла металу і захисту футерівки.
        * **Стабілізація дуги:** Спінений шлак допомагає стабілізувати дугу, зменшуючи її коливання та покращуючи перенос енергії.

**Аналіз варіантів:**

* A) Зменшення струму дуги разом із застосуванням технології спіненого шлаку: Зменшення струму (при незмінній потужності) означає збільшення напруги та довжини дуги. Це призводить до більших променистих втрат, що негативно впливає на ефективність, навіть зі спіненим шлаком.
* B) Зменшення струму дуги без застосування технології спіненого шлаку: Це призводить до найгірших умов, оскільки довгі дуги без захисту шлаку викликають великі втрати тепла і інтенсивне руйнування футерівки.
* C) Збільшення струму дуги без застосування технології спіненого шлаку: Збільшення струму сприяє кращому переносу тепла, але відсутність спіненого шлаку означає значні променисті втрати тепла в атмосферу та футерівку.
* D) Збільшення струму дуги разом із застосуванням технології спіненого шлаку: Це найбільш ефективна комбінація. Збільшення струму (що призводить до коротших, стабільніших дуг) забезпечує кращий перенос енергії в розплав. Застосування спіненого шлаку, у свою чергу, захищає футерівку від променистого тепла дуг і значно зменшує теплові втрати від розплаву, дозволяючи ще більше підвищити енергоефективність.

Правильна відповідь: $\boxed{D}$

# Питання 2-7
Вкажіть формулу, яка виражає основне рівняння гідростатики (у наведених виразах р-тиск, 𝜌 - щільність, m - маса, g - прискорення сили тяжіння, h - линійний розмір (глибина), F- сила).
* A) р = ρghF,
* B) p = ρgh,
* C) p = p0+pgh,
* D) р = mgh

Основне рівняння гідростатики описує залежність тиску в рідині (або газі, якщо його стисливість незначна і густина вважається сталою) від глибини під дією сили тяжіння.

Тиск на певній глибині $h$ у рідині складається з двох частин:
1.  **Тиску на поверхні рідини ($p_0$):** Це може бути атмосферний тиск, якщо рідина знаходиться у відкритій ємності, або тиск газу над рідиною у закритій ємності.
2.  **Гідростатичного тиску, створеного стовпом рідини ($p_{гідрост}$):** Цей тиск зумовлений вагою стовпа рідини над точкою вимірювання.

Гідростатичний тиск $p_{гідрост}$ розраховується за формулою:
$$p_{гідрост} = \rho g h$$
де:
* $\rho$ — густина рідини (кг/м$^3$)
* $g$ — прискорення вільного падіння (м/с$^2$)
* $h$ — глибина (лінійний розмір) від вільної поверхні рідини до точки, де вимірюється тиск (м).

Отже, повний тиск $p$ на глибині $h$ буде сумою тиску на поверхні та гідростатичного тиску:
$$p = p_0 + \rho g h$$

Проаналізуємо надані варіанти:
* A) $p = \rho ghF$: Ця формула включає силу $F$, що є некоректним для розрахунку тиску. Тиск - це сила, що діє на одиницю площі, а не сама сила.
* B) $p = \rho gh$: Ця формула представляє лише гідростатичний тиск стовпа рідини, але не враховує початковий тиск на поверхні рідини ($p_0$). Вона була б правильною, якщо $p_0$ дорівнював би нулю або якщо йшлося б про надлишковий тиск.
* C) $p = p_0 + \rho gh$: Ця формула точно відображає основне рівняння гідростатики, враховуючи як зовнішній тиск на поверхні, так і тиск, створюваний стовпом рідини.
* D) $p = mgh$: Ця формула виражає потенціальну енергію або вагу, а не тиск. Тиск - це сила, поділена на площу, а не просто маса, помножена на прискорення та глибину.

Правильна відповідь: $\boxed{C}$

# Питання 2-8
Вказати в якому з перелічених випадків стосовно коефіцієнту витрати повітря α коефіцієнт використання палива ηквп у нагрівальній печі максимальний:
* A) При $\alpha$=1,
* B) При $\alpha$<1,
* C) При $\alpha$>1,
* D) не впливає на $\eta_{квп}$

Коефіцієнт використання палива ($\eta_{квп}$) у нагрівальній печі відображає ефективність перетворення хімічної енергії палива в корисне тепло, що передається об'єкту нагріву. Він досягає максимуму за умов, що забезпечують найбільш повне згоряння палива з мінімальними втратами теплоти.

Коефіцієнт витрати повітря $\alpha$ (альфа) - це відношення фактичної кількості повітря, поданої для згоряння палива, до теоретично необхідної (стехіометричної) кількості повітря.

1.  **При $\alpha < 1$ (недостатній надлишок повітря):**
    * Згоряння палива відбувається неповністю. Це означає, що частина палива не згоряє (наприклад, утворюється чадний газ CO замість вуглекислого газу CO₂ або сажа), і його хімічна енергія не виділяється як тепло.
    * Наявність продуктів неповного згоряння (CO, H₂, C) у димових газах призводить до значних хімічних втрат теплоти.
    * Ефективність використання палива різко знижується.

2.  **При $\alpha > 1$ (надлишкове повітря):**
    * Згоряння палива відбувається повністю, але в піч подається більше повітря, ніж потрібно.
    * Надлишкове повітря, яке не бере участі в процесі згоряння, нагрівається в печі до температури димових газів.
    * Це нагріте надлишкове повітря відводиться з димовими газами, виносячи з собою значну кількість фізичного тепла, яке є втратами. Чим більше надлишкового повітря (чим більше $\alpha$), тим більші ці втрати.
    * Ефективність використання палива також знижується через зростання фізичних втрат теплоти з відхідними газами.

3.  **При $\alpha = 1$ (теоретично необхідна кількість повітря):**
    * Цей режим теоретично відповідає повному згорянню палива без надлишку повітря.
    * Мінімізуються як хімічні втрати (бо паливо згоряє повністю), так і фізичні втрати (бо немає надлишкового повітря, що виносить тепло).
    * Саме в цьому випадку коефіцієнт використання палива $\eta_{квп}$ є максимальним, оскільки вся хімічна енергія палива перетворюється на тепло, яке потім максимально ефективно передається об'єкту нагріву, а втрати з відхідними газами є мінімальними.

На практиці ідеального $\alpha = 1$ важко досягти і підтримувати, тому зазвичай працюють з невеликим надлишком повітря ($\alpha$ дещо більше 1, наприклад, 1.05 - 1.15) для гарантії повного згоряння, жертвуючи невеликою частиною ефективності заради безпеки та стабільності процесу. Однак, теоретичний максимум ефективності досягається саме при $\alpha = 1$.

Правильна відповідь: $\boxed{A}$

Вказати, який з елементів сопла Лаваля є камерою гальмування.
* A) Деталь 1
* B) Деталь 3
* C) Деталь 2
* D) Деталь F

---
Сопло Лаваля є термодинамічним пристроєм, призначеним для прискорення потоку газу до надзвукових швидкостей. Конструктивно воно складається з трьох основних частин, що послідовно розташовані за напрямком руху робочого тіла:

1.  **Камера гальмування (передсопло, ресивер):** Це початкова частина сопла, де газ перед входом у звужувальну частину має відносно низьку швидкість. В ідеальному випадку, швидкість газу в цій камері прямує до нуля. Тут макроскопічна кінетична енергія потоку мінімальна, а параметри газу (тиск $P_0$ та температура $T_0$) називаються **параметрами гальмування** або **повними параметрами**. На наданій схемі ця частина позначена цифрою **1**. Вона розташована перед конфузором і слугує для стабілізації потоку та забезпечення рівномірного надходження робочого тіла в наступні частини сопла.

2.  **Конфузор (звужувальна частина):** Це частина сопла, де площа поперечного перерізу поступово зменшується. В конфузорі дозвуковий потік прискорюється, а його статичний тиск та температура знижуються. На схемі позначена цифрою **2**.

3.  **Горловина:** Це місце з мінімальною площею поперечного перерізу, розташоване між конфузором та дифузором. У горловині, за умови досягнення критичного співвідношення тисків, швидкість потоку досягає **локальної швидкості звуку** (число Маха $M=1$). На схемі позначена літерою **F** та діаметром **d**.

4.  **Дифузор (розширювальна частина):** Це частина сопла, де площа поперечного перерізу поступово збільшується. Якщо в горловині потік досяг звукової швидкості, то в дифузорі він продовжує розширюватися та прискорюватися до надзвукових швидкостей, при цьому його статичний тиск та температура продовжують знижуватися. На схемі позначена цифрою **3**.

**Висновок:**
Деталь, позначена цифрою **1** на схемі, відповідає **камері гальмування**. Це вхідна частина сопла, де газ має початкові (загальмовані) параметри перед тим, як почати прискорюватися в конфузорі. Тому варіант відповіді **A) Деталь 1** є правильним.

**Правильна відповідь:** A) Деталь 1

Теплотворність (або нижча теплота згоряння, $Q_н$) є ключовою характеристикою палива, що вказує на кількість теплоти, яка виділяється при повному згорянні одиниці об'єму (для газоподібного палива, зазвичай м³) або маси палива. Розглянемо орієнтовні значення теплотворності для вказаних газів:

1.  **Доменний газ:** Це побічний продукт доменного виробництва чавуну. 2.  **Конвертерний газ:** Це газ, що утворюється під час конвертерного процесу виробництва сталі. Його основний горючий компонент – оксид вуглецю ($CO$), також містить значну кількість азоту. Він має дещо вищу теплотворність порівняно з доменним газом, оскільки концентрація $CO$ зазвичай вища, а баластних газів (якщо не використовується кисневе дуття в чистому вигляді) може бути менше, або ж він проходить очистку та збагачення.

3.  **Коксовий газ:** Це побічний продукт процесу коксування кам'яного вугілля. Він містить значну кількість горючих компонентів, таких як водень ($H_2$) та метан ($CH_4$), а також оксид вуглецю ($CO$).

    * Орієнтовна теплотворність: **31 - 38 МДж/м³** (для газу з високим вмістом метану, часто приймають середнє значення близько 35 МДж/м³).

# Питання 2-11
Встановіть у порядку зменшення по масі одиниці об'єму наступні характеристики палива.
* A) Робоча маса - суха маса - горюча маса - органічна маса,
* B) Органічна маса - горюча маса - робоча маса - суха маса,
* C) Органічна маса - суха маса - робоча маса - горюча маса,
* D) суха маса - робоча маса - горюча маса - органічна маса.

---
Для встановлення правильного порядку зменшення маси одиниці об'єму для різних характеристик палива, необхідно звернутися до їх визначень та складу. Розглянемо ці характеристики для твердого або рідкого палива, виходячи з того, що "робоча маса" є вихідною точкою, що представляє паливо в тому стані, в якому воно надходить до споживача. Маса одиниці об'єму в даному контексті означає, що ми розглядаємо фіксований об'єм палива в його початковому (робочому) стані і визначаємо, яку масу становлять його різні складові або стани.

1.  **Робоча маса ($M_P$ або $m^r$):** Це маса палива у тому вигляді, в якому воно надходить до споживача. Вона включає всі компоненти: горючу частину, внутрішню та зовнішню вологу ($W^P$), а також мінеральні домішки, які при спалюванні утворюють золу ($A^P$).
    Згідно з наданим зображенням: $M_P = C_P + H_P + O_P + N_P + S_P + A_P + W_P$. Це є 100% маси палива "як є".

2.  **Суха маса ($M_C$ або $m^c$):** Це маса палива після видалення з нього усієї вологи ($W^P$). Тобто, суха маса не містить води.
    $M_C = M_P - W^P$.
    Оскільки робоче паливо зазвичай містить деяку кількість вологи ($W^P > 0$), то суха маса завжди буде меншою (або рівною, якщо $W^P = 0$) за робочу масу.
    $M_P \ge M_C$.

3.  **Горюча маса ($M_Г$ або $m^г$):** Це маса палива після видалення з нього вологи ($W^P$) та золи ($A^P$). Горюча маса складається з органічної частини палива та горючої сірки ($S_Г$).
    $M_Г = M_C - A^C = M_P - W^P - A^P$ (де $A^C$ - вміст золи в сухій масі, $A^P$ - вміст золи в робочій масі).
    Оскільки паливо (окрім деяких спеціальних випадків) містить золу ($A^P > 0$), то горюча маса буде меншою за суху масу.
    $M_C \ge M_Г$.

4.  **Органічна маса ($M_O$ або $m^o$):** Це частина горючої маси, що залишається після умовного видалення з неї сірки ($S_Г$). Вона складається переважно з вуглецю ($C$), водню ($H$), кисню ($O$) та азоту ($N$), що входять до складу органічних сполук палива.
    $M_O = M_Г - S_Г$.
    Оскільки паливо часто містить сірку ($S_Г > 0$), органічна маса буде меншою (або рівною, якщо $S_Г = 0$) за горючу масу.
    $M_Г \ge M_O$.

**Ранжування за зменшенням маси:**

Виходячи з наведених визначень, для одиниці об'єму палива в його робочому стані, маса різних його характеристик буде зменшуватися в такому порядку:
1.  **Робоча маса** (найбільша, включає вологу, золу та горючі компоненти).
2.  **Суха маса** (менша за робочу на величину маси вологи).
3.  **Горюча маса** (менша за суху на величину маси золи).
4.  **Органічна маса** (менша за горючу на величину маси сірки).

Таким чином, послідовність за зменшенням маси:
Робоча маса $\ge$ Суха маса $\ge$ Горюча маса $\ge$ Органічна маса.

Зазвичай, для твердих та рідких палив вміст вологи, золи та сірки є суттєвим, тому нерівності будуть строгими:
Робоча маса > Суха маса > Горюча маса > Органічна маса.

# Питання 2-12
Втрати тиску на тертя та місцеві опори при русі середовища в каналах пропорційні
* A) Швидкості,
* B) Квадрату швидкості,
* C) Логарифму швидкості,
* D) Кореню квадратному із швидкості.
---
Втрати тиску при русі рідин або газів у каналах (трубопроводах, повітроводах тощо) складаються з двох основних компонентів: втрат на тертя по довжині каналу та втрат на місцевих опорах (вигини, звуження, розширення, арматура тощо).

1.  **Втрати тиску на тертя:**
    Ці втрати описуються **формулою Дарсі-Вейсбаха**:
    $$\Delta P_f = \lambda \frac{L}{D_h} \frac{\rho v^2}{2}$$
    де:
    * $\Delta P_f$ – втрати тиску на тертя, Па;
    * $\lambda$ – коефіцієнт гідравлічного тертя (коефіцієнт Дарсі), безрозмірний;
    * $L$ – довжина каналу, м;
    * $D_h$ – гідравлічний діаметр каналу, м;
    * $\rho$ – густина середовища, кг/м³;
    * $v$ – середня швидкість потоку, м/с.

    З цієї формули видно, що втрати тиску на тертя ($\Delta P_f$) прямо пропорційні **квадрату середньої швидкості потоку ($v^2$)**, за умови, що інші параметри (зокрема $\lambda$, який може мати слабку залежність від швидкості через число Рейнольдса в певних режимах течії, але для розвиненого турбулентного потоку ця залежність не змінює основного характеру пропорційності до $v^2$) залишаються відносно сталими.

2.  **Втрати тиску на місцеві опори:**
    Ці втрати розраховуються за **формулою Вейсбаха (для місцевих опорів)**:
    $$\Delta P_m = \zeta \frac{\rho v^2}{2}$$
    де:
    * $\Delta P_m$ – втрати тиску на місцевому опорі, Па;
    * $\zeta$ – коефіцієнт місцевого опору, безрозмірний (залежить від геометрії опору);
    * $\rho$ – густина середовища, кг/м³;
    * $v$ – характерна швидкість потоку (зазвичай у перерізі перед або після місцевого опору, або в звуженому перерізі), м/с.

    Аналогічно до втрат на тертя, втрати тиску на місцевих опорах ($\Delta P_m$) також прямо пропорційні **квадрату швидкості потоку ($v^2$)**.

**Загальний висновок:**
Обидві складові гідравлічних втрат – як втрати на тертя по довжині, так і втрати на місцевих опорах – пропорційні динамічному тиску потоку ($\rho v^2 / 2$), а отже, пропорційні **квадрату швидкості** середовища. Це є фундаментальним положенням гідродинаміки для турбулентних потоків, які найчастіше зустрічаються в інженерній практиці. Для ламінарних потоків залежність втрат тиску на тертя від швидкості є лінійною, однак питання зазвичай стосується загального випадку або турбулентного режиму, де втрати значно вищі та мають квадратичну залежність.

---
Правильна відповідь: $\boxed{B}$

Дайте якісну характеристику режимів електродугової генерації теплоти в дуговій сталеплавильній печі при незмінній електричній потужності:

* D) Робота на коротких дугах в енергетичний період плавки та на довгих дугах в технологічний період.

---
Розглянемо якісну характеристику режимів електродугової генерації теплоти в дуговій сталеплавильній печі (ДСП) при незмінній електричній потужності:

Електродугова сталеплавильна піч має два основні періоди плавки: енергетичний (період плавлення шихти) та технологічний (період доведення металу до потрібного складу й температури). У кожен період режими роботи дуги (довгі або короткі) впливають на теплові та технологічні процеси.

**Енергетичний період.** У цей період основне завдання — швидко розплавити шихту (металобрухт або інші матеріали). Робота на довгих дугах є оптимальною, оскільки довга дуга забезпечує більший тепловий потік, спрямований на шихту, через високу напругу та інтенсивне випромінювання. Це сприяє швидкому плавленню, хоча й супроводжується більшими тепловтратами через розсіювання енергії та підвищений знос футерівки печі. При незмінній електричній потужності довгі дуги дозволяють ефективно передати енергію для нагрівання й плавлення матеріалів.

**Технологічний період.** На цьому етапі шихта вже розплавлена, і основна мета — підтримувати температуру рідкого металу, проводити рафінування (видалення домішок) і легування. Робота на коротких дугах є кращою, оскільки коротка дуга забезпечує більш концентроване тепловиділення в зоні металу, зменшуючи тепловтрати на стінки печі та футерівку. Це сприяє стабільному нагріванню ванни, кращому перемішуванню металу та ефективному протіканню хімічних реакцій (наприклад, окислення домішок). Короткі дуги також знижують електромагнітний шум і знос електродів.

**Аналіз варіантів.**
- **A (довгі дуги в обидва періоди):** У технологічний період довгі дуги призводять до надмірних тепловтрат і зниження ефективності рафінування, що не відповідає вимогам процесу.
- **B (довгі дуги в енергетичний період, короткі в технологічний):** Цей режим є оптимальним, оскільки відповідає потребам обох періодів: швидке плавлення на довгих дугах і контрольоване рафінування на коротких.
- **C (короткі дуги в обидва періоди):** У енергетичний період короткі дуги менш ефективні для плавлення шихти, оскільки теплопередача менш інтенсивна, що уповільнює процес.
- **D (короткі дуги в енергетичний період, довгі в технологічний):** Цей режим суперечить технологічним потребам, адже короткі дуги в енергетичний період уповільнюють плавлення, а довгі дуги в технологічний період погіршують контроль над процесом рафінування.

Таким чином, відповідь B є правильною, оскільки вона враховує особливості теплопередачі та технологічні вимоги на різних етапах плавки в ДСП, забезпечуючи оптимальну ефективність процесу при незмінній електричній потужності.

---
Правильна відповідь: $\boxed{B}$

Електроенергія використовується лише для плавлення та рафінування (5–10 ГДж/т сталі). Це робить цей процес найбільш енергоефективним серед розглянутих, з витратами близько 10–12 ГДж/т.

Отже, послідовність енерговитрат у порядку зменшення виглядає так: А (20–25 ГДж/т) > Г (18–22 ГДж/т) > Б (15–20 ГДж/т) > В (10–12 ГДж/т). Отже правильна відповідь A: А–Г–Б–В, оскільки вона коректно відображає спадну послідовність питомих енерговитрат для зазначених технологічних маршрутів.

# Питання 2-16
Для реалізації розрахункового режиму сопла Лаваля щодо продувки рідкої сталі киснем одна із перелічених умов є зайвою. Яка?
* A) Локальна швидкість звуку в критичному перерізі,
* B) Тиск в камері гальмування вище критичного,
* C) Витрата кисню відповідає технологічному регламенту,
* D) Тиск у вихідному перерізі дифузора дорівнює тиску в робочому просторі печі.

---
Розглянемо умови, необхідні для реалізації розрахункового режиму сопла Лаваля при продувці рідкої сталі киснем, та визначимо, яка з них є зайвою:

Сопло Лаваля є конічним соплом, яке забезпечує надзвукову швидкість потоку газу (кисню) завдяки особливостям його геометрії: звуження до критичного перерізу (шийки) і подальше розширення в дифузорі. Для досягнення розрахункового режиму, коли швидкість потоку перевищує звукову, ключовими є такі параметри:

- **A) Локальна швидкість звуку в критичному перерізі.** Ця умова є необхідною, оскільки в критичному перерізі швидкість потоку досягає локальної швидкості звуку (Маха = 1), що є основою переходу до надзвукового режиму в дифузорі. Без цього параметра розрахунок режиму сопла неможливий.

- **B) Тиск у камері гальмування вище критичного.** Критичний тиск визначається співвідношенням із тиском у вихідному перерізі та залежить від співвідношення площ сопла. Для забезпечення надзвукового потоку тиск у камері гальмування має перевищувати критичне значення, що дозволяє газу прискорюватися. Ця умова є фундаментальною для роботи сопла Лаваля.

- **D) Тиск у вихідному перерізі дифузора дорівнює тиску в робочому просторі печі.** Це умова збігу тисків, яка забезпечує безперервний надзвуковий потік без ударних хвиль. Якщо тиск у вихідному перерізі не відповідає тиску в печі, режим потоку може порушитися (наприклад, виникнення зворотних хвиль), що робить цю умову важливою для розрахунку.

- **C) Витрата кисню відповідає технологічному регламенту.** Хоча витрата кисню є важливим параметром для технологічного процесу продувки сталі (наприклад, для окислення домішок), вона не є прямою умовою реалізації розрахункового режиму сопла Лаваля. Режим роботи сопла визначається геометрією, тиском і швидкістю потоку, а не конкретною витратою, яка регулюється зовнішніми системами подачі газу й залежить від металу та технології, а не від гідродинаміки сопла.

Отже, умова C є зайвою для реалізації розрахункового режиму сопла Лаваля, оскільки вона стосується технологічного аспекту, а не фізичних параметрів, що визначають гідродинаміку потоку в соплі.

---
Правильна відповідь: $\boxed{C}$

# Питання 2-17
З підвищенням швидкості руху реальної рідини в каналі в 3 рази, наскільки зміниться втрата тиску на тертя
* A) в 3 рази,
* B) в 1,73 рази,
* C) в 9 разів,
* D) не зміниться.

---
Розглянемо, як змінюється втрата тиску на тертя при підвищенні швидкості руху реальної рідини в каналі в 3 рази:

Втрата тиску на тертя в каналі для реальної рідини описується рівнянням Дарсі-Вейсбаха або аналогічними емпіричними формулами, такими як рівняння Гagena-Пуазейля для ламінарного потоку чи залежність від коефіцієнта тертя для турбулентного потоку. У загальному випадку для турбулентного потоку, який є типовим для реальних рідин у промислових умовах, втрата тиску на тертя пропорційна квадрату швидкості рідини (v²). Це випливає з рівняння Дарсі-Вейсбаха:

$$
\Delta P = f \cdot \frac{L}{D} \cdot \frac{\rho v^2}{2},
$$

де:
- $\Delta P$ — втрата тиску,
- $f$ — коефіцієнт тертя,
- $L$ — довжина каналу,
- $D$ — діаметр каналу,
- $\rho$ — густина рідини,
- $v$ — швидкість рідини.

Якщо швидкість $v$ збільшується в 3 рази, то $v^2$ зросте в $3^2 = 9$ разів, за умови, що інші параметри (коефіцієнт тертя $f$, довжина $L$, діаметр $D$ і густина $\rho$) залишаються сталими. Отже, втрата тиску на тертя пропорційно збільшиться в 9 разів.

Аналіз варіантів:
- **A) У 3 рази:** Відповідає лінійній залежності, що характерно для ламінарного потоку, але не для реальної рідини в типових умовах.
- **B) У 1,73 рази:** Ця величина близька до $\sqrt{3}$, що не має фізичного обґрунтування для тертя в даному контексті.
- **C) У 9 разів:** Відповідає квадратичній залежності від швидкості, що є правильним для турбулентного потоку.
- **D) Не зміниться:** Протирічить фізичним законам, оскільки втрата тиску залежить від швидкості.

Таким чином, правильна відповідь C, оскільки підвищення швидкості в 3 рази призводить до збільшення втрати тиску на тертя в 9 разів.

---
Правильна відповідь: $\boxed{C}$

# Питання 2-18
Модуль стиснення середовища означає
* A) Відношення маси до об'єму,
* B) Відношення об'єму до маси,
* C) Зміни тиску до відносної зміни об'єму,
* D) Зміни тиску до відносної зміни маси.

---
Розглянемо, що означає модуль стиснення середовища:

Модуль стиснення (або об’ємний модуль пружності) є фізичною характеристикою, яка описує здатність середовища протистояти стисненню під дією тиску. Він визначається як відношення приросту тиску до відносної зміни об’єму середовища, що виражається формулою:

$$
K = - \frac{\Delta P}{\Delta V / V},
$$

де $K$ — модуль стиснення, $\Delta P$ — зміна тиску, $\Delta V / V$ — відносна зміна об’єму, а знак "мінус" вказує на те, що при збільшенні тиску об’єм зменшується.

- **A) Відношення маси до об’єму:** Це визначення густини ($\rho = m/V$), а не модуля стиснення.
- **B) Відношення об’єму до маси:** Це обернена величина до густини ($V/m$), що також не стосується модуля стиснення.
- **C) Зміни тиску до відносної зміни об’єму:** Це коректне визначення модуля стиснення, яке відповідає формулі $K = - \frac{\Delta P}{\Delta V / V}$.
- **D) Зміни тиску до відносної зміни маси:** Маса середовища зазвичай не змінюється при стисненні (для нестисливих або слабко стисливих середовищ), тому це визначення некоректне.

Отже, правильна відповідь C, оскільки модуль стиснення середовища визначається як відношення зміни тиску до відносної зміни об’єму.

---
Правильна відповідь: $\boxed{C}$

# Питання 2-19
На якій глибині гідростатичний тиск води буде дорівнювати атмосферному тиску на рівні моря.
* A) 15 м,
* B) 10 м,
* C) 760 мм,
* D) 1,48 м.

---
Розглянемо, на якій глибині гідростатичний тиск води дорівнює атмосферному тиску на рівні моря:

Гідростатичний тиск на певній глибині визначається формулою:

$$
P = \rho g h,
$$

де:
- $P$ — гідростатичний тиск,
- $\rho$ — густина води (для прісної води приблизно $1000 \, \text{кг/м}^3$),
- $g$ — прискорення вільного падіння ($9,81 \, \text{м/с}^2$),
- $h$ — глибина.

Атмосферний тиск на рівні моря становить приблизно $101325 \, \text{Па}$ (або $101,325 \, \text{кПа}$), що еквівалентно $760 \, \text{мм рт. ст.}$. Для спрощення розрахунків приймемо, що $1 \, \text{атм} \approx 100000 \, \text{Па}$.

Тиск на глибині $h$ включає атмосферний тиск плюс гідростатичний тиск, але в завданні йдеться про глибину, де гідростатичний тиск дорівнює атмосферному тиску. Отже, потрібно знайти $h$, при якому:

$$
\rho g h = P_{\text{атм}}.
$$

Підставимо значення:

$$
1000 \cdot 9,81 \cdot h = 101325.
$$

Розв’яжемо для $h$:

$$
h = \frac{101325}{1000 \cdot 9,81} \approx \frac{101325}{9810} \approx 10,33 \, \text{м}.
$$

Оскільки $10,33 \, \text{м}$ найближче до значення $10 \, \text{м}$ з наданих варіантів, а розрахунок залежить від округлення та умов (наприклад, прісна вода vs морська вода, де $\rho \approx 1025 \, \text{кг/м}^3$), приймаємо $10 \, \text{м}$ як практичне значення для прісної води з урахуванням стандартних наближень.

> **Примітка:**
>
> * Для морської (солоної) води з густиною $\rho\approx1025\;\mathrm{кг/м^3}$ глибина трохи менша:
>
>   $$
>   h \approx \frac{101\,325}{1025\cdot9{,}81}\approx10{,}1\;\mathrm{м}.
>   $$
> * Часто в практиці приймають «10 м» як приблизне значення.


Аналіз варіантів:
- **A) 15 м:** Надмірна глибина, де тиск значно перевищить атмосферний.
- **B) 10 м:** Близьке значення, яке відповідає розрахунку.
- **C) 760 мм:** Це одиниця тиску ($760 \, \text{мм рт. ст.}$), а не глибина.
- **D) 1,48 м:** Недостатня глибина, де тиск значно менший за атмосферний.

Отже, правильна відповідь B: 10 м, оскільки гідростатичний тиск на цій глибині приблизно дорівнює атмосферному тиску на рівні моря.

---
Правильна відповідь: $\boxed{B}$

# Питання 2-21
Назвіть основну причину використання електричного струму зниженої частоти в індукційних тигельних сталеплавильних печах великої місткості.
* A) Збільшення глибини проникнення електромагнітної енергії в рідкий метал,
* B) Покращення коефіцієнту потужності електричного ланцюгу (косинус фі),
* C) Можливість нагріву шлаку виділенням електромагнітної енергії,
* D) Зниження потоку розсіяння енергії в системі індуктор - тигель.

---
Основна причина використання електричного струму зниженої частоти в індукційних тигельних сталеплавильних печах великої місткості полягає в **збільшенні глибини проникнення електромагнітної енергії в рідкий метал**. Давайте розберемо це детальніше.

### Що таке глибина проникнення?
Глибина проникнення ($δ$) — це відстань від поверхні металу, на якій інтенсивність електромагнітного поля зменшується приблизно на 63%. Вона обчислюється за формулою:

$$
\delta = \sqrt{\frac{\rho}{\pi f \mu}}
$$

де:
- $ρ$ — питомий опір металу,
- $f$ — частота струму,
- $μ$ — магнітна проникність металу.

З формули видно, що при **зменшенні частоти $f$** значення $δ$ **збільшується**. Це означає, що електромагнітна енергія проникає глибше в об’єм рідкого металу.

### Чому це важливо для печей великої місткості?
У печах великої місткості об’єм металу значний, тому для ефективного плавлення необхідно забезпечити **рівномірний нагрів** по всьому об’єму, а не лише на поверхні. Струм зниженої частоти дозволяє:
- Забезпечити глибше проникнення електромагнітного поля.
- Рівномірно розподілити тепло в металі.
- Покращити перемішування рідкого металу.
- Підвищити ефективність плавлення та якість сталі.

### Чому інші варіанти менш релевантні?
1. **Покращення коефіцієнта потужності ($cos \phi$):**
   Зниження частоти може незначно вплинути на коефіцієнт потужності, але це не є основною причиною, оскільки ефект мінімальний порівняно з впливом на глибину проникнення.

2. **Нагрів шлаку електромагнітною енергією:**
   Шлак — діелектрик, який не проводить струм, тому індукційний нагрів на нього не діє. Зміна частоти тут не відіграє ролі.

3. **Зниження потоку розсіяння енергії:**
   Зниження частоти може дещо зменшити втрати енергії, але це побічний ефект, а не головна мета.

### Висновок
Використання струму зниженої частоти в індукційних печах великої місткості дозволяє **збільшити глибину проникнення електромагнітної енергії в рідкий метал**, що забезпечує ефективний і рівномірний нагрів. Тому правильна відповідь — **A**.

---

# Питання 2-22
Одиниця виміру коефіцієнту дифузії

---
Коефіцієнт дифузії характеризує інтенсивність молекулярного переносу речовини в середовищі під дією градієнта концентрації. З позицій Міжнародної системи одиниць (СІ), одиницею виміру коефіцієнта дифузії є м²/с (квадратний метр на секунду).

Фізичний зміст цієї величини безпосередньо випливає з першого закону Фіка, який описує потік речовини через одиничну площу за одиницю часу:

J = -D × (∂c/∂x)

де:
- J — потік речовини (кг/(м²·с))
- D — коефіцієнт дифузії (м²/с)
- ∂c/∂x — градієнт концентрації (кг/м⁴)

Коефіцієнт дифузії можна інтерпретувати як площу, через яку здійснюється дифузійний перенос речовини за одиницю часу. Величина D залежить від багатьох факторів, включаючи:

1. Температуру системи
2. Тиск (для газів)
3. В'язкість середовища
4. Розмір молекул або частинок, що дифундують
5. Природу взаємодій між частинками

У газах коефіцієнт дифузії має порядок 10⁻⁵ м²/с, у рідинах — 10⁻⁹ м²/с, а в твердих тілах може досягати значень порядку 10⁻¹² м²/с, що відображає значну різницю в рухливості частинок у різних агрегатних станах речовини.

# Питання 2-23
Охарактеризуйте якісний вплив потенціалу іонізації розрядного проміжку на температуру електричної дуги в дуговій сталеплавильній печі
* A) Підвищений потенціал іонізації призводить до зменшення температури дуги,
* B) Зміна потенціалу іонізації практично не впливає на температуру дуги в межах практичних умов роботи ДСП,
* C) Зростання потенціалу іонізації сприяє збільшенню температури дуги,
* D) Підвищений потенціал іонізації призводить до зникнення дуги.

---

Температура електричної дуги в дуговій сталеплавильній печі (ДСП) безпосередньо пов'язана з енергією, що виділяється в розрядному проміжку. Ця енергія, в свою чергу, залежить від процесів іонізації газів та парів металів, що знаходяться в цьому проміжку.

**Потенціал іонізації** – це мінімальна енергія, необхідна для відриву електрона від атома або молекули, перетворюючи їх на позитивно заряджений іон. Цей параметр є ключовим для розуміння процесів у плазмі дуги.

1.  **Процес іонізації та провідність дуги:** Для підтримання електричної дуги необхідна наявність достатньої кількості вільних носіїв заряду (електронів та іонів) у розрядному проміжку. Ці носії заряду утворюються в результаті іонізації атомів і молекул газу (зазвичай повітря, а також парів металу та шлаку). Чим вищий потенціал іонізації, тим більше енергії потрібно для створення кожного іона.

2.  **Енергетичний баланс дуги:** Електрична дуга є високотемпературною плазмою. Енергія, що підводиться до дуги від джерела живлення, витрачається на нагрівання електродів, розплавлення металу, теплові втрати в навколишнє середовище, а також на процеси іонізації та рекомбінації в самій дузі.

3.  **Зв'язок потенціалу іонізації та температури:**
    * Якщо потенціал іонізації газів у розрядному проміжку **вищий**, то для досягнення тієї ж концентрації заряджених частинок (а отже, тієї ж електропровідності плазми, необхідної для проходження заданого струму) потрібно витратити **більше енергії** на сам процес іонізації.
    * Ця додаткова енергія, що "вкладається" в подолання вищого іонізаційного бар'єру, призводить до **збільшення загальної енергії плазми** і, як наслідок, до **підвищення її температури**. Іншими словами, при вищому потенціалі іонізації дуговий розряд стає "енергетично більш насиченим" для підтримки свого існування при заданому струмі, що проявляється у вищій температурі.

    Математично, хоча пряма залежність температури від потенціалу іонізації є складною і описується рівняннями стану плазми (наприклад, рівнянням Саха), якісний зв'язок очевидний: вищий енергетичний поріг для утворення носіїв заряду вимагає більшої середньої кінетичної енергії частинок плазми, що і визначає її температуру.

4.  **Розгляд інших варіантів:**
    * **A) Підвищений потенціал іонізації призводить до зменшення температури дуги:** Це суперечить фізиці процесу, оскільки для іонізації речовин з вищим потенціалом іонізації потрібна більша енергія.
    * **B) Зміна потенціалу іонізації практично не впливає на температуру дуги в межах практичних умов роботи ДСП:** Хоча інші фактори (сила струму, напруга, склад атмосфери печі) також суттєво впливають на температуру, потенціал іонізації є фундаментальною характеристикою плазмоутворюючого середовища і має помітний вплив. Склад газів у дуговому проміжку може змінюватися (наприклад, за рахунок випаровування компонентів шихти або електродів), що призводить до зміни ефективного потенціалу іонізації та, відповідно, температури.
    * **D) Підвищений потенціал іонізації призводить до зникнення дуги:** Це можливо лише в тому випадку, якщо напруга джерела живлення недостатня для подолання значно зрослого потенціалу іонізації і підтримки розряду. Однак, за умови достатньої потужності джерела, дуга буде горіти, але при вищій температурі.

**Висновок:** Зростання потенціалу іонізації компонентів розрядного проміжку вимагає більших енергетичних затрат на утворення та підтримання плазми, що веде до підвищення температури електричної дуги. Це важливо враховувати при аналізі теплових процесів та оптимізації режимів роботи дугових сталеплавильних печей.

# Питання 2-24
Перенос імпульсу в рухомому середовищі забезпечується
* A) Силами в'язкого тертя,
* B) Силою тяжіння,
* C) Градієнтом температури,
* D) Силою поверхневого натягу.
ANSWER: A

---
Перенос імпульсу в рухомому середовищі (рідині або газі) є фундаментальним процесом, що описує, як рух передається від одних шарів середовища до інших. Цей процес тісно пов'язаний з внутрішніми силами, що діють між частинками середовища.

1.  **Визначення імпульсу та його переносу:** Імпульс (або кількість руху) є векторною фізичною величиною, що дорівнює добутку маси тіла (або елементарного об'єму середовища) на його швидкість. В рухомому середовищі різні його частини можуть мати різні швидкості. Перенос імпульсу – це процес обміну кількістю руху між сусідніми шарами середовища, що рухаються з різними швидкостями.

2.  **Роль сил в'язкого тертя:**
    * **В'язкість** – це властивість рідин і газів чинити опір переміщенню одних їхніх частин відносно інших. Цей опір проявляється у вигляді **сил в'язкого (внутрішнього) тертя**.
    * Коли сусідні шари рідини або газу рухаються з різними швидкостями (тобто існує градієнт швидкості), між ними виникають дотичні напруження, зумовлені силами в'язкого тертя.
    * Шар, що рухається швидше, намагається "прискорити" повільніший шар завдяки цим силам, передаючи йому частину свого імпульсу. Навпаки, повільніший шар "гальмує" швидший, забираючи у нього імпульс.
    * Цей обмін імпульсом між шарами, опосередкований силами в'язкого тертя, і є механізмом переносу імпульсу в рухомому середовищі. Математично цей процес описується законом в'язкості Ньютона (для ньютонівських рідин), який пов'язує дотичні напруження з градієнтом швидкості та динамічною в'язкістю середовища: $\tau = \mu \frac{du}{dy}$, де $\tau$ – дотичне напруження, $\mu$ – динамічна в'язкість, а $\frac{du}{dy}$ – градієнт швидкості.

3.  **Розгляд інших варіантів:**
    * **B) Силою тяжіння:** Сила тяжіння діє на всі елементи середовища, що мають масу, і може бути причиною руху (наприклад, конвекція або течія під дією гравітації). Однак, вона не є безпосереднім механізмом переносу імпульсу *всередині* рухомого середовища між його шарами, що рухаються відносно один одного. Перенос імпульсу стосується внутрішніх взаємодій.
    * **C) Градієнтом температури:** Градієнт температури є рушійною силою для переносу теплоти (теплопровідність, конвекція). Він може спричиняти рух середовища (природна конвекція) через зміну густини, але сам по собі не є силою, що безпосередньо забезпечує обмін імпульсом між шарами рідини. Хоча в деяких випадках (наприклад, ефект термофорезу) градієнт температури може викликати рух частинок, основний механізм переносу імпульсу в суцільному середовищі пов'язаний з в'язкістю.
    * **D) Силою поверхневого натягу:** Сили поверхневого натягу діють на межі поділу фаз (наприклад, рідина-газ) і прагнуть скоротити площу цієї поверхні. Вони важливі для капілярних явищ, утворення крапель тощо, але не є основним механізмом переносу імпульсу в об'ємі рухомого середовища.

**Висновок:** Саме сили в'язкого тертя, що виникають між шарами рідини або газу, які рухаються з різними швидкостями, відповідають за перерозподіл та перенос імпульсу в рухомому середовищі. Це ключовий процес, що визначає характер течії, профіль швидкостей та опір руху.

# Питання 2-25
Пірометричний коефіцієнт в процесі горіння палива враховує
* A) Втрати енергії випромінюванням продуктами згоряння,
* B) Втрати енергії палива теплопровідністю при згорянні,
* C) Конвективні втрати енергії палива, що згоряє,
* D) Умови змішування палива з окиснювачем при згорянні.

---
В теорії горіння та теплотехнічних розрахунках, пов'язаних з високотемпературними процесами, важливо точно визначати температуру продуктів згоряння. Теоретична (адіабатна) температура горіння розраховується за умови, що вся теплота, яка виділяється при згорянні, йде на нагрівання продуктів реакції, і немає жодних тепловтрат. Однак в реальних умовах завжди існують втрати теплоти.

1.  **Реальна температура горіння:** Фактична температура полум'я або продуктів згоряння завжди нижча за теоретичну через різні види тепловтрат. Одним із значущих механізмів тепловтрат у високотемпературних процесах, таких як горіння, є **теплове випромінювання**.

2.  **Пірометричний коефіцієнт ($\varphi$):** Для врахування зниження температури продуктів згоряння через тепловтрати, зокрема через випромінювання, вводиться поняття **пірометричного коефіцієнта** (також іноді називають коефіцієнтом тепловтрат або коефіцієнтом збереження тепла). Цей коефіцієнт показує, яка частка теплоти, що виділилася, дійсно йде на нагрівання продуктів згоряння, а яка втрачається.

    * Дійсна (або калориметрична) температура продуктів згоряння ($T_д$) пов'язана з теоретичною (адіабатною) температурою горіння ($T_a$) через цей коефіцієнт. Хоча точна формула може варіюватися залежно від контексту розрахунку, пірометричний коефіцієнт по суті відображає ефективність використання теплоти згоряння.

3.  **Роль випромінювання:** Продукти згоряння, особливо такі гази, як $CO_2$ та $H_2O$ (водяна пара), а також частинки сажі (якщо вони утворюються), є активними випромінювачами в інфрачервоному діапазоні при високих температурах. Це випромінювання переносить енергію від гарячих газів до холодніших стінок камери згоряння, теплообмінних поверхонь або навколишнього середовища. Ці втрати енергії випромінюванням безпосередньо впливають на кінцеву температуру продуктів згоряння.

4.  **Розгляд інших варіантів:**
    * **B) Втрати енергії палива теплопровідністю при згорянні:** Теплопровідність всередині самого полум'я або між продуктами згоряння та стінками також є механізмом тепловтрат, але пірометричний коефіцієнт у класичному розумінні частіше асоціюється саме з інтегральними втратами, де випромінювання відіграє домінуючу роль при високих температурах. Втрати через теплопровідність стінок враховуються загальним тепловим балансом установки.
    * **C) Конвективні втрати енергії палива, що згоряє:** Конвективний теплообмін між продуктами згоряння та оточуючими поверхнями також є важливим механізмом тепловтрат. Однак, пірометричний коефіцієнт, як правило, узагальнює вплив основних незворотних втрат, що знижують максимальну температуру, і випромінювання від високотемпературних газів є ключовим серед них.
    * **D) Умови змішування палива з окиснювачем при згорянні:** Умови змішування впливають на повноту згоряння та розподіл температур у зоні горіння, але не є прямою "втратою енергії" в тому сенсі, як це враховує пірометричний коефіцієнт, що коригує теоретичну температуру. Недосконале змішування призведе до неповного згоряння (хімічний недопал), що зменшить кількість виділеної теплоти, а отже, і температуру. Пірометричний коефіцієнт же більше стосується тепловтрат вже виділеної теплоти.

---

**Висновок:** Пірометричний коефіцієнт в процесі горіння палива призначений для корекції теоретичної температури горіння з урахуванням реальних тепловтрат, серед яких **втрати енергії випромінюванням продуктами згоряння** є одними з найважливіших при високих температурах.

Для встановлення правильного порядку температур горіння палива за зростанням потрібно розуміти сутність кожного терміну та фактори, які впливають на їх значення.

1. **Дійсна температура (tₐ)**
   Це температура, яка реалізується при справжньому горінні палива, враховуючи **втрати теплоти** (Q<sub>втр</sub>) у зоні горіння та **дисоціацію продуктів згоряння** (Q<sub>дис</sub>). Формула:
   $$
   t_{\text{д}} = \frac{Q_{\text{н}}^{\text{п}} + Q_{\text{ф}} - Q_{\text{дис}} - Q_{\text{втр}}}{V_{\text{д}} \cdot C_{\text{д}}}
   $$
   Вона є **найнижчою**, оскільки враховує всі фактичні втрати та процеси, що зменшують ефективність конверсії теплоти.

2. **Жаропродуктивність (t<sub>ж</sub>)**
   Розрахункова температура, яка враховує лише теплоту, виділену при горінні палива (**Q<sub>н</sub><sup>п</sup>**), без урахування теплоти, постаченої через підігріття повітря та палива (Q<sub>ф</sub>). Формула:
   $$
   t_{\text{ж}} = \frac{Q_{\text{н}}^{\text{п}}}{V_{\text{д}} \cdot C_{\text{д}}}
   $$
   Вона вища за дійсну, оскільки не враховуються втрати теплоти, але нижча за теоретичну, оскільки не враховується дисоціація.

3. **Теоретична температура (t<sub>т</sub>)**
   Розрахункова температура, яка враховує **всю теплоту горіння** (Q<sub>н</sub><sup>п</sup> + Q<sub>ф</sub>), але **захистає дисоціацію продуктів згоряння** (Q<sub>дис</sub>). Формула:
   $$
   t_{\text{т}} = \frac{Q_{\text{н}}^{\text{п}} + Q_{\text{ф}} - Q_{\text{дис}}}{V_{\text{д}} \cdot C_{\text{д}}}
   $$
   Вона вища за жаропродуктивність, оскільки враховує теплоту, постачену через підігріття, але нижча за калориметричну, оскільки включає втрати на дисоціацію.

4. **Калориметрична температура (t<sub>к</sub>)**
   Розрахункова температура, яка **не враховує дисоціацію** та втрати теплоти, припускаючи, що вся теплота горіння (Q<sub>н</sub><sup>п</sup> + Q<sub>ф</sub>) йде на нагрівання продуктів згоряння. Формула:
   $$
   t_{\text{к}} = \frac{Q_{\text{н}}^{\text{п}} + Q_{\text{ф}}}{V_{\text{д}} \cdot C_{\text{д}}}
   $$
   Вона є **найвищою**, оскільки є ідеальним випадком без будь-яких втрат.

**Порядок зростання:**
**Дійсна < Жаропродуктивність < Теоретична < Калориметрична** (варіант B).

Цей результат відповідає логіці, що ідеальні розрахунки (калориметрична) завжди вищі за реальні умови, а теоретична температура знаходиться між жаропродуктивністю та калориметричною через врахування дисоціації.

# Питання 2-29
Процес массопереносу знаходиться в кінетичній області. Це означає, що такий процес лімітується

Розглянемо питання щодо лімітуючих факторів процесу масопереносу в кінетичній області.

**Теоретичні основи:**

У процесах масопереносу з хімічною реакцією розрізняють три основні області контролю процесу:

1. **Кінетична область** - процес лімітується швидкістю хімічної реакції
2. **Дифузійна область** - процес лімітується масопереносом (дифузією) речовин
3. **Перехідна область** - змішаний контроль (і кінетика, і дифузія впливають на загальну швидкість)

**Розгляд варіанта C - Хімічною реакцією:**

У кінетичній області швидкість загального процесу визначається швидкістю власне хімічної реакції. Це означає, що:

- Підведення реагентів до зони реакції відбувається достатньо швидко
- Відведення продуктів з зони реакції також не створює суттєвих обмежень
- Дифузійні процеси протікають значно швидше порівняно з хімічним перетворенням
- **Лімітуючим фактором є швидкість власне хімічної реакції**

**Математичний опис:**

Загальна швидкість процесу описується рівнянням:
1/W_заг = 1/W_хім + 1/W_диф

У кінетичній області: W_диф >> W_хім, тому W_заг ≈ W_хім

**Критичний аналіз інших варіантів:**

- **Варіант A** (Підведенням реагентів): характерно для дифузійної області, коли масоперенос реагентів до поверхні реакції є повільним
- **Варіант B** (Відведенням продуктів): також відноситься до дифузійних обмежень
- **Варіант D** (Дифузійними процесами): це характеристика дифузійної області, а не кінетичної

**Практичні наслідки:**

У кінетичній області:
- Збільшення поверхні контакту фаз слабо впливає на швидкість процесу
- Підвищення температури значно збільшує швидкість (через експоненційну залежність константи швидкості реакції)
- Перемішування має менший вплив на загальну швидкість

**Висновок:**

Правильна відповідь C обґрунтована тим, що в кінетичній області швидкість хімічної реакції є найповільнішою стадією загального процесу масопереносу, тому саме вона визначає (лімітує) швидкість всього процесу. Дифузійні процеси при цьому протікають достатньо швидко і не створюють суттєвих обмежень.

Правильна відповідь: $\boxed{C}$

**Теоретичні основи:**

При вакуумній обробці рідкого металу створюється різниця тисків, що може призводити до підняття рівня металу в вакуумній камері. Висота підняття визначається рівнянням гідростатики:

h = ΔP/(ρ·g)

де:
- ΔP - різниця тисків
- ρ - густина рідкої сталі (~7000 кг/м³)
- g - прискорення вільного падіння (9,81 м/с²)

**Аналіз схем з графіка:**

З наданого графіка видно залежність потужності перемішування від тиску над ванною для двох варіантів витрати аргону:
- Варіант 1: 60 м³/год
- Варіант 2: 20 м³/год

**Варіант 2 (h₂ = 1,48 м):**

ΔP = P₀ - P_вак ≈ P₀ (при абсолютному вакуумі)
ΔP = 101325 Па

1. **Інтенсивне барботування** (варіант 1) створює турбулентність, яка перешкоджає формуванню стійкого вакууму

- Видалення неметалевих включень
- Контролю хімічного складу

# Питання 2-36
Швидкість спливання газової бульбашки в рідині пропорційна
* A) Радіусу бульбашки,
* B) Глибині, на якій бульбашка утворюється,
* C) Кореню квадратному із радіуса бульбашки,
* D) Логарифму радіуса бульбашки.

---

## Пояснення швидкості спливання бульбашки

Швидкість спливання газової бульбашки в рідині визначається балансом сил, головним чином плавучості, опору та поверхневого натягу. Надане зображення демонструє дві формули для розрахунку цієї швидкості ($u_б$).

Перша, більш загальна формула:

$u_б = \sqrt{\frac{\sigma}{\rho \cdot r_{екв}} + g \cdot r_{екв}}$ , м/с

Де:
* $\sigma$ – поверхневий натяг рідини (Н/м)
* $\rho$ – щільність рідини (кг/м³)
* $r_{екв}$ – еквівалентний радіус бульбашки (м)
* $g$ – прискорення вільного падіння (м/с²)

Друга формула є спрощенням для більших бульбашок (зокрема, для більшості бульбашок у сталевій ванні, де $r_{екв} > 0.5-1 \text{ см}$), коли вплив поверхневого натягу стає менш значущим порівняно із силами плавучості та опору:

$u_б = \sqrt{g \cdot r_{екв}}$ , м/с

---

**Обґрунтування відповіді:**

Розглядаючи спрощену формулу, яка застосовується для більших бульбашок, ми можемо безпосередньо побачити залежність між швидкістю бульбашки ($u_б$) та її еквівалентним радіусом ($r_{екв}$).

$u_б = \sqrt{g \cdot r_{екв}}$

Оскільки $g$ (прискорення вільного падіння) є постійною величиною в цьому контексті, формула показує, що швидкість бульбашки ($u_б$) прямо пропорційна **кореню квадратному з еквівалентного радіуса бульбашки** ($\sqrt{r_{екв}}$).

Отже, правильною відповіддю є **C) Кореню квадратному із радіуса бульбашки**.

Навіть у більш загальній формулі, хоча залежність є складнішою через член поверхневого натягу ($\frac{\sigma}{\rho \cdot r_{екв}}$), для бульбашок значного розміру, де домінує член $g \cdot r_{екв}$ (як це випливає з умови для спрощеної формули), швидкість все одно буде найсильніше залежати від квадратного кореня радіуса. Питання передбачає загальну пропорційність, і домінуючий фактор для бульбашок достатнього розміру призводить до цього висновку.

# Питання 2-37
Яка ціль не реалізується при допаленні СО, що відходить зі сталеплавильного агрегату
* A) Підвищення продуктивності,
* B) Покращення екології,
* C) Утилізація теплоти відхідних газів та підвищення енергоефективності.

---
Допалення монооксиду вуглецю (СО), що утворюється в процесі сталеваріння, є важливою технологічною операцією, спрямованою на досягнення кількох ключових цілей. Розглянемо кожну з опцій:

* **B) Покращення екології:** Монооксид вуглецю (СО) є токсичною речовиною та забруднювачем атмосфери. Його допалення до діоксиду вуглецю (СО2) (хімічна реакція: $2\text{CO} + \text{O}_2 \rightarrow 2\text{CO}_2$) значно знижує шкідливі викиди в атмосферу, тим самим **покращуючи екологічну ситуацію**. * **C) Утилізація теплоти відхідних газів та підвищення енергоефективності:** Реакція окиснення СО до СО2 є екзотермічною, тобто відбувається з виділенням значної кількості теплоти. Ця теплота може бути утилізована (наприклад, для підігріву шихти, генерації пари, підігріву повітря для дуття тощо), що **підвищує загальну енергоефективність** сталеплавильного процесу. 🔥 Таким чином, відбувається утилізація теплової енергії, що міститься у відхідних газах.

* **A) Підвищення продуктивності:** Хоча утилізація теплоти може опосередковано впливати на економіку виробництва (наприклад, через зниження витрат на енергоносії), **прямого та безпосереднього впливу на продуктивність** сталеплавильного агрегату (тобто на кількість виплавленої сталі за одиницю часу) процес допалення СО зазвичай не чинить. Продуктивність агрегату визначається іншими факторами, такими як потужність трансформатора (для дугових печей), інтенсивність подачі кисню (для конвертерів), швидкість завантаження та розвантаження, організація виробничих процесів тощо. Допалення СО – це, перш за все, енергозберігаючий та екологічний захід. ⚙️

**Висновок:** Допалення СО спрямоване на покращення екологічних показників та підвищення енергоефективності за рахунок утилізації теплоти. Безпосереднє підвищення продуктивності сталеплавильного агрегату не є основною та прямою ціллю цього процесу.

# Питання 2-38
Яка, с позицій використання вторинних енергетичних ресурсів, причина впровадження відведення газового середовища із сталеплавильного кисневого конвертеру без допалення СО.
* A) Низька стійкість газоходів через високу температуру газу при допаленні СО,
* B) Зменшення викидів у атмосферу парникових газів,
* C) Накопичення газової суміші для подальшого використання в якості палива,
* D) Низька ефективність допалення СО в газоходах.

---
Киснево-конвертерний процес супроводжується утворенням великої кількості відхідних газів, що містять значну частку монооксиду вуглецю (СО). Цей газ має високу теплотворну здатність, що робить його цінним **вторинним енергетичним ресурсом**.

* **C) Накопичення газової суміші для подальшого використання в якості палива:** Це є ключовою причиною з точки зору використання вторинних енергоресурсів. 🔥 Відведення конвертерних газів **без допалення СО** (тобто без його окиснення до СО2 в системі газовідведення конвертера) дозволяє зібрати цей СО-вмісний газ. Після очищення його можна використовувати як паливо на інших дільницях металургійного підприємства (наприклад, для опалення печей, виробництва пари, в енергетичних установках), заміщуючи природний газ або інші види палива. Це прямий шлях до утилізації хімічної енергії СО.

Розглянемо інші варіанти:

* **A) Низька стійкість газоходів через високу температуру газу при допаленні СО:** Хоча високі температури при допаленні СО (реакція $2\text{CO} + \text{O}_2 \rightarrow 2\text{CO}_2$ є сильно екзотермічною) дійсно створюють проблеми для матеріалів газоходів та вимагають складних систем охолодження, це є радше технологічною трудністю, яку потрібно вирішувати, а не основною причиною відмови від допалення на користь збору газу як палива. Якби головною метою було лише збереження газоходів, можна було б інтенсивно охолоджувати продукти згоряння.

* **B) Зменшення викидів у атмосферу парникових газів:** Якщо СО збирається і потім спалюється як паливо, він все одно перетворюється на СО2 (парниковий газ). Прямого значного зменшення загальних викидів СО2 лише за рахунок відмови від допалення в конвертерному газоході не відбувається. Однак, якщо зібраний конвертерний газ заміщує інше, більш "брудне" викопне паливо, то опосередковано це може позитивно вплинути на загальний баланс викидів. Але первинна мета в контексті питання – енергетична утилізація. 🌍

* **D) Низька ефективність допалення СО в газоходах:** Сучасні системи допалення СО в газоходах конвертерів (якщо вони застосовуються з метою утилізації тепла на місці, наприклад, у котлах-утилізаторах) можуть бути досить ефективними. Рішення не допалювати СО для подальшого використання газу як палива приймається не через низьку ефективність самого процесу допалення, а через бажання використати його хімічну енергію більш раціонально в інших місцях.

**Висновок:** З позицій використання вторинних енергетичних ресурсів, головна причина відведення конвертерного газу без допалення СО полягає в можливості **накопичення цього газу та його подальшого використання як хімічного палива**, що дозволяє ефективно утилізувати його значний енергетичний потенціал.

# Питання 2-40
Яке з формулювань є законом Бернуллі для ідеального середовища, що рухається в каналі.
* A) Сума атмосферного, статичного та динамічного тиску є величина постійна,
* B) Сума статичного, динамічного та геометричного тиску є величина постійна,
* C) Сума статичного, динамічного та капілярного тиску є величина постійна,
* D) Сума статичного, динамічного та геометричного тиску є величина постійна за відсутністю джерел та стоків речовини.

---
Рівняння Бернуллі є фундаментальним законом гідродинаміки, що описує стаціонарний потік ідеальної (нев'язкої та нестисливої) рідини або газу. Воно є наслідком закону збереження енергії для рухомої рідини.

### Рівняння Бернуллі

Для ідеального середовища, що рухається в каналі, рівняння Бернуллі вздовж лінії току має вигляд:

$$P + \frac{\rho v^2}{2} + \rho g h = \text{const}$$

де:
* $P$ – **статичний тиск**. Це тиск, який чинить рідина на стінки каналу або на поверхню тіла, зануреного в потік, коли рідина перебуває у відносному спокої відносно цих поверхонь.
* $\frac{\rho v^2}{2}$ – **динамічний тиск**. Він представляє кінетичну енергію одиниці об'єму рідини, де $\rho$ – густина середовища, а $v$ – швидкість потоку.
* $\rho g h$ – **геометричний (або гідростатичний) тиск**. Він представляє потенційну енергію одиниці об'єму рідини, зумовлену її положенням у полі тяжіння, де $g$ – прискорення вільного падіння, а $h$ – висота (геометрична висота) над деяким обраним нульовим рівнем.

![img_11.png](img_11.png)
### Обґрунтування відповіді

**Відповідь D) "Сума статичного, динамічного та геометричного тиску є величина постійна за відсутністю джерел та стоків речовини"** є найбільш повним та коректним формулюванням закону Бернуллі для зазначених умов.

Розглянемо інші варіанти:

* **A) Сума атмосферного, статичного та динамічного тиску є величина постійна.** Атмосферний тиск є частиною статичного тиску, якщо система відкрита до атмосфери, але це не окремий доданок у загальному рівнянні Бернуллі, і ця опція пропускає геометричний тиск.
* **B) Сума статичного, динамічного та геометричного тиску є величина постійна.** Це формулювання близьке до правильного, однак воно не враховує важливе обмеження застосовності рівняння Бернуллі.
* **C) Сума статичного, динамічного та капілярного тиску є величина постійна.** Капілярний тиск пов'язаний з поверхневим натягом і не є складовою класичного рівняння Бернуллі для макроскопічних потоків у каналах.

**Ключовим аспектом у відповіді D є умова "за відсутністю джерел та стоків речовини".** Це означає, що в розглядуваній ділянці потоку маса рідини зберігається, тобто немає додаткового надходження або відведення рідини. Рівняння Бернуллі виводиться з припущення про неперервність потоку та збереження енергії для елементарної маси рідини, що рухається вздовж лінії току. Наявність джерел або стоків порушила б ці умови.

Таким чином, варіант D найточніше відображає суть закону Бернуллі для ідеального середовища, що рухається в каналі, включаючи необхідні умови його застосування.

---
Правильна відповідь: $\boxed{D}$

# Питання 2-41
Який із перелічених чинників негативно впливає на коефіцієнт використання палива $η_{квп}$ у нагрівальній печі.
* A) Збільшення теплоти згоряння палива,
* B) Підігрів палива та окислювача,
* C) Зменшення теплового опору футерівки печі,
* D) Збагачення киснем повітря на пальник.

---

**Коефіцієнт використання палива ($η_{квп}$)**, також відомий як коефіцієнт корисної дії печі (ККД печі) за теплотою, визначається як відношення кількості теплоти, корисно використаної на нагрівання матеріалу (або на технологічний процес), до загальної кількості теплоти, що виділилася при згорянні палива.

$$η_{квп} = \frac{Q_{корисна}}{Q_{палива}}$$

Чинник, що негативно впливає на $η_{квп}$, призводить до його зменшення, тобто до менш ефективного використання палива.
![img_12.png](img_12.png)

### Обґрунтування відповіді

**C) Зменшення теплового опору футерівки печі:**

* **Футерівка печі** – це внутрішній шар вогнетривких та теплоізоляційних матеріалів, що обмежують робочий простір печі. Її основне призначення – мінімізувати втрати теплоти в навколишнє середовище через стінки печі.
* **Тепловий опір ($R_{th}$)** – це здатність матеріалу чи конструкції перешкоджати проходженню теплового потоку. Чим вищий тепловий опір футерівки, тим краще вона ізолює піч і тим менші втрати теплоти.
* **Зменшення теплового опору футерівки** означає, що футерівка гірше ізолює тепло. Це призводить до **збільшення втрат теплоти ($Q_{втрат}$) через стінки печі** в навколишнє середовище.
* Оскільки корисна теплота ($Q_{корисна}$) є різницею між теплотою, що виділилася при згорянні палива ($Q_{палива}$), та сумою всіх втрат теплоти (втрати з відхідними газами, втрати через футерівку, тощо), збільшення втрат через футерівку безпосередньо зменшує $Q_{корисна}$.
* Як наслідок, при незмінній $Q_{палива}$ та збільшенні втрат, коефіцієнт використання палива ηквп **зменшується**.

Отже, зменшення теплового опору футерівки є чинником, що **негативно впливає** на коефіцієнт використання палива.

### Чому інші варіанти не є правильними:

* **A) Збільшення теплоти згоряння палива:**
    Збільшення теплоти згоряння палива означає, що при спалюванні тієї ж кількості палива виділяється більше енергії. За інших рівних умов, це може призвести до **збільшення** ηквп, оскільки потенційно більша кількість теплоти може бути передана нагрівальному матеріалу, або ж можна використати менше палива для отримання тієї ж кількості корисної теплоти.

* **B) Підігрів палива та окислювача:**
    Підігрів палива та окислювача (зазвичай повітря) перед подачею в зону горіння, особливо за рахунок утилізації теплоти відхідних газів (рекуперація, регенерація), дозволяє повернути частину втраченої теплоти в процес. Це призводить до підвищення температури горіння, поліпшення умов теплообміну та, як результат, до **збільшення** ηквп.

* **D) Збагачення киснем повітря на пальник:**
    Збагачення повітря киснем зменшує кількість азоту (інертного баласту), який нагрівається в процесі горіння і виноситься з відхідними газами. Це призводить до підвищення температури полум'я, зменшення об'єму продуктів згоряння та зменшення втрат теплоти з відхідними газами. Внаслідок цього ηквп **збільшується**.

Таким чином, єдиним чинником з перелічених, що однозначно негативно впливає на коефіцієнт використання палива, є зменшення теплового опору футерівки печі.

# Питання 2-42
Який тепловий режим найбільш ефективно забезпечує нейтралізацію токсичних PCDD/F, що виникають при нагріванні скрапу.
* A) Нагрів до 1800°С та різке охолодження до температури <200°С,
* B) Витримка при температурі > 1600°С впродовж певного часу,
* C) Введення сорбентів,
* D) Комплекс варіантів А,В,
* E) Комплекс варіантів Б,В.

---

Питання стосується ефективних методів термічної нейтралізації (руйнування) поліхлорованих дибензо-п-діоксинів (PCDD) та поліхлорованих дибензофуранів (PCDF), які є стійкими органічними забруднювачами (СОЗ) і можуть утворюватися при нагріванні металевого брухту, особливо якщо він містить хлоровмісні сполуки та органічні домішки.

### Механізми утворення та руйнування PCDD/F

* **Утворення:** PCDD/F можуть утворюватися в діапазоні температур приблизно 200-650°C (так зване вікно "de novo синтезу"), з піком утворення близько 300-400°C, за наявності попередників (наприклад, хлорфенолів, хлорбензолів), кисню та каталізаторів (наприклад, міді).
* **Руйнування:** Для ефективного термічного руйнування цих стійких сполук потрібні значно вищі температури. Загальноприйнято, що температури вище 850°C з достатнім часом перебування забезпечують їх розклад. Чим вища температура, тим швидше відбувається руйнування.

### Аналіз варіантів відповіді

* **A) Нагрів до 1800°С та різке охолодження до температури <200°С:**
    * **Нагрів до 1800°С:** Така надзвичайно висока температура гарантує швидке та практично повне термічне руйнування молекул PCDD/F.
    * **Різке охолодження до температури <200°С (квенчинг):** Це критично важливий крок. Після високотемпературного руйнування необхідно швидко пройти температурне вікно (200-650°C), де можливе повторне утворення PCDD/F (de novo синтез) з продуктів неповного згоряння або попередників, що можуть бути присутні у відхідних газах. Охолодження нижче 200°C мінімізує ризик такого повторного утворення.

* **B) Витримка при температурі > 1600°С впродовж певного часу:**
    * **Витримка при температурі > 1600°С:** Підтримання газів при такій високій температурі протягом певного часу (час перебування) забезпечує повноту реакцій розкладу PCDD/F. Чим вища температура, тим менший необхідний час перебування для досягнення того ж ступеня руйнування. 1600°C – це температура, що значно перевищує мінімально необхідну, забезпечуючи високий ступінь деструкції.

* **C) Введення сорбентів:**
    * Введення сорбентів (наприклад, активованого вугілля) у потік відхідних газів є ефективним методом *адсорбції* PCDD/F, що вже утворилися або не повністю зруйнувалися. Це метод очищення газів, а не термічний режим їх нейтралізації. Зазвичай сорбенти вводяться при нижчих температурах (наприклад, <200°C) після основного процесу горіння або термічної обробки.

* **D) Комплекс варіантів А,В:**
    * Цей варіант об'єднує переваги обох підходів:
        1.  Досягнення дуже високої температури (1600-1800°C) для ефективного руйнування PCDD/F.
        2.  Забезпечення достатнього часу перебування при цій високій температурі (з варіанту В) для повноти процесу.
        3.  Подальше різке охолодження газів (з варіанту А) для запобігання їх повторному утворенню.
    * Такий комплексний підхід є найбільш надійним та ефективним термічним методом для нейтралізації PCDD/F.

* **E) Комплекс варіантів Б,В:**
    * Цей варіант поєднує високотемпературну витримку (руйнування) із сорбційною очисткою. Хоча це ефективна стратегія контролю викидів PCDD/F, питання фокусується саме на *тепловому режимі*, що забезпечує *нейтралізацію* (тобто руйнування). Сорбція – це видалення, а не руйнування на молекулярному рівні внаслідок теплового впливу.

### Обґрунтування відповіді

**Відповідь D) "Комплекс варіантів А,В"** є найбільш коректною, оскільки вона поєднує ключові аспекти ефективного термічного руйнування PCDD/F:

1.  **Досягнення екстремально високих температур (1600-1800°C):** Це забезпечує високу швидкість та повноту розкладу стійких молекул PCDD/F.
2.  **Достатній час перебування (витримка):** Гарантує, що всі молекули встигнуть зазнати термічного впливу та зруйнуватися.
3.  **Різке охолодження (квенчинг):** Мінімізує можливість повторного утворення PCDD/F у більш холодному газовому тракті.

Цей комплексний підхід забезпечує як максимальне руйнування існуючих PCDD/F, так і запобігання їхньому de novo синтезу, що робить його найбільш ефективним *тепловим режимом* нейтралізації.

---
Правильна відповідь: $\boxed{D}$

# Питання 2-43
Яким фактором, перш за все, обумовлена тяга димової труби
* A) Швидкістю руху димових газів,
* B) Силою тяжіння Землі,
* C) Різницею щільності димових газів і навколишнього середовища у основи труби,
* D) Висотою труби.

---
**Тяга димової труби** (або природна тяга) – це різниця тисків, яка виникає між основою димової труби (або входом у неї) та її виходом нагорі, що змушує гази рухатися вгору через трубу. Цей ефект є ключовим для функціонування багатьох систем спалювання, забезпечуючи видалення продуктів згоряння та підведення свіжого повітря.

### Обґрунтування відповіді

**B) Силою тяжіння Землі:**

Природна тяга димової труби є проявом **явища плавучості (або Архімедової сили)** в газовому середовищі, яке, в свою чергу, безпосередньо зумовлене **силою тяжіння Землі**. Ось як це працює:

1.  **Різниця густин:** Гарячі димові гази всередині труби мають значно нижчу густину ($\rho_{газ}$), ніж холодне навколишнє повітря ($\rho_{пов}$). Це відбувається через теплове розширення газів при нагріванні.
2.  **Дія сили тяжіння:** Сила тяжіння діє на всі маси, включаючи стовп гарячих газів у трубі та еквівалентний за висотою стовп холодного повітря ззовні. Через меншу густину, стовп гарячих газів всередині труби має меншу вагу ($m_{газ} \cdot g$), ніж стовп холодного повітря такої ж висоти ($m_{пов} \cdot g$).
3.  **Виникнення різниці тисків:** Ця різниця у вазі стовпів газів (на одиницю площі) створює різницю тисків біля основи труби. Тиск стовпа холодного зовнішнього повітря на рівні основи труби виявляється більшим, ніж тиск стовпа гарячих газів всередині. Ця різниця тисків і є тягою:
    $\Delta P = (\rho_{пов} - \rho_{газ}) \cdot g \cdot H$
    де:
    * $\Delta P$ – створювана тяга (різниця тисків),
    * $\rho_{пов}$ – густина навколишнього повітря,
    * $\rho_{газ}$ – середня густина димових газів у трубі,
    * $g$ – прискорення вільного падіння (характеристика сили тяжіння Землі),
    * $H$ – висота труби.

**Ключова роль сили тяжіння:** Якби не було сили тяжіння ($g=0$), то поняття "ваги" та "густини" не призводили б до виникнення плавучості чи різниці тисків. Саме гравітація змушує більш щільне холодне повітря "виштовхувати" менш щільні гарячі гази вгору. Таким чином, сила тяжіння Землі є **першопричиною**, що уможливлює сам механізм природної тяги.

### Чому інші варіанти менш первинні:

* **A) Швидкістю руху димових газів:** Швидкість руху газів є **наслідком** тяги, а не її причиною. Більша тяга призведе до більшої швидкості.
* **C) Різницею щільності димових газів і навколишнього середовища у основи труби:** Різниця густин є **необхідною умовою** для виникнення тяги, оскільки вона визначає, наскільки "легшими" будуть гази в трубі порівняно з навколишнім повітрям. Однак ця різниця густин створює тягу лише *завдяки* дії сили тяжіння. Без гравітації різниця густин не мала б значення для вертикального руху.
* **D) Висотою труби:** Висота труби є важливим фактором, що **впливає на величину** тяги (чим вища труба, тим більша тяга за інших рівних умов). Проте висота лише масштабує ефект, який виникає завдяки різниці густин та силі тяжіння.

Отже, хоча різниця густин та висота труби є критично важливими параметрами для розрахунку величини тяги, **фундаментальним фактором, що обумовлює саму можливість виникнення природної тяги, є сила тяжіння Землі.**

---
Правильна відповідь: $\boxed{B}$

# Питання 2-44
Які з перелічених вторинних енергетичних ресурсів (ВЕР) не відносяться, за класифікацією, до теплових ВЕР.
* A) Відхідні гази нагрівальних печей,
* B) Доменний газ,
* C) Конвертерний шлак,
* D) Рідкий чавун

---
**Вторинні енергетичні ресурси (ВЕР)** – це енергетичний потенціал продукції, відходів, побічних і проміжних продуктів, який утворюється в технологічних агрегатах (установках, процесах) і не використовується безпосередньо в самому агрегаті, але може бути частково або повністю використаний для енергопостачання інших агрегатів (процесів).

ВЕР класифікують за видами енергії, що міститься в них, на:
1.  **Теплові (або теплові відходи):** Це фізична теплота відхідних газів, основної та побічної продукції, відходів виробництва, гарячої води, пари тощо. Енергія вилучається шляхом охолодження цих носіїв.
2.  **Горючі (або паливні):** Це хімічно зв'язана енергія горючих газів, рідких та твердих відходів виробництва, які можуть бути використані як паливо.
3.  **Надлишкового тиску (або барометричні):** Це потенційна енергія газів, що виходять з технологічних установок з надлишковим тиском, який може бути використаний в утилізаційних турбінах.

Тепер проаналізуємо запропоновані варіанти:

* **A) Відхідні гази нагрівальних печей:**
    Ці гази мають високу температуру і їхня енергія – це, перш за все, фізична теплота (ентальпія). Їх використовують для підігріву повітря, палива, садки в рекуператорах або регенераторах, або для виробництва пари. Це класичний приклад **теплових ВЕР**.

* **B) Доменний газ:**
    Доменний газ є побічним продуктом виплавки чавуну в доменних печах. Він містить значну кількість горючих компонентів, головним чином оксид вуглецю (CO) та невелику кількість водню (H₂), а також негорючі гази (азот N₂, діоксид вуглецю CO₂). Хоча доменний газ виходить з печі гарячим (і має певну фізичну теплоту), його основна цінність полягає саме в **хімічній енергії горючих складових**. Його використовують як паливо для нагрівання повітронагрівачів (кауперів), парових котлів, промислових печей та в енергетичних установках. За класифікацією, доменний газ є, в першу чергу, **горючим (паливним) ВЕР**.

* **C) Конвертерний шлак:**
    Рідкий шлак, що утворюється в кисневих конвертерах при виробництві сталі, має дуже високу температуру (1500-1700°C). Його енергетичний потенціал – це фізична теплота. Існують технології утилізації цієї теплоти (наприклад, суха грануляція з відбором тепла повітрям). Це **тепловий ВЕР**.

* **D) Рідкий чавун:**
    Рідкий чавун, що виходить з доменної печі, має високу температуру (близько 1350-1500°C). Його фізична теплота (ентальпія) є значним енергетичним ресурсом. Частково ця теплота використовується в наступному переділі (сталеплавильному), але будь-які втрати цієї теплоти або спеціальні заходи з її утилізації (якщо б він розглядався як побічний продукт з точки зору енерговикористання) класифікували б його як **тепловий ВЕР**.

**Висновок:**

З перелічених варіантів **доменний газ (B)** виділяється тим, що його основна енергетична цінність полягає в його горючості (хімічній енергії), а не лише у фізичній теплоті. Тому за стандартною класифікацією ВЕР, він відноситься до групи горючих (паливних) ВЕР, а не суто теплових. Інші ж варіанти (відхідні гази, конвертерний шлак, теплота рідкого чавуну) є переважно носіями фізичної теплоти і класифікуються як теплові ВЕР.

---
Правильна відповідь: $\boxed{B}$

# Питання 2-45
Якому з критеріїв, що характеризують процеси теплообміну відповідає по змісту критерій Шервуда, що описує процеси масообміну.
* A) Рейнольдса,
* B) Прандтля,
* C) Фур'є,
* D) Нусельта.
ANSWER: D

---

#### Критерій Шервуда (Sh)
Критерій **Шервуда** (Sh), також відомий як масообмінний критерій Нуссельта, є безрозмірнісною величиною, що використовується в аналізі процесів масообміну. Він характеризує співвідношення між інтенсивністю конвективного масопереносу та інтенсивністю молекулярного масопереносу (дифузії). Математично він виражається як:

$$Sh = \frac{\beta \cdot L}{D}$$

де:
* $\beta$ – коефіцієнт масовіддачі (м/с),
* $L$ – характерний лінійний розмір (м),
* $D$ – коефіцієнт молекулярної дифузії (м²/с).

Отже, критерій Шервуда показує, у скільки разів конвективний масоперенос інтенсивніший за молекулярну дифузію.

---

#### Критерій Нуссельта (Nu)
Критерій **Нуссельта** (Nu) є безрозмірнісною величиною, що використовується в аналізі процесів теплообміну. Він характеризує співвідношення між інтенсивністю конвективного теплообміну та інтенсивністю молекулярного теплообміну (теплопровідності) через шар рідини або газу. Математично він виражається як:

$Nu = \frac{\alpha \cdot L}{\lambda}$

де:
* $\alpha$ – коефіцієнт тепловіддачі (Вт/(м²·К)),
* $L$ – характерний лінійний розмір (м),
* $\lambda$ – коефіцієнт теплопровідності середовища (Вт/(м·К)).

Отже, критерій Нуссельта показує, у скільки разів конвективний теплообмін інтенсивніший за теплопровідність.

### Аналогія між критеріями Шервуда та Нуссельта

Як видно з визначень та математичних виразів, критерій **Шервуда** в масообміні є прямим аналогом критерію **Нуссельта** в теплообміні. Обидва критерії:

1.  Є безрозмірнісними.
2.  Характеризують відношення інтенсивності конвективного переносу (маси або тепла) до інтенсивності молекулярного переносу (дифузії або теплопровідності).
3.  Мають схожу математичну структуру, де в чисельнику стоїть добуток коефіцієнта переносу (масовіддачі або тепловіддачі) на характерний розмір, а в знаменнику – коефіцієнт молекулярного переносу (дифузії або теплопровідності).

### Розгляд інших варіантів

* **A) Критерій Рейнольдса (Re)**:
    $Re = \frac{\rho \cdot v \cdot L}{\mu}$
    Визначає режим руху рідини або газу (ламінарний або турбулентний) і характеризує співвідношення сил інерції до сил в'язкості. Він є важливим як для теплообміну, так і для масообміну, але не є прямим аналогом Шервуда за фізичним змістом інтенсивності переносу.

* **B) Критерій Прандтля (Pr)**:
    $Pr = \frac{\nu}{a} = \frac{\mu \cdot c_p}{\lambda}$
    Характеризує співвідношення між кінематичною в'язкістю (імпульсопереносом) та температуропровідністю (теплопереносом). Він описує фізичні властивості самого середовища і є специфічним для теплових процесів. Його аналогом у масообмінних процесах є критерій Шмідта (Sc).

* **C) Критерій Фур'є (Fo)**:
    $Fo = \frac{a \cdot \tau}{L^2}$
    Характеризує безрозмірнісний час у нестаціонарних процесах теплопровідності (або дифузії, якщо використовується коефіцієнт дифузії замість температуропровідності). Він описує відносну швидкість зміни температурного поля (або концентраційного поля) в часі.

---

### Висновок

Таким чином, за своїм фізичним змістом та математичним вираженням критерій **Шервуда (Sh)**, що описує процеси масообміну, є аналогом критерію **Нуссельта (Nu)**, що характеризує процеси теплообміну. Обидва критерії показують відносну роль конвективного переносу порівняно з молекулярним переносом відповідної субстанції (маси або теплоти).
